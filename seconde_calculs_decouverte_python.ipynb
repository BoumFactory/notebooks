## üìö Notion 1 : Python comme calculatrice

### üìñ Cours

Python utilise les op√©rateurs suivants :

| Op√©rateur | Op√©ration | Exemple | R√©sultat |
|-----------|-----------|---------|----------|
| `+` | Addition | `5 + 3` | 8 |
| `-` | Soustraction | `10 - 7` | 3 |
| `*` | Multiplication | `6 * 4` | 24 |
| `/` | Division | `15 / 3` | 5.0 |
| `**` | Puissance | `2 ** 5` | 32 |
| `//` | Quotient (division euclidienne) | `17 // 5` | 3 |
| `%` | Reste (modulo) | `17 % 5` | 2 |

**Priorit√©s** : Comme en maths ! Parenth√®ses > Puissances > √ó √∑ > + ‚àí

**Exemples :**

````python
# Les 4 op√©rations de base
print("Addition :", 15 + 23)
print("Soustraction :", 50 - 18)
print("Multiplication :", 7 * 9)
print("Division :", 64 / 8)
````

````python
# Puissances
print("3 au carr√© :", 3 ** 2)
print("2 puissance 10 :", 2 ** 10)
````

````python
# Division euclidienne de 17 par 5
print("Quotient :", 17 // 5)   # 3
print("Reste :", 17 % 5)       # 2
# V√©rification : 17 = 5 √ó 3 + 2
````

````python
# Priorit√©s op√©ratoires
print("Sans parenth√®ses : 3 + 5 * 2 =", 3 + 5 * 2)   # 13
print("Avec parenth√®ses : (3 + 5) * 2 =", (3 + 5) * 2) # 16
````

### üìñ Comprendre print() et les guillemets

La fonction `print()` permet d'**afficher** des r√©sultats √† l'√©cran.

**R√®gle importante** :
- **Texte** ‚Üí entre guillemets `"..."` : Python affiche le texte tel quel
- **Calcul** ‚Üí sans guillemets : Python **calcule** d'abord, puis affiche le r√©sultat

**Pourquoi ?** Les guillemets indiquent √† Python : "ceci est du texte, ne cherche pas √† le calculer".

**Exemples :**

````python
# Avec guillemets : Python affiche le texte tel quel
print("5 + 3")   # Affiche : 5 + 3

# Sans guillemets : Python calcule d'abord
print(5 + 3)     # Affiche : 8

# On peut m√©langer texte et calcul (s√©par√©s par des virgules)
print("La somme de 5 et 3 est", 5 + 3)
````

## üìö Notion 2 : Les variables

### üìñ Cours

Une **variable** est un nom qui stocke une valeur en m√©moire. C'est comme une bo√Æte √©tiquet√©e.

**Cr√©er une variable** (affectation) :
```python
nom_variable = valeur
```

**R√®gles de nommage** :
- Lettres, chiffres, underscores `_`
- Ne commence **pas** par un chiffre
- Pas d'espaces ni d'accents
- Sensible √† la casse (`age` ‚â† `Age`)

**Exemples :**

````python
# Cr√©er des variables
age = 16
prix = 19.99
nom = "Alice"

print(age)
print(prix)
print(nom)
````

````python
# Utiliser des variables dans des calculs
longueur = 5
largeur = 3

aire = longueur * largeur
perimetre = 2 * (longueur + largeur)

print("Aire :", aire)
print("P√©rim√®tre :", perimetre)
````

````python
# Modifier une variable
compteur = 10
print("Avant :", compteur)

compteur = compteur + 5  # On peut aussi √©crire : compteur += 5
print("Apr√®s :", compteur)
````

## üìö Notion 3 : Fonctions math√©matiques utiles

### üìñ Cours

Python dispose de fonctions int√©gr√©es pratiques :

| Fonction | Description | Exemple |
|----------|-------------|---------|
| `abs(x)` | Valeur absolue | `abs(-15)` ‚Üí 15 |
| `round(x)` | Arrondi | `round(3.7)` ‚Üí 4 |
| `round(x, n)` | Arrondi √† n d√©cimales | `round(3.14159, 2)` ‚Üí 3.14 |

**Exemples :**

````python
# Valeur absolue : distance √† z√©ro
print(abs(-15))     # 15
print(abs(8 - 15))  # 7 (distance entre 8 et 15)
````

````python
# Arrondis
print(round(3.7))        # 4
print(round(3.14159, 2)) # 3.14
````
