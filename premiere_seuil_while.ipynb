## ğŸ“š Notion 1 : Syntaxe de la boucle while

### ğŸ“– Cours

Une boucle `while` rÃ©pÃ¨te des instructions **tant qu'une condition est vraie**.

**Syntaxe :**

**DiffÃ©rence avec `for` :**

| `for` | `while` |

## ğŸ“š Notion 2 : Seuil d'une suite croissante

### ğŸ“– Cours

Un **seuil** est le premier rang $n$ Ã  partir duquel une suite vÃ©rifie une condition.

**MÃ©thode pour une suite croissante :**
1. Initialiser le terme $u$ et le rang $n$
2. Tant que la condition n'est **pas** atteinte, calculer le terme suivant
3. IncrÃ©menter le rang Ã  chaque itÃ©ration

**Structure type :**
```python

n = 0        # Rang initial

while u <= seuil:     # Tant que NON atteint
    u = f(u)          # Terme suivant
    n = n + 1         # Rang suivant

print(f"Seuil atteint au rang {n}")
```

````python
# Exemple : Suite u_0 = 1, u_{n+1} = 2*u_n
# Trouver le premier rang oÃ¹ u_n > 100

u = 1   # u_0
n = 0

while u <= ...:       # ComplÃ©tez la condition
    u = ...           # ComplÃ©tez : terme suivant
    n = ...           # ComplÃ©tez : rang suivant

print(f"Premier terme > 100 : u_{n} = {u}")
print(f"Seuil : n = {n}")
````
