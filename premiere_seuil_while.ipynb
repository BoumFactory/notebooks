## üí° Introduction

En math√©matiques, on cherche souvent √† d√©terminer un **seuil** : le premier rang n √† partir duquel une suite v√©rifie une certaine condition.

<div style="background-color: #f0f4c3; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ü§î Probl√®me :</strong> On ne sait pas √† l'avance combien de termes calculer<br>
    <strong>‚ú® Solution :</strong> La boucle <strong>while</strong> permet de r√©p√©ter jusqu'√† ce qu'une condition soit satisfaite !
</div>

**Exemple** : Pour la suite $u_0 = 1$ et $u_{n+1} = 2u_n$, √† partir de quel rang a-t-on $u_n > 100$ ?

On ne peut pas utiliser une boucle `for` car on ne conna√Æt pas le nombre d'it√©rations n√©cessaires.

## üìö Notion 1 : Syntaxe de la boucle while

### üìñ Cours

Une boucle `while` r√©p√®te des instructions **tant qu'une condition est vraie**.

**Syntaxe :**
```python
while condition:
    # Instructions √† r√©p√©ter
    # Faire √©voluer la variable test√©e !
```

**Diff√©rence avec `for` :**
- `for` : nombre d'it√©rations connu √† l'avance
- `while` : on r√©p√®te jusqu'√† satisfaction d'une condition

‚ö†Ô∏è **Attention** : Il faut **toujours** faire √©voluer la variable test√©e, sinon boucle infinie !

**Exemple r√©solu :**

````python
# Compter de 1 √† 5
compteur = 1

while compteur <= 5:
    print(compteur)
    compteur = compteur + 1  # Faire √©voluer la variable !

print("Fin de la boucle")
````

## üìö Notion 2 : D√©terminer un seuil (suite croissante)

### üìñ Cours

Un **seuil** est le premier rang n √† partir duquel une suite v√©rifie une condition.

**M√©thode pour une suite croissante :**
1. Initialiser le terme u et le rang n
2. Tant que la condition n'est pas atteinte, calculer le terme suivant
3. Incr√©menter le rang √† chaque it√©ration
4. Afficher le r√©sultat quand la boucle s'arr√™te

**Exemple r√©solu :** Suite g√©om√©trique $u_0 = 1$, $u_{n+1} = 2u_n$. Trouver le premier rang o√π $u_n > 100$.

````python
# D√©terminer le seuil pour u_n > 100
u = 1    # Premier terme u_0
n = 0    # Rang du terme

# Tant que u <= 100, on continue
while u <= 100:
    u = 2 * u    # Calcul du terme suivant
    n = n + 1    # Rang suivant

print(f"Le premier terme > 100 est u_{n} = {u}")
print(f"Le seuil est n = {n}")
````

## üìö Notion 4 : Condition de proximit√© (convergence)

### üìñ Cours

Certaines suites **convergent** vers une limite L. On cherche souvent le premier rang o√π la suite est "assez proche" de L.

**Condition de proximit√©** : $|u_n - L| < \epsilon$ (epsilon repr√©sente la pr√©cision souhait√©e)

**Valeur absolue en Python** : `abs(u - L)`

**Exemple r√©solu :** Suite $w_0 = 100$, $w_{n+1} = \dfrac{w_n + 10}{2}$. Cette suite converge vers 10. Trouver le premier rang o√π $|w_n - 10| < 0.001$.

````python
# Convergence vers 10 avec pr√©cision 0.001
w = 100
n = 0
L = 10         # Limite
epsilon = 0.001  # Pr√©cision

while abs(w - L) >= epsilon:
    w = (w + 10) / 2
    n = n + 1

print(f"Convergence atteinte √† n = {n}")
print(f"w_{n} = {w}")
print(f"Distance √† la limite : {abs(w - 10)}")
````

## üìö Notion 5 : Applications en contexte

### üìñ Cours

Les boucles while sont tr√®s utiles pour r√©soudre des **probl√®mes concrets** :
- Placement financier (capital avec int√©r√™ts)
- Croissance d'une population
- Sommes partielles

**Principe** : Mod√©liser la situation par une suite, puis d√©terminer le seuil.

**Exemple r√©solu :** Un capital de 1000 ‚Ç¨ est plac√© √† 3% par an. Combien d'ann√©es faut-il pour doubler le capital ?

````python
# Placement avec int√©r√™ts compos√©s
capital = 1000
objectif = 2000  # Doubler
annees = 0
taux = 1.03      # +3% par an

while capital < objectif:
    capital = capital * taux
    annees = annees + 1

print(f"Il faut {annees} ans pour doubler le capital")
print(f"Capital final : {capital:.2f} ‚Ç¨")
````

### Algorithme de Babylone (racine carr√©e)

Pour calculer $\sqrt{a}$, l'algorithme de Babylone utilise la suite :

$$u_0 = 1 \quad \text{et} \quad u_{n+1} = \dfrac{1}{2}\left(u_n + \dfrac{a}{u_n}\right)$$

Cette suite converge vers $\sqrt{a}$.

**Votre mission :**
1. Calculer $\sqrt{50}$ avec une pr√©cision de $10^{-6}$ (c'est-√†-dire $|u_n - \sqrt{50}| < 10^{-6}$)
2. Comparer avec la valeur donn√©e par Python : `50**0.5`
3. Afficher le nombre d'it√©rations n√©cessaires

*Indice : Utilisez `import math` puis `math.sqrt(50)` pour la valeur exacte.*

````python
# √âcrivez votre code ici
````

<details>
<summary>üèÜ Correction du challenge</summary>

```python
import math

# Param√®tres
a = 50
u = 1  # Valeur initiale
n = 0
epsilon = 1e-6  # Pr√©cision
racine_exacte = math.sqrt(a)

# Algorithme de Babylone
while abs(u - racine_exacte) >= epsilon:
    u = 0.5 * (u + a / u)
    n = n + 1
    print(f"It√©ration {n} : u = {u}")

# R√©sultats
print(f"\nConvergence atteinte en {n} it√©rations")
print(f"Approximation : {u}")
print(f"Valeur exacte : {racine_exacte}")
print(f"Erreur : {abs(u - racine_exacte)}")
```

**Explication :**
- L'algorithme converge tr√®s rapidement (environ 5-6 it√©rations)
- La formule $u_{n+1} = \dfrac{1}{2}\left(u_n + \dfrac{a}{u_n}\right)$ est une moyenne entre $u_n$ et $\dfrac{a}{u_n}$
- C'est une m√©thode ancienne mais tr√®s efficace pour calculer des racines carr√©es
