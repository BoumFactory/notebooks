## üìö Notion 1 : D√©finir une suite par r√©currence en Python

### üìñ Cours

Une suite est **d√©finie par r√©currence** quand on conna√Æt :
1. Le **premier terme** $u_0$ (valeur initiale)
2. Une **relation de r√©currence** : $u_{n+1} = f(u_n)$

**M√©thode Python :**
- Cr√©er une liste contenant le premier terme : `u = [u0]`
- Boucle for pour calculer les termes suivants
- Ajouter chaque nouveau terme avec `append()`

**Exemple r√©solu :** $u_0 = 10$, $u_{n+1} = 0.5 \times u_n + 2$

````python
# Calculer les 10 premiers termes
u = [10]  # Liste contenant u_0

# Boucle pour calculer u_1, u_2, ..., u_10
for n in range(10):
    nouveau_terme = 0.5 * u[n] + 2
    u.append(nouveau_terme)

# Affichage
print("Les 11 premiers termes :")
for n in range(len(u)):
    print(f"u_{n} = {u[n]}")
````

## üìö Notion 2 : Suite arithm√©tique

### üìñ Cours

Une **suite arithm√©tique** a une raison constante r. Chaque terme s'obtient en ajoutant r au pr√©c√©dent.

**D√©finition par r√©currence :** $u_{n+1} = u_n + r$

**Formule explicite :** $u_n = u_0 + n \times r$

**En Python** : On peut calculer soit par r√©currence (boucle), soit directement avec la formule explicite.

**Exemple r√©solu :** $u_0 = 7$, raison $r = 4$

````python
# M√©thode 1 : Par r√©currence
u = [7]
r = 4

for n in range(10):
    u.append(u[n] + r)

print("M√©thode r√©currence :", u)

# M√©thode 2 : Formule explicite
u0 = 7
v = [u0 + n*r for n in range(11)]

print("Formule explicite :", v)
````

## üìö Notion 3 : Suite g√©om√©trique

### üìñ Cours

Une **suite g√©om√©trique** a une raison constante q. Chaque terme s'obtient en multipliant le pr√©c√©dent par q.

**D√©finition par r√©currence :** $u_{n+1} = u_n \times q$

**Formule explicite :** $u_n = u_0 \times q^n$

**Propri√©t√©s :**
- Si $q > 1$ : suite croissante (croissance exponentielle)
- Si $0 < q < 1$ : suite d√©croissante (converge vers 0)
- Si $q < 0$ : suite altern√©e

**Exemple r√©solu :** $u_0 = 100$, raison $q = 0.9$

````python
# Suite g√©om√©trique d√©croissante
u = [100]
q = 0.9

for n in range(20):
    u.append(u[n] * q)

# Affichage de quelques termes
print(f"u_0 = {u[0]}")
print(f"u_5 = {u[5]:.2f}")
print(f"u_10 = {u[10]:.2f}")
print(f"u_20 = {u[20]:.2f}")
````

## üìö Notion 4 : Calculer les termes d'une suite avec une fonction

### üìñ Cours

Pour **r√©utiliser** le code, on peut cr√©er une **fonction** qui calcule les termes d'une suite.

**Structure d'une fonction de suite :**
```python
def calculer_suite(u0, nombre_termes):
    u = [u0]
    for n in range(nombre_termes):
        nouveau = f(u[n])  # Formule de r√©currence
        u.append(nouveau)
    return u
```

**Avantages :**
- Code r√©utilisable
- Param√®tres modifiables facilement
- Plus lisible

**Exemple r√©solu :** Fonction pour suite $u_{n+1} = \dfrac{u_n + 6}{2}$

````python
def suite_convergente(u0, n):
    """
    Calcule n termes de la suite u_{n+1} = (u_n + 6) / 2
    qui converge vers 6
    """
    u = [u0]
    for i in range(n):
        nouveau = (u[i] + 6) / 2
        u.append(nouveau)
    return u

# Test avec diff√©rentes valeurs initiales
suite1 = suite_convergente(0, 10)
suite2 = suite_convergente(100, 10)

print("Depuis u_0 = 0 :", suite1[-5:])  # 5 derniers termes
print("Depuis u_0 = 100 :", suite2[-5:])  # Convergent vers 6
````

    """Calcule les n premiers termes de Fibonacci"""
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib

## üìö Notion 5 : Repr√©sentation graphique d'une suite

### üìñ Cours

Visualiser une suite permet de mieux comprendre son **comportement** (croissance, d√©croissance, convergence, oscillation).

**M√©thode avec matplotlib :**
```python
import matplotlib.pyplot as plt
plt.plot(u, 'o-')  # Points reli√©s
plt.show()
```

**Types de graphiques :**
- Nuage de points : `plt.plot(u, 'o')`
- Points reli√©s : `plt.plot(u, 'o-')`
- Ligne continue : `plt.plot(u)`

**Exemple r√©solu :** Comparaison de deux suites

````python
import matplotlib.pyplot as plt

# Suite arithm√©tique
arith = [10 + 3*n for n in range(20)]

# Suite g√©om√©trique
geom = [10 * (1.15**n) for n in range(20)]

# Graphique
plt.figure(figsize=(10, 5))
plt.plot(arith, 'o-', label='Arithm√©tique (r=3)')
plt.plot(geom, 's-', label='G√©om√©trique (q=1.15)')
plt.title("Comparaison : suite arithm√©tique vs g√©om√©trique")
plt.xlabel("Rang n")
plt.ylabel("Valeur")
plt.legend()
plt.grid(True)
plt.show()
````

### √âtude compl√®te d'une suite

Soit $(u_n)$ d√©finie par : $u_0 = 1$, $u_{n+1} = \sqrt{2 + u_n}$

**Votre mission compl√®te :**

1. **Calcul** : Calculer les 50 premiers termes
2. **Visualisation** : Tracer le graphique avec une ligne horizontale √† $y = 2$ (limite suppos√©e)
3. **Convergence** : Afficher $u_{10}, u_{20}, u_{30}, u_{40}, u_{50}$
4. **Analyse** : Calculer la distance √† la limite 2 pour ces termes : $|u_n - 2|$
5. **Conclusion** : La suite semble-t-elle converger vers 2 ?

*Indice : Utilisez `import math` et `math.sqrt()`.*

````python
# √âcrivez votre code ici
````

<details>
<summary>üèÜ Correction du challenge</summary>

```python
import math
import matplotlib.pyplot as plt

# 1. Calcul des 50 premiers termes
u = [1]
for n in range(50):
    u.append(math.sqrt(2 + u[n]))

# 2. Visualisation
plt.figure(figsize=(12, 6))
plt.plot(u, 'o-', color='blue', label='Suite')
plt.axhline(y=2, color='red', linestyle='--', linewidth=2, label='Limite suppos√©e = 2')
plt.title("√âtude de la suite u_{n+1} = ‚àö(2 + u_n)")
plt.xlabel("Rang n")
plt.ylabel("Valeur")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 3. Termes particuliers
print("\nTermes particuliers :")
indices = [10, 20, 30, 40, 50]
for i in indices:
    print(f"u_{i} = {u[i]:.10f}")

# 4. Distance √† la limite
print("\nDistance √† la limite 2 :")
for i in indices:
    distance = abs(u[i] - 2)
    print(f"|u_{i} - 2| = {distance:.2e}")

# 5. Conclusion
print("\nüìä CONCLUSION :")
print(f"La suite converge vers 2.")
print(f"Distance finale |u_50 - 2| ‚âà {abs(u[50] - 2):.2e}")
print(f"V√©rification : si L = ‚àö(2 + L), alors L¬≤ = 2 + L")
print(f"Donc L¬≤ - L - 2 = 0, solutions : L = 2 ou L = -1")
print(f"Comme u_0 = 1 > 0 et u_{n+1} > 0, la limite est L = 2.")
```
