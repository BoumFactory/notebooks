## üìö Notion 1 : D√©finir une suite par r√©currence

### üìñ Cours

Une suite est **d√©finie par r√©currence** quand on conna√Æt :
1. Le **premier terme** $u_0$ (valeur initiale)
2. Une **relation de r√©currence** : $u_{n+1} = f(u_n)$

**M√©thode Python :**
```python
u = [u0]                    # Liste contenant le premier terme
for n in range(nb_termes):
    nouveau = f(u[n])       # Calcul du terme suivant
    u.append(nouveau)       # Ajout √† la liste
```

````python
# Exemple : u_0 = 10, u_{n+1} = 0.5 * u_n + 2

u = [10]  # Premier terme

for n in range(10):
    nouveau = 0.5 * u[n] + 2
    u.append(nouveau)

# Affichage
for n in range(len(u)):
    print(f"u_{n} = {u[n]:.2f}")
````

## üìö Notion 2 : Suite arithm√©tique

### üìñ Cours

Une **suite arithm√©tique** a une raison constante $r$. Chaque terme s'obtient en **ajoutant** $r$ au pr√©c√©dent.

| D√©finition | Formule |
|------------|--------|
| Par r√©currence | $u_{n+1} = u_n + r$ |
| Explicite | $u_n = u_0 + n \times r$ |

**Propri√©t√©s :**
- Si $r > 0$ : suite croissante
- Si $r < 0$ : suite d√©croissante
- Si $r = 0$ : suite constante

````python
# Exemple : Suite arithm√©tique u_0 = 5, r = 3

# M√©thode 1 : Par r√©currence
u = [5]
r = 3
for n in range(10):
    u.append(u[n] + r)
print("R√©currence :", u)

# M√©thode 2 : Formule explicite
u0, r = 5, 3
v = [u0 + n*r for n in range(11)]
print("Explicite  :", v)
````

# === SOLUTION ATTENDUE ===
def fibonacci(n):
    """Renvoie les n premiers termes de Fibonacci"""
    if n <= 0:
        return []
    if n == 1:
        return [0]
    
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])

## üìö Notion 5 : Repr√©sentation graphique

### üìñ Cours

Visualiser une suite permet de comprendre son **comportement** (croissance, convergence, oscillation).

```python

```

**Options :**
- `'o'` : points seuls
- `'-'` : ligne continue
- `'o-'` : points reli√©s

````python
import matplotlib.pyplot as plt

# Comparaison arithm√©tique vs g√©om√©trique
arith = [10 + 5*n for n in range(20)]
geom = [10 * (1.2**n) for n in range(20)]

plt.figure(figsize=(10, 5))
plt.plot(arith, 'o-', label='Arithm√©tique (r=5)')
plt.plot(geom, 's-', label='G√©om√©trique (q=1.2)')
plt.title("Comparaison des croissances")
plt.xlabel("Rang n")
plt.ylabel("Valeur")
plt.legend()
plt.grid(True)
plt.show()
````

import math
import matplotlib.pyplot as plt

# === SOLUTION ATTENDUE ===

# 1. Calcul
u = [1]
for n in range(29):
    u.append(math.sqrt(2 + u[n]))

# 2. Graphique
plt.figure(figsize=(10, 5))
plt.plot(u, 'o-', label='Suite')
plt.axhline(y=2, color='r', linestyle='--', label='Limite = 2')
plt.title("Suite : u_{n+1} = ‚àö(2 + u_n)")
plt.xlabel("Rang n")
plt.ylabel("Valeur")
plt.legend()
plt.grid(True)
plt.show()

# 3. Distance √† la limite
print("Distance √† la limite 2 :")
for n in [20, 25, 29]:
    print(f"  |u_{n} - 2| = {abs(u[n] - 2):.2e}")

# === BONUS (non demand√©) ===
# 4. Preuve alg√©brique : si L = sqrt(2 + L), alors L¬≤ = 2 + L
# L¬≤ - L - 2 = 0 ‚Üí (L-2)(L+1) = 0 ‚Üí L = 2 ou L = -1
# Comme tous les termes sont positifs, L = 2
print("\nPreuve : L¬≤ = 2 + L ‚Üí L = 2 (solution positive)")
