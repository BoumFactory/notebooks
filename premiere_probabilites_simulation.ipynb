## üìö Notion 1 : D√©couvrir le module random

### üìñ Cours

Le module **random** permet de g√©n√©rer des nombres al√©atoires en Python.

**Import du module :**
```python
import random
```

**Fonctions principales :**

| Fonction | Description | Exemple |
|----------|-------------|----------|
| `random.randint(a, b)` | Nombre al√©atoire entre a et b (inclus) | `random.randint(1, 6)` ‚Üí 1, 2, 3, 4, 5 ou 6 |
| `random.choice(liste)` | √âl√©ment al√©atoire d'une liste | `random.choice(['pile', 'face'])` |
| `random.random()` | Nombre d√©cimal entre 0 et 1 | `random.random()` ‚Üí 0.7234... |

**Exemple r√©solu :**

````python
import random

# G√©n√©rer un nombre entre 1 et 10
nombre = random.randint(1, 10)
print(f"Nombre al√©atoire : {nombre}")

# Choisir une couleur au hasard
couleurs = ['rouge', 'bleu', 'vert', 'jaune']
couleur = random.choice(couleurs)
print(f"Couleur tir√©e : {couleur}")
````

## üìö Notion 4 : Compter les issues favorables

### üìñ Cours

Une **issue favorable** est un r√©sultat qui v√©rifie une condition donn√©e.

**M√©thode :**
1. Initialiser un compteur √† 0
2. Simuler N exp√©riences
3. Si le r√©sultat est favorable, incr√©menter le compteur
4. Afficher le nombre d'issues favorables

**Exemple r√©solu :** Sur 100 lancers de d√©, combien de fois obtient-on un 6 ?

````python
import random

nb_six = 0  # Compteur d'issues favorables

for i in range(100):
    de = random.randint(1, 6)
    if de == 6:
        nb_six = nb_six + 1

print(f"Nombre de 6 obtenus : {nb_six}/100")
print(f"Probabilit√© th√©orique : 1/6 ‚âà {1/6:.3f}")
````

## üìö Notion 5 : Calculer des fr√©quences

### üìñ Cours

La **fr√©quence** d'un √©v√©nement est le rapport entre le nombre de fois o√π il se r√©alise et le nombre total d'exp√©riences.

$$\text{Fr√©quence} = \dfrac{\text{Nombre de succ√®s}}{\text{Nombre total d'exp√©riences}}$$

**Lien avec la probabilit√© :** Plus on r√©p√®te l'exp√©rience, plus la fr√©quence se rapproche de la probabilit√© th√©orique.

**Exemple r√©solu :** Calculer la fr√©quence d'obtenir un 6 sur 1000 lancers.

````python
import random

nb_six = 0
nb_lancers = 1000

for i in range(nb_lancers):
    de = random.randint(1, 6)
    if de == 6:
        nb_six = nb_six + 1

frequence = nb_six / nb_lancers

print(f"Nombre de 6 : {nb_six}/{nb_lancers}")
print(f"Fr√©quence : {frequence:.3f}")
print(f"Probabilit√© th√©orique : {1/6:.3f}")
````
