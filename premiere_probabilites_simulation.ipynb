  - Plat (100% des clients) : entre 15‚Ç¨ et 28‚Ç¨
  - Dessert (60% des clients) : entre 6‚Ç¨ et 12‚Ç¨
  - Boisson (75% des clients) : entre 3‚Ç¨ et 8‚Ç¨

**Mission :**

import random

chiffre_affaires = 0
nb_entrees = 0
nb_desserts = 0

    addition = 0

        addition = addition + random.uniform(3, 8)
        nb_boissons = nb_boissons + 1
    
    chiffre_affaires = chiffre_affaires + addition

ticket_moyen = chiffre_affaires / 50
