## üìö Notion 5 : Simulation et estimation de probabilit√©s

### üìñ Cours

En math√©matiques, on peut **estimer une probabilit√©** en r√©p√©tant une exp√©rience al√©atoire un grand nombre de fois.

**Principe :**
```python
# Compter le nombre de succ√®s
succes = 0
for i in range(nombre_simulations):
    if experience_reussie():
        succes = succes + 1

probabilite = succes / nombre_simulations
```

**Exemple r√©solu :**

````python
# Estimer la probabilit√© d'obtenir 6 avec un d√©
import random

succes = 0
simulations = 10000

for i in range(simulations):
    de = random.randint(1, 6)
    if de == 6:
        succes = succes + 1

probabilite = succes / simulations
print(f"Probabilit√© estim√©e : {probabilite:.4f}")
print(f"Probabilit√© th√©orique : {1/6:.4f}")
````

import random

succes = 0
simulations = 10000

for i in range(simulations):
    # G√©n√©rer 23 anniversaires al√©atoires
    anniversaires = []
    for j in range(23):
        jour = random.randint(1, 365)
        anniversaires.append(jour)

    if len(anniversaires) != len(set(anniversaires)):

### Simulation de Monte-Carlo : Estimer œÄ

On peut estimer œÄ en tirant des points al√©atoires dans un carr√© :

**M√©thode :**
1. Tirer 100000 points al√©atoires (x, y) dans le carr√© [0, 1] √ó [0, 1]
2. Compter combien sont dans le quart de cercle de rayon 1 (condition : x¬≤ + y¬≤ ‚â§ 1)
3. Estimation de œÄ : 4 √ó (nombre de points dans le cercle) / (nombre total de points)

**Indice :**
```python
x = random.random()  # Entre 0 et 1
y = random.random()
if x**2 + y**2 <= 1:
    # Point dans le cercle
```

````python
# √âcrivez votre code ici
````

<details>
<summary>üèÜ Correction du challenge</summary>

```python
import random

dans_cercle = 0
total = 100000

for i in range(total):
    x = random.random()
    y = random.random()
    
    # V√©rifier si le point est dans le quart de cercle
    if x**2 + y**2 <= 1:
        dans_cercle = dans_cercle + 1

# Estimation de œÄ
pi_estime = 4 * dans_cercle / total

print(f"Estimation de œÄ : {pi_estime:.6f}")
print(f"Valeur r√©elle de œÄ : 3.141593")
print(f"Erreur : {abs(pi_estime - 3.141593):.6f}")
```

**Explication math√©matique :**
- Aire du carr√© = 1
- Aire du quart de cercle = œÄ/4
- Proportion de points dans le cercle ‚âà (œÄ/4) / 1 = œÄ/4
- Donc œÄ ‚âà 4 √ó proportion
</details>
