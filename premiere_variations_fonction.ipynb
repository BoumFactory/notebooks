## üí° Introduction

En math√©matiques, **√©tudier les variations** d'une fonction signifie comprendre son comportement :
- O√π augmente-t-elle ?
- O√π diminue-t-elle ?
- Quels sont ses extremums (maximum et minimum) ?

<div style="background-color: #f0f4c3; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ü§î Probl√®me :</strong> Comment d√©terminer si une fonction est croissante ou d√©croissante sur un intervalle ?<br>
    <strong>‚ú® Solution :</strong> On calcule $f(x+h) - f(x)$ : si c'est positif, $f$ augmente ; si c'est n√©gatif, $f$ diminue !
</div>

### üîë D√©finitions math√©matiques

- **Fonction croissante** sur $I$ : $\forall x_1 < x_2 \in I, f(x_1) \leq f(x_2)$
- **Fonction d√©croissante** sur $I$ : $\forall x_1 < x_2 \in I, f(x_1) \geq f(x_2)$
- **Extremum** : valeur maximale ou minimale de la fonction sur un intervalle

## üìö Notion 1 : Calculer les valeurs d'une fonction sur un intervalle

### üìñ Cours

Pour √©tudier une fonction, on commence par calculer ses valeurs en plusieurs points d'un intervalle.

**M√©thode :**
1. D√©finir la fonction math√©matique
2. Choisir un intervalle $[a, b]$ et un pas $h$
3. Parcourir l'intervalle et calculer $f(x)$ pour chaque valeur

**Nombre de valeurs :** Sur $[a, b]$ avec un pas $h$, on obtient $\frac{b-a}{h} + 1$ valeurs.

**Exemple :** Pour $f(x) = x^2 - 4$ sur $[-3, 3]$ avec un pas de 1 :

````python
# Exemple : valeurs de f(x) = x¬≤ - 4 sur [-3, 3]
def f(x):
    return x**2 - 4

print("  x    |  f(x)")
print("-------+-------")

x = -3
pas = 1

while x <= 3:
    fx = f(x)
    print(f"{x:5.1f}  | {fx:6.2f}")
    x += pas
````

## üìö Notion 2 : D√©terminer le signe d'une fonction

### üìñ Cours

√âtudier le **signe** d'une fonction signifie d√©terminer pour quelles valeurs de $x$ la fonction est positive, n√©gative ou nulle.

**Interpr√©tation :**
- $f(x) > 0$ : la courbe est **au-dessus** de l'axe des abscisses
- $f(x) < 0$ : la courbe est **en-dessous** de l'axe des abscisses
- $f(x) = 0$ : la courbe **coupe** l'axe des abscisses (racines)

**M√©thode Python :**
```python
def signe(valeur):
    if valeur > 0:
        return "+"
    elif valeur < 0:
        return "-"
    else:
        return "0"
```

````python
# Exemple : tableau de signes de f(x) = x¬≤ - 4
def f(x):
    return x**2 - 4

def signe(valeur):
    if valeur > 0:
        return "+"
    elif valeur < 0:
        return "-"
    else:
        return "0"

print("  x    |  f(x)  | Signe")
print("-------+--------+-------")

x = -3
while x <= 3:
    fx = f(x)
    s = signe(fx)
    print(f"{x:5.1f}  | {fx:6.2f} |   {s}")
    x += 1

print("\nRacines : x = -2 et x = 2 (o√π f(x) = 0)")
````

## üìö Notion 3 : √âtudier la monotonie (croissance/d√©croissance)

### üìñ Cours

Pour d√©terminer si une fonction **cro√Æt** ou **d√©cro√Æt**, on √©tudie le signe de la **variation** :

$$\Delta f = f(x+h) - f(x)$$

**Interpr√©tation :**
- Si $\Delta f > 0$ : $f$ **augmente** entre $x$ et $x+h$ ‚Üí fonction **croissante** (‚Üó)
- Si $\Delta f < 0$ : $f$ **diminue** entre $x$ et $x+h$ ‚Üí fonction **d√©croissante** (‚Üò)
- Si $\Delta f = 0$ : $f$ est **constante** entre $x$ et $x+h$ (‚Üí)

**Lien avec la d√©riv√©e :** La variation $\Delta f$ est li√©e au taux d'accroissement !

````python
# Exemple : √©tude de la monotonie de f(x) = x¬≤ sur [-3, 3]
def f(x):
    return x**2

print("  x   |  f(x)  |   Œîf   | Sens")
print("------+--------+--------+------")

x = -3
pas = 0.5

while x < 3:
    fx = f(x)
    delta = f(x + pas) - f(x)
    
    if delta > 0:
        sens = "‚Üó"
    elif delta < 0:
        sens = "‚Üò"
    else:
        sens = "‚Üí"
    
    print(f"{x:5.1f} | {fx:6.2f} | {delta:6.2f} |  {sens}")
    x += pas

print("\nConclusion : f d√©cro√Æt sur ]-‚àû, 0] et cro√Æt sur [0, +‚àû[")
````

# === SOLUTION ATTENDUE ===
def etudier_monotonie(f, debut, fin, pas):
    """Analyse la monotonie d'une fonction sur un intervalle"""
    x = debut
    variations = []
    
    while x < fin:
        delta = f(x + pas) - f(x)
        variations.append(delta)
        x += pas
    
    if all(v > 0 for v in variations):
        return "Croissante"
    elif all(v < 0 for v in variations):
        return "D√©croissante"
    else:
        return "Variable"

prev_delta = None

while x < 5:

    if prev_delta is not None:
