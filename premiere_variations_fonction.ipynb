## ğŸ’¡ Introduction

En mathÃ©matiques, connaÃ®tre les **variations** d'une fonction permet de comprendre son comportement :
- OÃ¹ augmente-t-elle ?
- OÃ¹ diminue-t-elle ?
- Quels sont ses extremums ?

<div style="background-color: #f0f4c3; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ¤” ProblÃ¨me :</strong> Comment dÃ©terminer si une fonction est croissante ou dÃ©croissante sur un intervalle ?<br>
    <strong>âœ¨ Solution :</strong> On calcule $f(x+h) - f(x)$ : si c'est positif, $f$ augmente ; si c'est nÃ©gatif, $f$ diminue !
</div>

### ğŸ”‘ DÃ©finitions

- **Fonction croissante** sur un intervalle : $\forall x_1 < x_2, f(x_1) \leq f(x_2)$
- **Fonction dÃ©croissante** sur un intervalle : $\forall x_1 < x_2, f(x_1) \geq f(x_2)$

## 1ï¸âƒ£ Calculer des variations

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– DÃ©finition :</strong> La <strong>variation</strong> de $f$ entre $x$ et $x+h$ est :
    $$\Delta f = f(x+h) - f(x)$$
</div>

### ğŸ”§ InterprÃ©tation

- Si $\Delta f > 0$ : $f$ **augmente** entre $x$ et $x+h$
- Si $\Delta f < 0$ : $f$ **diminue** entre $x$ et $x+h$
- Si $\Delta f = 0$ : $f$ est **constante** entre $x$ et $x+h$

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Calculer les variations de f(x) = xÂ²

def f(x):
    return x**2

def variation(f, x, h):
    """Calcule f(x+h) - f(x)"""
    return f(x + h) - f(x)

# Tests
print("ğŸ“Š Variations de f(x) = xÂ²\n")
print("x   | h   | f(x+h) - f(x) | InterprÃ©tation")
print("-" * 60)

tests = [
    (2, 1),
    (3, 0.5),
    (-2, 1),
    (-3, 0.5)
]

for x, h in tests:
    var = variation(f, x, h)
    if var > 0:
        interpretation = "f augmente"
    elif var < 0:
        interpretation = "f diminue"
    else:
        interpretation = "f constante"
    
    print(f"{x:3} | {h:3} | {var:13.2f} | {interpretation}")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li>Pour $x > 0$, la variation est positive â†’ $f$ augmente</li>
        <li>Pour $x < 0$, la variation est nÃ©gative â†’ $f$ diminue</li>
        <li>Cela confirme que $x^2$ dÃ©croÃ®t sur $]-\infty, 0]$ et croÃ®t sur $[0, +\infty[$</li>
    </ul>
</div>

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Fonction d'analyse du signe

def analyser_variation(f, x, h):
    """Analyse si f augmente, diminue ou est constante entre x et x+h"""
    var = f(x + h) - f(x)
    
    if var > 0:
        return "croissante", var
    elif var < 0:
        return "dÃ©croissante", var
    else:
        return "constante", var

# Test avec f(x) = xÂ³
def f(x):
    return x**3

print("ğŸ“Š Analyse de f(x) = xÂ³\n")

for x in [-2, -1, 0, 1, 2]:
    sens, var = analyser_variation(f, x, 0.1)
    print(f"En x = {x:2}, f est {sens:12} (Î”f = {var:6.3f})")
````

## 3ï¸âƒ£ Conjecturer la monotonie sur un intervalle

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– DÃ©finition :</strong> Une fonction est <strong>monotone</strong> sur un intervalle si elle est soit toujours croissante, soit toujours dÃ©croissante sur cet intervalle.
</div>

### ğŸ” MÃ©thode

1. On teste la variation en plusieurs points de l'intervalle
2. Si toutes les variations sont du mÃªme signe, on conjecture la monotonie

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Ã‰tudier la monotonie de f(x) = xÂ² - 6x + 5 sur [0, 6]

def f(x):
    return x**2 - 6*x + 5

def etudier_monotonie(f, debut, fin, pas=0.5):
    """Ã‰tudie la monotonie de f sur [debut, fin]"""
    x = debut
    variations = []
    
    print(f"Ã‰tude de f sur [{debut}, {fin}]\n")
    print("x     | Î”f    | Sens")
    print("-" * 30)
    
    while x < fin:
        delta_f = f(x + pas) - f(x)
        variations.append(delta_f)
        
        if delta_f > 0:
            sens = "â†— croissante"
        elif delta_f < 0:
            sens = "â†˜ dÃ©croissante"
        else:
            sens = "â†’ constante"
        
        print(f"{x:5.1f} | {delta_f:5.2f} | {sens}")
        x += pas
    
    # Analyse globale
    print("\nğŸ“Š Analyse :")
    if all(v > 0 for v in variations):
        print("âœ… La fonction est CROISSANTE sur tout l'intervalle")
    elif all(v < 0 for v in variations):
        print("âœ… La fonction est DÃ‰CROISSANTE sur tout l'intervalle")
    else:
        print("âš ï¸ La fonction change de sens (ni croissante ni dÃ©croissante)")

etudier_monotonie(f, 0, 6)
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li>La fonction $f(x) = x^2 - 6x + 5$ a un minimum en $x = 3$</li>
        <li>Elle dÃ©croÃ®t sur $[0, 3]$ puis croÃ®t sur $[3, 6]$</li>
        <li>Le changement de signe des variations rÃ©vÃ¨le ce changement de monotonie</li>
    </ul>
</div>

### ğŸ§ª Exemple 2 : Ã€ vous de complÃ©ter !

````python
# ğŸ”¨ Ã€ COMPLÃ‰TER : Ã‰tudier g(x) = 2x - 3 sur [0, 5]

def g(x):
    return ____  # 2x - 3

def etudier_monotonie(f, debut, fin, pas=0.5):
    x = debut
    variations = []
    
    while x < fin:
        delta_f = f(x + pas) - f(x)
        variations.append(delta_f)
        x += pas
    
    if all(v > 0 for v in variations):
        print("âœ… La fonction est CROISSANTE")
    elif all(v < 0 for v in variations):
        print("âœ… La fonction est DÃ‰CROISSANTE")
    else:
        print("âš ï¸ La fonction change de sens")

etudier_monotonie(____, 0, 5)  # ComplÃ©tez avec g
````

## 4ï¸âƒ£ Dresser un tableau de variations

<div style="background-color: #fff9c4; padding: 15px; border-radius: 10px; border-left: 5px solid #FFC107;">
    <strong>ğŸ¯ Objectif :</strong> CrÃ©er automatiquement un tableau de variations simplifiÃ©.
</div>

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Tableau de variations de f(x) = -xÂ² + 4x - 1

def f(x):
    return -x**2 + 4*x - 1

def tableau_variations(f, debut, fin, pas=0.5):
    """Dresse un tableau de variations simplifiÃ©"""
    print(f"ğŸ“Š Tableau de variations de f sur [{debut}, {fin}]\n")
    print("x      | f(x)  | Variation | Sens")
    print("-" * 45)
    
    x = debut
    while x <= fin:
        fx = f(x)
        
        if x < fin:
            delta = f(x + pas) - f(x)
            if delta > 0:
                sens = "â†—"
            elif delta < 0:
                sens = "â†˜"
            else:
                sens = "â†’"
        else:
            delta = 0
            sens = ""
        
        print(f"{x:6.1f} | {fx:5.2f} | {delta:9.3f} | {sens}")
        x += pas

tableau_variations(f, 0, 4)
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li>On affiche pour chaque $x$ : la valeur $f(x)$, la variation vers le prochain point, et le sens</li>
        <li>Les flÃ¨ches â†— et â†˜ indiquent le sens de variation</li>
        <li>On observe un maximum autour de $x = 2$</li>
    </ul>
</div>
