- **Extremum** : valeur maximale ou minimale de la fonction sur un intervalle

## ğŸ“š Notion 2 : DÃ©terminer le signe d'une fonction

### ğŸ“– Cours

<div style="background-color: #fff9c4; padding: 15px; border-radius: 10px; border-left: 5px solid #FFC107;">

**Rappel mathÃ©matique :**
- $f(x) > 0$ : la fonction est **positive**
- $f(x) < 0$ : la fonction est **nÃ©gative**
- $f(x) = 0$ : la fonction s'annule (racines)

**MÃ©thode en Python :**
```python

    print("f est nÃ©gative")
```

**Application :** Cela permet de crÃ©er un tableau de signes automatiquement !

````python
# âœ… EXEMPLE COMPLET : Tableau de signes de f(x) = xÂ² - 4

def f(x):
    return x**2 - 4

def signe(valeur):
    """Retourne le signe d'une valeur"""
    if valeur > 0:
        return "+"
    elif valeur < 0:
        return "-"
    else:
        return "0"

print("ğŸ“Š Tableau de signes de f(x) = xÂ² - 4\n")
print("  x    |  f(x)  | Signe")
print("-------+--------+-------")

x = -3
while x <= 3:
    fx = f(x)
    s = signe(fx)
    print(f"{x:5.1f}  | {fx:6.2f} |   {s}")
    x += 0.5

print("\nğŸ’¡ On constate que f s'annule en x = -2 et x = 2")
````

## ğŸ“š Notion 3 : Ã‰tudier la monotonie (croissance/dÃ©croissance)

### ğŸ“– Cours

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“ Principe :</strong> Pour dÃ©terminer si une fonction croÃ®t ou dÃ©croÃ®t, on Ã©tudie le signe de $f(x+h) - f(x)$.
</div>

**DÃ©finition de la variation :**
$$\Delta f = f(x+h) - f(x)$$

**InterprÃ©tation :**
- Si $\Delta f > 0$ : $f$ **augmente** entre $x$ et $x+h$ â†’ fonction **croissante**
- Si $\Delta f < 0$ : $f$ **diminue** entre $x$ et $x+h$ â†’ fonction **dÃ©croissante**
- Si $\Delta f = 0$ : $f$ est **constante** entre $x$ et $x+h$

**MÃ©thode en Python :**
```python
def variation(f, x, h):
    return f(x + h) - f(x)

var = variation(f, x, 0.1)
if var > 0:
    print("Croissante")
elif var < 0:
    print("DÃ©croissante")
```

````python
# âœ… EXEMPLE COMPLET : Ã‰tudier la monotonie de f(x) = xÂ² sur [-3, 3]

def f(x):
    return x**2

def variation(f, x, h):
    """Calcule f(x+h) - f(x)"""
    return f(x + h) - f(x)

print("ğŸ“Š Ã‰tude de la monotonie de f(x) = xÂ²\n")
print("  x   |  f(x)  | Î”f  | Sens")
print("------+--------+-----+-----------")

x = -3
h = 0.5

while x < 3:
    fx = f(x)
    delta = variation(f, x, h)
    
    if delta > 0:
        sens = "â†— Croissante"
    elif delta < 0:
        sens = "â†˜ DÃ©croissante"
    else:
        sens = "â†’ Constante"
    
    print(f"{x:5.1f} | {fx:6.2f} | {delta:4.2f} | {sens}")
    x += h

print("\nğŸ’¡ La fonction xÂ² dÃ©croÃ®t sur ]-âˆ, 0] et croÃ®t sur [0, +âˆ[")
````

## ğŸ“š Notion 4 : Trouver les extremums (minimum/maximum)

### ğŸ“– Cours

<div style="background-color: #fff9c4; padding: 15px; border-radius: 10px; border-left: 5px solid #FFC107;">
    <strong>ğŸ“ Principe :</strong> Un **extremum** est une valeur minimale ou maximale d'une fonction sur un intervalle.
</div>

**DÃ©finitions :**
- **Minimum** : plus petite valeur de $f(x)$ sur l'intervalle
- **Maximum** : plus grande valeur de $f(x)$ sur l'intervalle
- **Point d'extremum** : valeur de $x$ oÃ¹ l'extremum est atteint

**MÃ©thode de dÃ©tection :**
1. Calculer toutes les valeurs de $f(x)$ sur l'intervalle
2. Comparer les valeurs pour trouver le min et le max
3. Retenir la valeur de $x$ correspondante

**MÃ©thode alternative :** Un extremum apparaÃ®t souvent quand la fonction **change de monotonie** :
- Si $\Delta f$ passe de positif Ã  nÃ©gatif â†’ **maximum**
- Si $\Delta f$ passe de nÃ©gatif Ã  positif â†’ **minimum**

````python
# âœ… EXEMPLE COMPLET : Trouver les extremums de f(x) = -xÂ² + 4x - 1

def f(x):
    return -x**2 + 4*x - 1

# MÃ©thode 1 : Recherche par comparaison
def trouver_extremums(f, debut, fin, pas):
    x = debut
    min_val = f(x)
    max_val = f(x)
    x_min = x
    x_max = x
    
    while x <= fin:
        fx = f(x)
        
        if fx < min_val:
            min_val = fx
            x_min = x
        
        if fx > max_val:
            max_val = fx
            x_max = x
        
        x += pas
    
    return (x_min, min_val), (x_max, max_val)

# Recherche sur [0, 4]
minimum, maximum = trouver_extremums(f, 0, 4, 0.1)

print("ğŸ“Š Extremums de f(x) = -xÂ² + 4x - 1 sur [0, 4]\n")
print(f"Minimum : f({minimum[0]:.1f}) = {minimum[1]:.2f}")
print(f"Maximum : f({maximum[0]:.1f}) = {maximum[1]:.2f}")

print("\nğŸ’¡ La fonction a un maximum en x = 2 (sommet de la parabole)")
````

````python
# âœ… EXEMPLE COMPLET : Tableau de variations de f(x) = xÂ² - 6x + 5

def f(x):
    return x**2 - 6*x + 5

def tableau_variations_complet(f, debut, fin, pas):
    """CrÃ©e un tableau de variations dÃ©taillÃ©"""
    print("ğŸ“Š TABLEAU DE VARIATIONS COMPLET\n")
    print("  x   |  f(x)  |   Î”f   | Sens  | Remarque")
    print("------+--------+--------+-------+-------------")
    
    x = debut
    min_val = f(x)
    x_min = x
    max_val = f(x)
    x_max = x
    prev_delta = None
    
    while x <= fin:
        fx = f(x)
        
        # Mise Ã  jour des extremums
        if fx < min_val:
            min_val = fx
            x_min = x
        if fx > max_val:
            max_val = fx
            x_max = x
        
        # Calcul de la variation
        if x < fin:
            delta = f(x + pas) - f(x)
            
            if delta > 0:
                sens = "  â†—  "
            elif delta < 0:
                sens = "  â†˜  "
            else:
                sens = "  â†’  "
            
            # DÃ©tection de changement
            remarque = ""
            if prev_delta is not None:
                if prev_delta > 0 and delta < 0:
                    remarque = "â† Maximum"
                elif prev_delta < 0 and delta > 0:
                    remarque = "â† Minimum"
            
            print(f"{x:5.1f} | {fx:6.2f} | {delta:6.3f} | {sens} | {remarque}")
            prev_delta = delta
        else:
            print(f"{x:5.1f} | {fx:6.2f} |    -   |   -   |")
        
        x += pas
    
    print("\n" + "="*60)
    print(f"ğŸ“ Minimum global : f({x_min:.1f}) = {min_val:.2f}")
    print(f"ğŸ“ Maximum global : f({x_max:.1f}) = {max_val:.2f}")

# Test sur [0, 6]
tableau_variations_complet(f, 0, 6, 0.5)
````
