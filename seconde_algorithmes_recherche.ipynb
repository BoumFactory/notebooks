## üìö Notion 1 : Parcourir une liste avec for

### üìñ Cours

Une liste est une collection ordonn√©e de valeurs. On peut parcourir tous ses √©l√©ments avec une boucle `for`.

**M√©thode 1 : Parcourir les valeurs directement**
```python
for valeur in liste:
    print(valeur)
```

**M√©thode 2 : Parcourir par indices** (n√©cessaire quand on veut la position)
```python
for i in range(len(liste)):
    print(i, liste[i])
```

| Fonction | Description | Exemple |
|----------|-------------|---------|
| `len(liste)` | Nombre d'√©l√©ments | `len([5,3,8])` ‚Üí 3 |
| `liste[i]` | √âl√©ment √† l'indice i | `[5,3,8][2]` ‚Üí 8 |

````python
# Exemple : Parcours direct vs par indices
notes = [12, 15, 8, 17]

print("Parcours direct :")
for note in notes:
    print(note)

print("\nParcours par indices :")
for i in range(len(notes)):
    print(f"Indice {i} ‚Üí {notes[i]}")
````

## üìö Notion 2 : Rechercher le maximum dans une liste

### üìñ Cours

**L'algorithme de recherche du maximum :**

1. **Initialiser** `maximum` avec le premier √©l√©ment de la liste
2. **Parcourir** tous les √©l√©ments de la liste
3. **Comparer** : si un √©l√©ment est plus grand que `maximum`, le remplacer
4. √Ä la fin, `maximum` contient la plus grande valeur

```python
maximum = liste[0]           # 1. Initialiser
for valeur in liste:         # 2. Parcourir
    if valeur > maximum:     # 3. Comparer
        maximum = valeur     # Mise √† jour
# maximum contient le plus grand
```

<div style="background-color: #ffecb3; padding: 10px; border-radius: 5px;">
‚ö†Ô∏è <strong>Important :</strong> On initialise avec <code>liste[0]</code> et non avec 0, car la liste pourrait ne contenir que des valeurs n√©gatives !
</div>

````python
# Exemple : Trouver la temp√©rature maximale
temperatures = [18, 22, 19, 25, 23, 20, 21]

maximum = temperatures[0]  # On part du premier : 18

for temp in temperatures:
    if temp > maximum:
        maximum = temp

print(f"Temp√©rature maximale : {maximum}¬∞C")
````

## üìö Notion 3 : Rechercher le minimum dans une liste

### üìñ Cours

L'algorithme de recherche du **minimum** est identique √† celui du maximum, mais on inverse le sens de la comparaison :

```python
minimum = liste[0]           # Initialiser
for valeur in liste:         # Parcourir
    if valeur < minimum:     # ‚Üê < au lieu de >
        minimum = valeur     # Mise √† jour
```

| Algorithme | Comparaison |
|------------|-------------|
| Maximum | `if valeur > maximum:` |
| Minimum | `if valeur < minimum:` |

````python
# Exemple : Trouver le prix le plus bas
prix = [45.99, 32.50, 28.90, 51.20, 39.00]

minimum = prix[0]

for p in prix:
    if p < minimum:
        minimum = p

print(f"Prix le plus bas : {minimum}‚Ç¨")
````

## üìö Notion 4 : Trouver l'indice du maximum

### üìñ Cours

Parfois, on veut savoir **o√π** se trouve le maximum, pas seulement sa valeur. On doit alors :
- Parcourir avec `range(len(liste))` pour avoir les indices
- Stocker deux informations : la valeur maximale ET sa position

```python
maximum = liste[0]
indice_max = 0

for i in range(len(liste)):
    if liste[i] > maximum:
        maximum = liste[i]
        indice_max = i       # On m√©morise aussi la position
```

````python
# Exemple : Trouver quel jour il a fait le plus chaud
temperatures = [18, 22, 19, 25, 23, 20, 21]
jours = ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"]

maximum = temperatures[0]
indice_max = 0

for i in range(len(temperatures)):
    if temperatures[i] > maximum:
        maximum = temperatures[i]
        indice_max = i

print(f"Jour le plus chaud : {jours[indice_max]} avec {maximum}¬∞C")
````
