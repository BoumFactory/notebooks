## üìñ Notion 1 : Parcourir une liste avec for

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>üìö Cours :</strong> Une liste est une collection ordonn√©e de valeurs. On peut parcourir tous ses √©l√©ments avec une boucle <code>for</code>.
</div>

### Deux m√©thodes de parcours

**M√©thode 1 : Parcourir les valeurs directement**
```python
for valeur in liste:
    print(valeur)
```

**M√©thode 2 : Parcourir par indices**
```python
for i in range(len(liste)):
    print(liste[i])
```

### üìù Exemple r√©solu

````python
# ‚úÖ EXEMPLE R√âSOLU : Afficher toutes les temp√©ratures d'une semaine
temperatures = [18, 22, 19, 25, 23, 20, 21]

print("üå°Ô∏è Temp√©ratures de la semaine :")
jours = ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"]

for i in range(len(temperatures)):
    print(f"{jours[i]} : {temperatures[i]}¬∞C")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>üí¨ Explication :</strong>
    <ul>
        <li><code>len(temperatures)</code> renvoie la longueur de la liste (7)</li>
        <li><code>range(7)</code> g√©n√®re 0, 1, 2, 3, 4, 5, 6</li>
        <li><code>temperatures[i]</code> acc√®de √† l'√©l√©ment √† l'indice i</li>
    </ul>
</div>

## üìñ Notion 2 : Rechercher le maximum dans une liste

<div style="background-color: #fff9c4; padding: 15px; border-radius: 10px; border-left: 5px solid #FFC107;">
    <strong>üìö Cours :</strong> Pour trouver la plus grande valeur dans une liste, on utilise un algorithme :
    <ol>
        <li>On initialise <code>maximum</code> avec le premier √©l√©ment de la liste</li>
        <li>On parcourt tous les √©l√©ments</li>
        <li>Si on trouve une valeur plus grande que <code>maximum</code>, on met √† jour <code>maximum</code></li>
        <li>√Ä la fin, <code>maximum</code> contient la plus grande valeur</li>
    </ol>
</div>

### üìù Exemple r√©solu

````python
# ‚úÖ EXEMPLE R√âSOLU : Trouver la temp√©rature maximale
temperatures = [18, 22, 19, 25, 23, 20, 21]

# Initialisation avec le premier √©l√©ment
maximum = temperatures[0]

# Parcours de la liste
for temp in temperatures:
    if temp > maximum:
        maximum = temp

print(f"üî• Temp√©rature maximale : {maximum}¬∞C")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>üí¨ Explication :</strong>
    <ul>
        <li><code>maximum = temperatures[0]</code> : on initialise avec le premier √©l√©ment (18)</li>
        <li><code>if temp > maximum</code> : on teste si la temp√©rature actuelle est plus grande</li>
        <li><code>maximum = temp</code> : si oui, on met √† jour le maximum</li>
    </ul>
</div>

## üìñ Notion 3 : Rechercher le minimum dans une liste

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>üìö Cours :</strong> L'algorithme de recherche du minimum est identique √† celui du maximum, mais on teste <code>if valeur < minimum</code> au lieu de <code>></code>.
</div>

### üìù Exemple r√©solu

````python
# ‚úÖ EXEMPLE R√âSOLU : Trouver le prix le plus bas
prix = [45.99, 32.50, 28.90, 51.20, 39.00]

minimum = prix[0]

for p in prix:
    if p < minimum:
        minimum = p

print(f"üí∞ Prix le plus bas : {minimum}‚Ç¨")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>üí¨ Explication :</strong>
    <ul>
        <li>Le principe est le m√™me que pour le maximum</li>
        <li>On change juste <code>></code> en <code><</code></li>
        <li>On initialise toujours avec le premier √©l√©ment</li>
    </ul>
</div>

## üìñ Notion 4 : Trouver l'indice du maximum

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>üìö Cours :</strong> En plus de la valeur maximale, on veut parfois savoir <strong>o√π elle se trouve</strong> dans la liste (son indice/position). On utilise alors deux variables : <code>maximum</code> pour la valeur et <code>indice_max</code> pour la position.
</div>

### üìù Exemple r√©solu

````python
# ‚úÖ EXEMPLE R√âSOLU : Trouver quel jour il a fait le plus chaud
temperatures = [18, 22, 19, 25, 23, 20, 21]
jours = ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"]

maximum = temperatures[0]
indice_max = 0

for i in range(len(temperatures)):
    if temperatures[i] > maximum:
        maximum = temperatures[i]
        indice_max = i

print(f"üî• Jour le plus chaud : {jours[indice_max]} avec {maximum}¬∞C")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>üí¨ Explication :</strong>
    <ul>
        <li>On parcourt avec <code>range(len(...))</code> pour avoir les indices</li>
        <li>On met √† jour <code>maximum</code> ET <code>indice_max</code> simultan√©ment</li>
        <li><code>indice_max</code> garde la position de la valeur maximale</li>
    </ul>
</div>
