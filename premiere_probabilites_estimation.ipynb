## üí° Introduction

Certains probl√®mes de probabilit√©s sont difficiles √† r√©soudre par le calcul th√©orique.

<div style="background-color: #f0f4c3; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ü§î Probl√®me :</strong> Comment conna√Ætre la probabilit√© d'un √©v√©nement complexe ?<br>
    <strong>‚ú® Solution :</strong> La <strong>simulation de grande ampleur</strong> : r√©p√©ter l'exp√©rience des milliers de fois !

**Principe :** Plus on r√©p√®te l'exp√©rience, plus la fr√©quence observ√©e se rapproche de la probabilit√© r√©elle. C'est la **loi des grands nombres**.

$$P(\text{√©v√©nement}) \approx \dfrac{\text{Nombre de succ√®s}}{\text{Nombre total de simulations}}$$

## üìö Notion 3 : Loi des grands nombres

### üìñ Cours

La **loi des grands nombres** est un th√©or√®me fondamental en probabilit√©s :

<div style="background-color: #e3f2fd; padding: 15px; border-radius: 10px; margin: 10px 0;">
    <strong>üìú Loi des grands nombres :</strong><br>
    Lorsque le nombre de simulations augmente, la fr√©quence observ√©e se stabilise et converge vers la probabilit√© th√©orique.
</div>

**Cons√©quence pratique :** Pour estimer pr√©cis√©ment une probabilit√©, il faut un grand nombre de simulations.

**Exemple :** Observer la convergence vers P(Face) = 0.5.

````python
import random

simulations = [100, 500, 1000, 5000, 10000, 50000]

print("Observation de la convergence vers P(Face) = 0.5")
print("-" * 55)

for n in simulations:
    nb_face = 0
    for i in range(n):
        tirage = random.choice(['Pile', 'Face'])
        if tirage == 'Face':
            nb_face = nb_face + 1
    
    frequence = nb_face / n
    ecart = abs(frequence - 0.5)
    print(f"N = {n:6} | Fr√©quence = {frequence:.5f} | √âcart = {ecart:.5f}")
````

## üèÜ Challenge final : Estimer une aire par simulation

On peut estimer l'**aire** situ√©e **sous une courbe** par une m√©thode probabiliste !

**Principe math√©matique :**

On consid√®re la fonction $f(x) = 4x(1-x)$ sur l'intervalle $[0, 1]$.

Cette fonction forme une parabole qui :
- vaut 0 en $x = 0$ et $x = 1$
- atteint son maximum 1 en $x = 0{,}5$
- reste donc enti√®rement dans le carr√© $[0, 1] \times [0, 1]$

**M√©thode :**
1. Tirer un point $(x, y)$ au hasard dans le carr√© unitaire
2. Si $y < f(x)$, le point est **sous la courbe** ‚Üí on le compte
3. L'aire sous la courbe $\approx \dfrac{\text{Nb points sous la courbe}}{\text{Nb total de points}}$

**Votre mission :**
1. Tirer **100 000** points $(x, y)$ al√©atoires dans le carr√© $[0, 1] \times [0, 1]$
2. Pour chaque point, calculer $f(x) = 4x(1-x)$ et comparer avec $y$
3. Compter les points situ√©s **sous** la courbe (ceux o√π $y < f(x)$)
4. Estimer l'aire et comparer avec la valeur th√©orique $\dfrac{2}{3}$

```python
# === SOLUTION ATTENDUE ===
import random

nb_sous_courbe = 0
nb_points = 100000

for i in range(nb_points):
    x = random.random()
    y = random.random()
    
    # Valeur de f(x) = 4x(1-x)
    f_x = 4 * x * (1 - x)
    
    # Le point est-il sous la courbe ?
    if y < f_x:
        nb_sous_courbe = nb_sous_courbe + 1

# Estimation de l'aire
aire_estimee = nb_sous_courbe / nb_points

print(f"Estimation de l'aire sous f(x) = 4x(1-x)")
print(f"Points tir√©s : {nb_points}")
print(f"Points sous la courbe : {nb_sous_courbe}")
print(f"Aire estim√©e : {aire_estimee:.4f}")
print(f"Aire th√©orique : {2/3:.4f}")
print(f"Erreur : {abs(aire_estimee - 2/3):.4f}")
```

**Pourquoi √ßa marche ?**

La probabilit√© qu'un point al√©atoire tombe sous la courbe est √©gale au rapport des aires :
$$P(\text{sous la courbe}) = \dfrac{\text{Aire sous la courbe}}{\text{Aire du carr√©}} = \dfrac{\text{Aire sous la courbe}}{1}$$

**Calcul th√©orique de l'aire :**

$$\int_0^1 4x(1-x)\, dx = \int_0^1 (4x - 4x^2)\, dx = \left[2x^2 - \frac{4x^3}{3}\right]_0^1 = 2 - \frac{4}{3} = \frac{2}{3}$$
