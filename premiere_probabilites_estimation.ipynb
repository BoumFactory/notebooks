## üìö Notion 1 : Principe de la simulation de grande ampleur

### üìñ Cours

Une **simulation de grande ampleur** consiste √† r√©p√©ter un tr√®s grand nombre de fois (1000, 10 000, voire 100 000 fois) une exp√©rience al√©atoire.

**M√©thode :**
1. Choisir le nombre de simulations (N grand, par exemple 10 000)
2. Initialiser un compteur de succ√®s √† 0
3. R√©p√©ter N fois l'exp√©rience
4. Si l'√©v√©nement se r√©alise, incr√©menter le compteur
5. Calculer la fr√©quence : `frequence = nb_succes / N`

**Avantage :** Permet d'estimer des probabilit√©s difficiles √† calculer th√©oriquement.

**Exemple r√©solu :** Estimer la probabilit√© d'obtenir Face avec 10 000 lancers.

````python
import random

nb_face = 0
nb_simulations = 10000

for i in range(nb_simulations):
    tirage = random.choice(['Pile', 'Face'])
    if tirage == 'Face':
        nb_face = nb_face + 1

frequence = nb_face / nb_simulations

print(f"Simulations : {nb_simulations}")
print(f"Nombre de Face : {nb_face}")
print(f"Fr√©quence observ√©e : {frequence:.4f}")
print(f"Probabilit√© th√©orique : 0.5000")
print(f"√âcart : {abs(frequence - 0.5):.4f}")
````

## üìö Notion 2 : Estimer la probabilit√© d'un √©v√©nement simple

### üìñ Cours

Pour estimer une probabilit√© inconnue, on utilise la **fr√©quence observ√©e** comme estimation.

**Principe :**
$$P(\text{√©v√©nement}) \approx \dfrac{\text{Nombre de fois o√π l'√©v√©nement se r√©alise}}{\text{Nombre total de simulations}}$$

Plus le nombre de simulations est grand, meilleure est l'estimation.

**Exemple r√©solu :** Estimer la probabilit√© d'obtenir une somme de 10 en lan√ßant deux d√©s.

````python
import random

nb_somme_10 = 0
nb_simulations = 20000

for i in range(nb_simulations):
    de1 = random.randint(1, 6)
    de2 = random.randint(1, 6)
    somme = de1 + de2
    if somme == 10:
        nb_somme_10 = nb_somme_10 + 1

frequence = nb_somme_10 / nb_simulations

print(f"Estimation de P(somme = 10) : {frequence:.4f}")
print(f"Probabilit√© th√©orique : {3/36:.4f}")
# Explication : 3 fa√ßons d'obtenir 10 sur 36 : (4,6), (5,5), (6,4)
````

## üìö Notion 3 : Loi des grands nombres

### üìñ Cours

La **loi des grands nombres** est un th√©or√®me fondamental en probabilit√©s :

<div style="background-color: #e3f2fd; padding: 15px; border-radius: 10px; margin: 10px 0;">
    <strong>üìú Loi des grands nombres :</strong><br>
    Lorsque le nombre de simulations augmente, la fr√©quence observ√©e se stabilise et se rapproche de la probabilit√© th√©orique.
</div>

**Cons√©quence pratique :** Pour estimer pr√©cis√©ment une probabilit√©, il faut un grand nombre de simulations.

**Exemple r√©solu :** Observer la convergence de la fr√©quence vers 0.5 pour Pile.

````python
import random

nb_pile = 0
simulations = [100, 500, 1000, 5000, 10000, 50000]

print("Observation de la convergence vers P(Pile) = 0.5")
print("-" * 50)

for n in simulations:
    nb_pile = 0
    for i in range(n):
        tirage = random.choice(['Pile', 'Face'])
        if tirage == 'Pile':
            nb_pile = nb_pile + 1
    
    frequence = nb_pile / n
    ecart = abs(frequence - 0.5)
    print(f"N = {n:6} | Fr√©quence = {frequence:.5f} | √âcart = {ecart:.5f}")
````

### üîç Exercice 1 : Comprendre un programme

Que fait le programme suivant ?

````python
import random

etapes = [10, 50, 100, 500, 1000, 5000]
nb_six_total = 0

print("Convergence vers P(6) = 1/6")
for n in etapes:
    # Simuler les lancers suppl√©mentaires
    nb_lancers_supplementaires = n - (etapes[etapes.index(n)-1] if etapes.index(n) > 0 else 0)
    for i in range(nb_lancers_supplementaires):
        de = random.randint(1, 6)
        if de == 6:
            nb_six_total = nb_six_total + 1
    
    freq = nb_six_total / n
    print(f"{n} lancers : fr√©quence = {freq:.4f}")
````

<details>
<summary>üí° Correction</summary>

Ce programme **observe la convergence de la fr√©quence** d'obtenir un 6 vers la probabilit√© th√©orique $\dfrac{1}{6} \approx 0.1667$. Il affiche la fr√©quence cumul√©e pour diff√©rents nombres de lancers croissants.

## üìö Notion 4 : Simulation avec plusieurs d√©s

### üìñ Cours

Lorsqu'on lance plusieurs d√©s simultan√©ment, on peut √©tudier des √©v√©nements plus complexes :
- La somme des d√©s
- Le maximum ou le minimum
- L'obtention de valeurs identiques (doublets, triplets)

**M√©thode :** Pour chaque simulation, g√©n√©rer autant de d√©s que n√©cessaire et tester la condition.

**Exemple r√©solu :** Probabilit√© d'obtenir au moins un 6 en lan√ßant 3 d√©s.

````python
import random

nb_au_moins_un_6 = 0
nb_simulations = 30000

for i in range(nb_simulations):
    de1 = random.randint(1, 6)
    de2 = random.randint(1, 6)
    de3 = random.randint(1, 6)
    
    # Au moins un 6 ?
    if de1 == 6 or de2 == 6 or de3 == 6:
        nb_au_moins_un_6 = nb_au_moins_un_6 + 1

proba_estimee = nb_au_moins_un_6 / nb_simulations

print(f"P(au moins un 6 sur 3 d√©s) ‚âà {proba_estimee:.4f}")
# Calcul th√©orique : 1 - (5/6)^3 ‚âà 0.4213
print(f"Probabilit√© th√©orique : {1 - (5/6)**3:.4f}")
````

## üìö Notion 5 : Applications : probl√®mes de probabilit√©s

### üìñ Cours

La simulation permet de r√©soudre des **probl√®mes complexes** difficiles √† calculer th√©oriquement.

**Strat√©gie :**
1. Mod√©liser le probl√®me (identifier les tirages al√©atoires)
2. D√©finir l'√©v√©nement favorable
3. Simuler un grand nombre de fois
4. Estimer la probabilit√© par la fr√©quence

**Exemple r√©solu :** Probl√®me de Monty Hall simplifi√© - Deux portes, une avec un prix. On choisit, puis on a le droit de changer. Vaut-il mieux changer ?

````python
import random

# Simulation : deux portes (0 ou 1), le prix est derri√®re une porte al√©atoire
nb_gains_sans_changer = 0
nb_gains_en_changeant = 0
nb_simulations = 10000

for i in range(nb_simulations):
    porte_prix = random.randint(0, 1)  # Porte avec le prix
    choix_initial = random.randint(0, 1)  # Premier choix du joueur
    
    # Strat√©gie 1 : Ne pas changer
    if choix_initial == porte_prix:
        nb_gains_sans_changer = nb_gains_sans_changer + 1
    
    # Strat√©gie 2 : Changer (l'autre porte)
    choix_final = 1 - choix_initial  # L'autre porte
    if choix_final == porte_prix:
        nb_gains_en_changeant = nb_gains_en_changeant + 1

print(f"Sans changer : {nb_gains_sans_changer / nb_simulations:.4f}")
print(f"En changeant : {nb_gains_en_changeant / nb_simulations:.4f}")
print("\nConclusion : Les deux strat√©gies sont √©quivalentes (‚âà 0.5).")
````

### Probl√®me de l'aiguille de Buffon (estimation de œÄ)

On peut estimer la valeur de œÄ par une simulation g√©om√©trique !

**M√©thode de Monte Carlo pour œÄ :**
- On tire N points al√©atoires dans un carr√© de c√¥t√© 1
- On compte combien sont √† l'int√©rieur du quart de cercle de rayon 1
- Le rapport des surfaces donne : $\dfrac{\text{Nb dans le cercle}}{N} \approx \dfrac{\pi/4}{1} = \dfrac{\pi}{4}$
- Donc $\pi \approx 4 \times \dfrac{\text{Nb dans le cercle}}{N}$

**Votre mission :**
1. Tirer 1 000 000 points al√©atoires (x, y) avec $0 \leq x \leq 1$ et $0 \leq y \leq 1$
2. Compter combien v√©rifient $x^2 + y^2 \leq 1$ (dans le quart de cercle)
3. Estimer œÄ
4. Comparer avec la valeur r√©elle de œÄ

*Indice : Utilisez `random.random()` pour g√©n√©rer des nombres d√©cimaux entre 0 et 1.*

````python
# √âcrivez votre code ici
````

<details>
<summary>üèÜ Correction du challenge</summary>

```python
import random
import math

nb_dans_cercle = 0
nb_points = 1000000

print(f"Estimation de œÄ par la m√©thode de Monte Carlo")
print(f"Tirage de {nb_points:,} points al√©atoires...")
print()

for i in range(nb_points):
    x = random.random()  # Entre 0 et 1
    y = random.random()  # Entre 0 et 1
    
    # Distance au centre (0, 0)
    distance_carree = x**2 + y**2
    
    # Dans le quart de cercle ?
    if distance_carree <= 1:
        nb_dans_cercle = nb_dans_cercle + 1

# Estimation de œÄ
pi_estime = 4 * nb_dans_cercle / nb_points

print(f"R√©sultats :")
print(f"Points dans le quart de cercle : {nb_dans_cercle:,}")
print(f"Estimation de œÄ : {pi_estime:.6f}")
print(f"Valeur r√©elle de œÄ : {math.pi:.6f}")
print(f"Erreur absolue : {abs(pi_estime - math.pi):.6f}")
print(f"Erreur relative : {abs(pi_estime - math.pi) / math.pi * 100:.4f}%")

print(f"\n‚ú® Avec 1 million de points, on obtient œÄ avec une pr√©cision d'environ 0.1% !")
```

**Explication de la m√©thode :**
- Aire du carr√© unitaire = 1
- Aire du quart de cercle de rayon 1 = œÄ/4
- Rapport des aires = œÄ/4
- En tirant des points al√©atoires uniform√©ment, le rapport du nombre de points dans le cercle sur le total tend vers œÄ/4
- C'est un exemple c√©l√®bre de m√©thode de Monte Carlo !
