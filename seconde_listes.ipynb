## üìö Notion 1 : Cr√©er une liste et acc√©der aux √©l√©ments

### üìñ Cours

Une **liste** est une collection ordonn√©e de valeurs, not√©e entre crochets `[]`.

```python
ma_liste = [valeur1, valeur2, valeur3, ...]
```

**Indices** : Chaque √©l√©ment a un **indice** (position) qui commence √† **0** :

| √âl√©ment | 12 | 15 | 8 | 20 | 11 |
|---------|----|----|---|----|----|  
| Indice | 0 | 1 | 2 | 3 | 4 |

Pour acc√©der √† un √©l√©ment : `liste[indice]`

‚ö†Ô∏è **Le premier √©l√©ment est √† l'indice 0, pas 1 !**

**Exemples :**

````python
# Cr√©er une liste de notes
notes = [12, 15, 8, 20, 11]

print(notes)       # Toute la liste
print(notes[0])    # Premier √©l√©ment : 12
print(notes[2])    # Troisi√®me √©l√©ment : 8
print(notes[-1])   # Dernier √©l√©ment : 11
````

````python
# Modifier un √©l√©ment
notes = [12, 15, 8, 20, 11]
notes[2] = 10  # Changer le 8 en 10
print(notes)
````

## üìö Notion 2 : Parcourir une liste avec for

### üìñ Cours

Pour traiter **tous** les √©l√©ments d'une liste, on utilise une boucle `for` :

```python
for element in liste:
    # Traiter element
```

**Exemples :**

````python
# Afficher tous les √©l√©ments
notes = [12, 15, 8, 20]

for note in notes:
    print(note)
````

````python
# Calculer la somme
notes = [12, 15, 8, 20]
total = 0

for note in notes:
    total = total + note

print("Somme :", total)
````

````python
# Compter les notes >= 10
notes = [12, 8, 15, 9, 18, 11]
nb_admis = 0

for note in notes:
    if note >= 10:
        nb_admis = nb_admis + 1

print("Nombre d'admis :", nb_admis)
````

## üìö Notion 3 : Fonctions utiles sur les listes

### üìñ Cours

Python fournit des fonctions pratiques pour les listes de nombres :

| Fonction | Description | Exemple |
|----------|-------------|---------|
| `len(liste)` | Nombre d'√©l√©ments | `len([1,2,3])` ‚Üí 3 |
| `sum(liste)` | Somme | `sum([1,2,3])` ‚Üí 6 |
| `min(liste)` | Minimum | `min([5,2,9])` ‚Üí 2 |
| `max(liste)` | Maximum | `max([5,2,9])` ‚Üí 9 |

**Exemples :**

````python
notes = [12, 15, 8, 20, 11]

print("Nombre de notes :", len(notes))
print("Somme :", sum(notes))
print("Moyenne :", sum(notes) / len(notes))
print("Note min :", min(notes))
print("Note max :", max(notes))
````

````python
# √âtendue (statistique)
temperatures = [18, 22, 25, 19, 21]
etendue = max(temperatures) - min(temperatures)
print("√âtendue :", etendue)
````
