## üèÜ Challenge final

### √âcart-type d'une s√©rie statistique

L'**√©cart-type** mesure la dispersion des valeurs autour de la moyenne.

**Formule** : $\sigma = \sqrt{\dfrac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2}$

**Algorithme** :
1. Calculer la moyenne $\bar{x}$
2. Pour chaque valeur, calculer $(x_i - \bar{x})^2$
3. Calculer la moyenne de ces carr√©s (variance)
4. Prendre la racine carr√©e (√©cart-type)

√âcrivez une fonction `ecart_type(liste)` et testez avec `[10, 12, 11, 13, 10]`.

<details>
<summary>üõ†Ô∏è Outils</summary>

- **Moyenne** : `sum(liste) / len(liste)`
- **√âcart** : `x - moyenne`
- **Carr√© de l'√©cart** : `ecart ** 2`
- **Racine carr√©e** : `resultat ** 0.5`
- **Variables** : `moyenne`, `somme_carres`, `variance`
</details>

````python
# √âcrivez votre code ici
````

<details>
<summary>üèÜ Correction</summary>

```python
# === SOLUTION ATTENDUE ===
def ecart_type(liste):
    # Calculer la moyenne
    moyenne = sum(liste) / len(liste)
    
    # Calculer la somme des carr√©s des √©carts
    somme_carres = 0
    for x in liste:
        ecart = x - moyenne
        somme_carres = somme_carres + ecart ** 2
    
    # Calculer la variance puis l'√©cart-type
    variance = somme_carres / len(liste)
    return variance ** 0.5

# Test
serie = [10, 12, 11, 13, 10]
print(f"Moyenne : {sum(serie)/len(serie)}")
print(f"√âcart-type : {ecart_type(serie):.2f}")  # ‚âà 1.10

# === BONUS (non demand√©) ===
# V√©rification : moyenne = 11.2, variance = 1.36, √©cart-type ‚âà 1.17
```

**Interpr√©tation :** Un √©cart-type faible indique que les valeurs sont proches de la moyenne.
</details>
