## üìö Notion 1 : Cr√©er une liste et acc√©der aux √©l√©ments

### üìñ Cours

Une **liste** est une collection ordonn√©e de valeurs. En Python, on cr√©e une liste avec des crochets `[]`.

**Syntaxe :**
```python
ma_liste = [valeur1, valeur2, valeur3, ...]
```

**Indices** : Chaque √©l√©ment a un **indice** (position) qui commence √† **0** :

| √âl√©ment | 12 | 15 | 8 | 20 | 11 |
|---------|----|----|---|----|----||
| Indice  | 0  | 1  | 2 | 3  | 4  |

Pour acc√©der √† un √©l√©ment : `liste[indice]`

‚ö†Ô∏è **Attention** : Le premier √©l√©ment est √† l'indice 0, pas 1 !

**Exemple r√©solu :**

````python
# Cr√©er une liste de notes
notes = [12, 15, 8, 20, 11]

print(notes)        # Affiche toute la liste
print(notes[0])     # Premier √©l√©ment : 12
print(notes[2])     # Troisi√®me √©l√©ment : 8
print(notes[4])     # Cinqui√®me √©l√©ment : 11
````

````python
# Indices n√©gatifs : compter depuis la fin
notes = [12, 15, 8, 20, 11]

print(notes[-1])    # Dernier √©l√©ment : 11
print(notes[-2])    # Avant-dernier : 20
print(notes[-5])    # Premier √©l√©ment : 12
````

## üìö Notion 2 : Modifier une liste

### üìñ Cours

Les listes sont **modifiables** : on peut changer, ajouter ou supprimer des √©l√©ments.

**Modifier un √©l√©ment :**
```python
liste[indice] = nouvelle_valeur
```

**Ajouter un √©l√©ment :**
- `liste.append(valeur)` : ajoute √† la **fin**
- `liste.insert(indice, valeur)` : ins√®re √† une position pr√©cise

**Supprimer un √©l√©ment :**
- `liste.remove(valeur)` : supprime la premi√®re occurrence
- `liste.pop(indice)` : supprime √† un indice (et renvoie la valeur)

**Exemple r√©solu :**

````python
# Modifier un √©l√©ment
notes = [12, 8, 15]
notes[1] = 10  # Changer le deuxi√®me √©l√©ment
print(notes)   # [12, 10, 15]
````

````python
# Ajouter des √©l√©ments
notes = [12, 15]
notes.append(18)      # Ajoute 18 √† la fin
print(notes)          # [12, 15, 18]

notes.insert(1, 14)   # Ins√®re 14 √† l'indice 1
print(notes)          # [12, 14, 15, 18]
````

````python
# Supprimer des √©l√©ments
notes = [12, 15, 8, 20]
notes.remove(8)    # Supprime la valeur 8
print(notes)       # [12, 15, 20]

valeur = notes.pop(1)  # Supprime l'√©l√©ment d'indice 1
print(notes)           # [12, 20]
print(valeur)          # 15 (valeur supprim√©e)
````

## üìö Notion 3 : Parcourir une liste

### üìñ Cours

Pour traiter **tous** les √©l√©ments d'une liste, on utilise une boucle `for`.

**Deux m√©thodes :**

**M√©thode 1 : Parcourir les valeurs directement**
```python
for element in liste:
    # Traiter element
```

**M√©thode 2 : Parcourir les indices**
```python
for i in range(len(liste)):
    # Traiter liste[i]
```

**Fonction len()** : renvoie le nombre d'√©l√©ments

**Exemple r√©solu :**

````python
# M√©thode 1 : Parcourir les valeurs
notes = [12, 15, 8, 20]

for note in notes:
    print(note)
````

````python
# M√©thode 2 : Parcourir les indices
notes = [12, 15, 8, 20]

for i in range(len(notes)):
    print(f"Note {i+1} : {notes[i]}")
````

````python
# Calculer la somme des √©l√©ments
nombres = [5, 10, 15, 20]
total = 0

for nombre in nombres:
    total = total + nombre

print(f"Somme : {total}")  # 50
````

## üìö Notion 4 : Fonctions sur les listes

### üìñ Cours

Python fournit des **fonctions pratiques** pour les listes de nombres :

| Fonction | Description | Exemple |
|----------|-------------|----------|
| `len(liste)` | Nombre d'√©l√©ments | `len([1,2,3])` ‚Üí 3 |
| `sum(liste)` | Somme des √©l√©ments | `sum([1,2,3])` ‚Üí 6 |
| `min(liste)` | Plus petit √©l√©ment | `min([5,2,9])` ‚Üí 2 |
| `max(liste)` | Plus grand √©l√©ment | `max([5,2,9])` ‚Üí 9 |

**Concat√©nation** : Fusionner deux listes avec `+`

**Tranches (slicing)** : Extraire une portion de liste
```python
liste[debut:fin]  # De l'indice debut (inclus) √† fin (exclus)
```

**Exemple r√©solu :**

````python
# Fonctions statistiques
notes = [12, 15, 8, 20, 11]

print(f"Nombre de notes : {len(notes)}")
print(f"Somme : {sum(notes)}")
print(f"Moyenne : {sum(notes) / len(notes)}")
print(f"Note minimale : {min(notes)}")
print(f"Note maximale : {max(notes)}")
````

````python
# Concat√©nation
liste1 = [1, 2, 3]
liste2 = [4, 5]
liste_complete = liste1 + liste2
print(liste_complete)  # [1, 2, 3, 4, 5]
````

````python
# Tranches (slicing)
valeurs = [10, 20, 30, 40, 50]

print(valeurs[1:4])   # [20, 30, 40] (indices 1, 2, 3)
print(valeurs[:3])    # [10, 20, 30] (du d√©but √† l'indice 3 exclus)
print(valeurs[2:])    # [30, 40, 50] (de l'indice 2 √† la fin)
````

## üìö Notion 5 : Filtrer et compter

### üìñ Cours

On combine souvent **listes** et **conditions** pour :
- **Compter** les √©l√©ments qui v√©rifient une propri√©t√©
- **Filtrer** pour cr√©er une nouvelle liste

**M√©thode :**
```python
# Compter
compteur = 0
for element in liste:
    if condition:
        compteur = compteur + 1

# Filtrer
nouvelle_liste = []
for element in liste:
    if condition:
        nouvelle_liste.append(element)
```

**Exemple r√©solu :**

````python
# Compter les notes >= 10
notes = [12, 8, 15, 9, 18, 11]
nb_admis = 0

for note in notes:
    if note >= 10:
        nb_admis = nb_admis + 1

print(f"Nombre d'admis : {nb_admis}")
````

````python
# Filtrer : garder seulement les nombres pairs
nombres = [12, 7, 18, 5, 20, 9]
pairs = []

for n in nombres:
    if n % 2 == 0:
        pairs.append(n)

print(pairs)  # [12, 18, 20]
````

````python
# Calculer la moyenne des notes >= 10
notes = [12, 8, 15, 9, 18, 11]
notes_admis = []

for note in notes:
    if note >= 10:
        notes_admis.append(note)

moyenne = sum(notes_admis) / len(notes_admis)
print(f"Moyenne des admis : {moyenne}")
````

### Exercice 11 : M√©diane d'une liste

√âcrivez une fonction `mediane(liste)` qui calcule la m√©diane d'une liste **tri√©e** de nombres.

**Rappel :**
- Si la liste a un nombre **impair** d'√©l√©ments : m√©diane = √©l√©ment central
- Si la liste a un nombre **pair** d'√©l√©ments : m√©diane = moyenne des deux √©l√©ments centraux

*Indice : Utilisez `len(liste)`, l'op√©rateur `%` et l'op√©rateur `//` (division enti√®re).*

````python
# √âcrivez votre code ici
````

<details>
<summary>üí° Correction</summary>

```python
def mediane(liste):
    n = len(liste)
    
    if n % 2 == 1:  # Nombre impair d'√©l√©ments
        indice_median = n // 2
        return liste[indice_median]
    else:  # Nombre pair d'√©l√©ments
        milieu = n // 2
        return (liste[milieu - 1] + liste[milieu]) / 2

print(mediane([1, 3, 5, 7, 9]))        # 5
print(mediane([2, 4, 6, 8]))           # 5.0
print(mediane([10, 20, 30, 40, 50]))  # 30
```

## üèÜ Challenge final : Suite de Fibonacci

La **suite de Fibonacci** est d√©finie par :
- u‚ÇÄ = 0

def fibonacci(n):
    if n == 0:

        return [0]
