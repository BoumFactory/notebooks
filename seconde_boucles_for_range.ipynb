## üìö Notion 1 : La fonction range()

### üìñ Cours

La fonction `range()` g√©n√®re une **s√©quence de nombres entiers**. Elle est tr√®s utile pour cr√©er des boucles.

**Trois syntaxes possibles :**

| Syntaxe | R√©sultat |
|---------|----------|
| `range(n)` | G√©n√®re les nombres de **0** √† **n-1** |
| `range(a, b)` | G√©n√®re les nombres de **a** √† **b-1** |
| `range(a, b, pas)` | G√©n√®re les nombres de **a** √† **b-1** avec un **pas** |

‚ö†Ô∏è **Important** : La borne sup√©rieure n'est **jamais atteinte** !

**Exemple r√©solu :**

````python
# Exemple 1 : range(5) g√©n√®re 0, 1, 2, 3, 4
print("Nombres de 0 √† 4 :")
print(list(range(5)))

# Exemple 2 : range(2, 7) g√©n√®re 2, 3, 4, 5, 6
print("\nNombres de 2 √† 6 :")
print(list(range(2, 7)))

# Exemple 3 : range(0, 11, 2) g√©n√®re 0, 2, 4, 6, 8, 10
print("\nNombres pairs de 0 √† 10 :")
print(list(range(0, 11, 2)))
````

## üìö Notion 2 : La boucle for simple

### üìñ Cours

Une **boucle for** permet de **r√©p√©ter** des instructions un certain nombre de fois.

**Syntaxe :**
```python

    instruction1
    instruction2
    # ...
```

‚ö†Ô∏è **L'indentation est obligatoire** : toutes les instructions √† r√©p√©ter doivent √™tre d√©cal√©es (avec Tab).

**Exemple r√©solu :**

````python
# Afficher les nombres de 0 √† 4
for i in range(5):
    print(f"Valeur de i : {i}")
````

## üìö Notion 3 : Calculer des images avec une boucle

### üìñ Cours

En math√©matiques, on cr√©e souvent des **tableaux de valeurs** pour √©tudier une fonction.

**Principe :** On utilise une boucle for pour calculer automatiquement les images.

**Exemple r√©solu :**

````python
# Tableau de valeurs de f(x) = 2x + 3
def f(x):
    return 2 * x + 3

print("Tableau de valeurs de f(x) = 2x + 3")
print("="*40)
for x in range(6):
    print(f"f({x}) = {f(x)}")
````

## üìö Notion 4 : Variables accumulatrices

### üìñ Cours

Une **variable accumulatrice** est une variable qui accumule des valeurs au fil d'une boucle.

**Types courants :**
- **Somme** : on initialise √† `0` et on ajoute avec `+=`
- **Produit** : on initialise √† `1` et on multiplie avec `*=`
- **Compteur** : on initialise √† `0` et on incr√©mente de 1

**Exemple r√©solu :**

````python
# Calculer la somme S = 1 + 2 + 3 + ... + 10
somme = 0  # Initialisation

for i in range(1, 11):
    somme = somme + i  # ou somme += i

print(f"La somme de 1 √† 10 est : {somme}")

# V√©rification avec la formule : S_n = n(n+1)/2
n = 10
formule = n * (n + 1) // 2
print(f"V√©rification avec la formule : {formule}")
````

## üìö Notion 5 : Boucles avec conditions

### üìñ Cours

On peut **combiner une boucle for avec des conditions** pour s√©lectionner certaines valeurs.

**Exemple : compter seulement les nombres pairs**

**Exemple r√©solu :**

````python
# Compter les nombres pairs entre 1 et 20
compteur = 0

for i in range(1, 21):
    if i % 2 == 0:  # Si i est pair
        compteur += 1

print(f"Il y a {compteur} nombres pairs entre 1 et 20")
````
