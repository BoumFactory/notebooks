## üìö Notion 1 : La fonction range() et la boucle for

### üìñ Cours

**`range()`** g√©n√®re une s√©quence de nombres entiers :

| Syntaxe | R√©sultat |
|---------|----------|
| `range(n)` | 0, 1, 2, ..., n-1 |
| `range(a, b)` | a, a+1, ..., b-1 |
| `range(a, b, pas)` | a, a+pas, ... (jusqu'√† b-1) |

‚ö†Ô∏è **La borne sup√©rieure n'est jamais atteinte !**

**Boucle for** : r√©p√®te des instructions pour chaque valeur.
```python
for variable in range(...):
    instruction  # Indentation obligatoire !
```

**Exemples :**

````python
# range(5) g√©n√®re 0, 1, 2, 3, 4
for i in range(5):
    print(i)
````

````python
# range(2, 7) g√©n√®re 2, 3, 4, 5, 6
for i in range(2, 7):
    print(i)
````

````python
# range(0, 11, 2) g√©n√®re les nombres pairs de 0 √† 10
for i in range(0, 11, 2):
    print(i)
````

## üìö Notion 2 : Variables accumulatrices (sommes)

### üìñ Cours

Une **variable accumulatrice** accumule des valeurs au fil de la boucle.

**Pour une somme** :
1. Initialiser √† 0
2. Ajouter √† chaque tour de boucle

```python
somme = 0
for i in range(...):
    somme = somme + valeur  # ou somme += valeur
```

**Exemple :**

````python
# Calculer S = 1 + 2 + 3 + ... + 10
somme = 0

for i in range(1, 11):
    somme = somme + i

print("Somme de 1 √† 10 :", somme)  # 55
````

## üìö Notion 3 : Aper√ßu des listes

### üìñ Cours

Une **liste** est une collection ordonn√©e de valeurs, not√©e entre crochets `[]`.

```python
ma_liste = [10, 20, 30, 40]
```

**Pourquoi c'est utile ?** Une boucle `for` peut parcourir directement une liste !

```python
for element in ma_liste:
    # Faire quelque chose avec element
```

**Exemples :**

````python
# Cr√©er une liste
notes = [12, 15, 8, 20, 11]
print(notes)
````

````python
# Parcourir une liste
notes = [12, 15, 8, 20, 11]

for note in notes:
    print(note)
````

````python
# Calculer la somme d'une liste
notes = [12, 15, 8, 20, 11]
total = 0

for note in notes:
    total = total + note

print("Somme des notes :", total)
````
