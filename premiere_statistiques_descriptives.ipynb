## üìö Notion 1 : Calculer la moyenne d'une s√©rie

### üìñ Cours

La **moyenne** est la somme des valeurs divis√©e par l'effectif total.

**Formule math√©matique :**
$$\bar{x} = \dfrac{x_1 + x_2 + \cdots + x_n}{n}$$

**Fonctions Python utiles :**
- `sum(liste)` : calcule la somme des √©l√©ments
- `len(liste)` : retourne le nombre d'√©l√©ments

**M√©thode :**
```python
def moyenne(donnees):
    return sum(donnees) / len(donnees)
```

**Exemple r√©solu :** Notes d'un contr√¥le

````python
# Calculer la moyenne de notes
def moyenne(donnees):
    """Calcule la moyenne d'une s√©rie de donn√©es"""
    return sum(donnees) / len(donnees)

# Test avec des notes
notes = [12, 15, 8, 14, 16, 9, 18, 11, 13, 17]

moy = moyenne(notes)

print(f"Notes : {notes}")
print(f"Somme : {sum(notes)}")
print(f"Nombre de notes : {len(notes)}")
print(f"Moyenne : {moy}")
````

## üìö Notion 2 : Calculer la m√©diane d'une s√©rie

### üìñ Cours

La **m√©diane** est la valeur qui s√©pare la s√©rie ordonn√©e en deux parties de m√™me effectif.

**M√©thode :**
1. Trier la s√©rie avec `sorted()`
2. Si $n$ est **impair** : m√©diane = valeur centrale
3. Si $n$ est **pair** : m√©diane = moyenne des deux valeurs centrales

**Impl√©mentation Python :**
```python
def mediane(donnees):
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    
    if n % 2 == 1:  # Impair
        return donnees_triees[n // 2]
    else:  # Pair
        return (donnees_triees[n // 2 - 1] + donnees_triees[n // 2]) / 2
```

**Exemple r√©solu :** Cas pair et impair

````python
# Calculer la m√©diane
def mediane(donnees):
    """Calcule la m√©diane d'une s√©rie"""
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    
    if n % 2 == 1:  # Impair
        indice = n // 2
        return donnees_triees[indice]
    else:  # Pair
        indice1 = n // 2 - 1
        indice2 = n // 2
        return (donnees_triees[indice1] + donnees_triees[indice2]) / 2

# Test avec effectif impair
serie1 = [15, 8, 12, 20, 10]
print(f"S√©rie : {serie1}")
print(f"Tri√©e : {sorted(serie1)}")
print(f"M√©diane : {mediane(serie1)}")

# Test avec effectif pair
serie2 = [12, 15, 8, 14, 16, 9]
print(f"\nS√©rie : {serie2}")
print(f"Tri√©e : {sorted(serie2)}")
print(f"M√©diane : {mediane(serie2)}")
````

## üìö Notion 3 : Calculer min, max et l'√©tendue

### üìñ Cours

Ces indicateurs mesurent les **valeurs extr√™mes** d'une s√©rie.

**D√©finitions :**
- **Minimum** : plus petite valeur de la s√©rie
- **Maximum** : plus grande valeur de la s√©rie
- **√âtendue** : diff√©rence entre le maximum et le minimum

**Formule de l'√©tendue :**
$$\text{√âtendue} = \max - \min$$

**Fonctions Python :**
- `min(liste)` : retourne le minimum
- `max(liste)` : retourne le maximum

**Exemple r√©solu :** Temp√©ratures d'une semaine

````python
# Calculer min, max et √©tendue
def etendue(donnees):
    """Calcule l'√©tendue d'une s√©rie"""
    return max(donnees) - min(donnees)

# Test avec des temp√©ratures
temperatures = [18, 22, 19, 25, 23, 20, 21]

print(f"Temp√©ratures : {temperatures}")
print(f"Minimum : {min(temperatures)}¬∞C")
print(f"Maximum : {max(temperatures)}¬∞C")
print(f"√âtendue : {etendue(temperatures)}¬∞C")
print(f"\nL'amplitude des temp√©ratures est de {etendue(temperatures)}¬∞C")
````

## üìö Notion 4 : Calculer la variance et l'√©cart-type

### üìñ Cours

L'**√©cart-type** mesure la **dispersion** des valeurs autour de la moyenne.

**Formule de la variance :**
$$V = \dfrac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2$$

**Formule de l'√©cart-type :**
$$\sigma = \sqrt{V}$$

**Interpr√©tation :**
- **√âcart-type faible** : valeurs proches de la moyenne (donn√©es **homog√®nes**)
- **√âcart-type √©lev√©** : valeurs dispers√©es (donn√©es **h√©t√©rog√®nes**)

**Impl√©mentation Python :**
```python
import math

def ecart_type(donnees):
    moy = moyenne(donnees)
    variance = sum((x - moy)**2 for x in donnees) / len(donnees)
    return math.sqrt(variance)
```

**Exemple r√©solu :** Comparer deux s√©ries

````python
# Calculer la variance et l'√©cart-type
import math

def moyenne(donnees):
    return sum(donnees) / len(donnees)

def variance(donnees):
    """Calcule la variance d'une s√©rie"""
    moy = moyenne(donnees)
    n = len(donnees)
    somme_ecarts_carres = sum((x - moy)**2 for x in donnees)
    return somme_ecarts_carres / n

def ecart_type(donnees):
    """Calcule l'√©cart-type d'une s√©rie"""
    return math.sqrt(variance(donnees))

# Test avec deux s√©ries
serie_homogene = [10, 11, 10, 11, 10]  # Valeurs proches
serie_heterogene = [5, 10, 15, 2, 18]  # Valeurs dispers√©es

print("S√©rie homog√®ne :", serie_homogene)
print(f"  Moyenne : {moyenne(serie_homogene):.2f}")
print(f"  Variance : {variance(serie_homogene):.2f}")
print(f"  √âcart-type : {ecart_type(serie_homogene):.2f}")

print("\nS√©rie h√©t√©rog√®ne :", serie_heterogene)
print(f"  Moyenne : {moyenne(serie_heterogene):.2f}")
print(f"  Variance : {variance(serie_heterogene):.2f}")
print(f"  √âcart-type : {ecart_type(serie_heterogene):.2f}")
````

````python
import math

def calculer_dispersion(serie):
    moy = sum(serie) / len(serie)
    ecarts_carres = [(x - moy)**2 for x in serie]
    var = sum(ecarts_carres) / len(serie)
    return math.sqrt(var)

notes_classe_a = [12, 13, 11, 12, 13]
notes_classe_b = [5, 15, 10, 20, 10]

print(f"Classe A : √©cart-type = {calculer_dispersion(notes_classe_a):.2f}")
print(f"Classe B : √©cart-type = {calculer_dispersion(notes_classe_b):.2f}")
````


**Exemple r√©solu :** Analyse compl√®te d'une s√©rie de notes

````python
# Analyse statistique compl√®te
import math

def moyenne(donnees):
    return sum(donnees) / len(donnees)

def mediane(donnees):
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    if n % 2 == 1:
        return donnees_triees[n // 2]
    else:
        return (donnees_triees[n // 2 - 1] + donnees_triees[n // 2]) / 2

def ecart_type(donnees):
    moy = moyenne(donnees)
    variance = sum((x - moy)**2 for x in donnees) / len(donnees)
    return math.sqrt(variance)

def quartiles(donnees):
    """Calcule Q1, Q2 (m√©diane), Q3"""
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    
    Q1 = donnees_triees[n // 4]
    Q2 = mediane(donnees)
    Q3 = donnees_triees[3 * n // 4]
    
    return Q1, Q2, Q3

def analyser_serie(donnees, nom="S√©rie"):
    """Analyse compl√®te d'une s√©rie de donn√©es"""
    print(f"üìä Analyse de {nom}")
    print("=" * 50)
    print(f"Donn√©es : {donnees}")
    print(f"Effectif : {len(donnees)}")
    print()
    print(f"Minimum : {min(donnees)}")
    print(f"Maximum : {max(donnees)}")
    print(f"√âtendue : {max(donnees) - min(donnees)}")
    print()
    print(f"Moyenne : {moyenne(donnees):.2f}")
    print(f"M√©diane : {mediane(donnees):.2f}")
    print(f"√âcart-type : {ecart_type(donnees):.2f}")
    print()
    Q1, Q2, Q3 = quartiles(donnees)
    print(f"Q1 (25%) : {Q1}")
    print(f"Q2 (50%) : {Q2}")
    print(f"Q3 (75%) : {Q3}")
    print(f"√âcart interquartile : {Q3 - Q1}")
    print("=" * 50)

# Test
notes = [12, 15, 8, 14, 16, 9, 18, 11, 13, 17]
analyser_serie(notes, "Notes du contr√¥le")
````
