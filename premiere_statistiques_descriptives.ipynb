## üìö Notion 1 : Calculer la moyenne d'une s√©rie

### üìñ Cours

La **moyenne** est la somme des valeurs divis√©e par l'effectif total.

**Formule math√©matique :**
$$\bar{x} = \dfrac{x_1 + x_2 + \cdots + x_n}{n}$$

**Fonctions Python utiles :**
- `sum(liste)` : calcule la somme des √©l√©ments
- `len(liste)` : retourne le nombre d'√©l√©ments

````python
# ‚úÖ EXEMPLE : Calcul de la moyenne

def moyenne(donnees):
    """Calcule la moyenne d'une s√©rie de donn√©es"""
    return sum(donnees) / len(donnees)

# Test avec des notes
notes = [12, 15, 8, 14, 16]
print(f"Moyenne : {moyenne(notes)}")
````

## üìö Notion 3 : Calculer min, max et l'√©tendue

### üìñ Cours

Ces indicateurs mesurent les **valeurs extr√™mes** d'une s√©rie.

**D√©finitions :**
- **Minimum** : plus petite valeur (`min(liste)`)
- **Maximum** : plus grande valeur (`max(liste)`)
- **√âtendue** : diff√©rence entre le maximum et le minimum

**Formule :** $\text{√âtendue} = \max - \min$

````python
# ‚úÖ EXEMPLE : Min, max et √©tendue

def etendue(donnees):
    """Calcule l'√©tendue d'une s√©rie"""
    return max(donnees) - min(donnees)

temperatures = [18, 22, 19, 25, 23, 20, 21]
print(f"Min : {min(temperatures)}¬∞C")
print(f"Max : {max(temperatures)}¬∞C")
print(f"√âtendue : {etendue(temperatures)}¬∞C")
````

## üìö Notion 4 : Calculer la variance et l'√©cart-type

### üìñ Cours

L'**√©cart-type** mesure la **dispersion** des valeurs autour de la moyenne.

**Formule de la variance :**
$$V = \dfrac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2$$

**Formule de l'√©cart-type :**
$$\sigma = \sqrt{V}$$

**Interpr√©tation :**
- **√âcart-type faible** : valeurs proches de la moyenne (donn√©es **homog√®nes**)
- **√âcart-type √©lev√©** : valeurs dispers√©es (donn√©es **h√©t√©rog√®nes**)

````python
# ‚úÖ EXEMPLE : Variance et √©cart-type
import math

def moyenne(donnees):
    return sum(donnees) / len(donnees)

def variance(donnees):
    """Calcule la variance d'une s√©rie"""
    moy = moyenne(donnees)
    somme_carres = sum((x - moy)**2 for x in donnees)
    return somme_carres / len(donnees)

def ecart_type(donnees):
    """Calcule l'√©cart-type d'une s√©rie"""
    return math.sqrt(variance(donnees))

# Comparaison de deux s√©ries
homogene = [10, 11, 10, 11, 10]
heterogene = [5, 10, 15, 2, 18]

print(f"S√©rie homog√®ne : œÉ = {ecart_type(homogene):.2f}")
print(f"S√©rie h√©t√©rog√®ne : œÉ = {ecart_type(heterogene):.2f}")
````

## üìö Notion 5 : Analyse statistique compl√®te et quartiles

### üìñ Cours

Une **analyse compl√®te** regroupe tous les indicateurs statistiques.

**Quartiles :**
- **Q1** (1er quartile) : 25% des donn√©es sont inf√©rieures √† Q1
- **Q2** (2√®me quartile) : m√©diane (50%)
- **Q3** (3√®me quartile) : 75% des donn√©es sont inf√©rieures √† Q3

**√âcart interquartile :** $\text{IQR} = Q3 - Q1$ (dispersion des 50% centraux)

````python
# ‚úÖ EXEMPLE : Quartiles

def quartiles(donnees):
    """Calcule Q1, Q2 (m√©diane), Q3"""
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    
    Q1 = donnees_triees[...] # A compl√©ter 
    Q2 = donnees_triees[...] # A compl√©ter 
    Q3 = donnees_triees[...] # A compl√©ter 
    
    return Q1, Q2, Q3

# Test
notes = [12, 15, 8, 14, 16, 9, 18, 11, 13, 17]
Q1, Q2, Q3 = quartiles(notes)
print(f"Q1 = {Q1}, Q2 = {Q2}, Q3 = {Q3}")
print(f"IQR = {Q3 - Q1}")
````

import math

# === SOLUTION ATTENDUE ===
def moyenne(donnees):
    return sum(donnees) / len(donnees)

def mediane(donnees):
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    if n % 2 == 1:
        return donnees_triees[n // 2]
    else:
        return (donnees_triees[n // 2 - 1] + donnees_triees[n // 2]) / 2

def ecart_type(donnees):
    moy = moyenne(donnees)
    variance = sum((x - moy)**2 for x in donnees) / len(donnees)
    return math.sqrt(variance)

def quartiles(donnees):
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    Q1 = donnees_triees[n // 4]
    Q2 = mediane(donnees)
    Q3 = donnees_triees[3 * n // 4]
    return Q1, Q2, Q3

salaires = [1800, 2200, 1950, 2500, 1700, 2100, 2300, 1900]

print("ANALYSE DES SALAIRES")
print("=" * 40)
print(f"Effectif : {len(salaires)}")
print(f"Moyenne : {moyenne(salaires):.2f}‚Ç¨")
print(f"M√©diane : {mediane(salaires):.2f}‚Ç¨")

print(f"√âtendue : {max(salaires) - min(salaires)}‚Ç¨")

import math

# === SOLUTION ATTENDUE ===
def moyenne(donnees):
    return sum(donnees) / len(donnees)

def ecart_type(donnees):
    moy = moyenne(donnees)
    variance = sum((x - moy)**2 for x in donnees) / len(donnees)
    return math.sqrt(variance)

groupe1 = [14, 15, 13, 16, 14, 15]
groupe2 = [10, 18, 12, 16, 11, 17]
groupe3 = [8, 9, 7, 10, 8, 9]

groupes = [groupe1, groupe2, groupe3]
noms = ["Groupe 1", "Groupe 2", "Groupe 3"]

print(f"{'Groupe':<12} | {'Moyenne':<8} | {'√âcart-type':<10}")
print("-" * 40)

for nom, groupe in zip(noms, groupes):
    moy = moyenne(groupe)
    std = ecart_type(groupe)
    print(f"{nom:<12} | {moy:<8.2f} | {std:<10.2f}")

print("\nConclusions :")
print("- Meilleur niveau : Groupe 1 (moyenne = 14.50)")
print("- Plus homog√®ne : Groupe 1 (œÉ ‚âà 0.96)")

ecarts = [(nom, ecart_type(groupe)) for nom, groupe in zip(noms, groupes)]

moyennes_triees = sorted(moyennes, key=lambda x: x[1], reverse=True)
ecarts_tries = sorted(ecarts, key=lambda x: x[1])

for i, (nom, moy) in enumerate(moyennes_triees, 1):
