## 1ï¸âƒ£ Calculer la moyenne

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– DÃ©finition :</strong> La moyenne $\bar{x}$ d'une sÃ©rie de $n$ valeurs est :
    \(\bar{x} = \dfrac{x_1 + x_2 + \cdots + x_n}{n}\)
</div>

### ğŸ”§ Fonctions Python utiles

- `sum(liste)` : calcule la somme des Ã©lÃ©ments
- `len(liste)` : renvoie le nombre d'Ã©lÃ©ments

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Calculer la moyenne de notes

def moyenne(donnees):
    """Calcule la moyenne d'une sÃ©rie de donnÃ©es"""
    return sum(donnees) / len(donnees)

# Test avec des notes
notes = [12, 15, 8, 14, 16, 9, 18, 11, 13, 17]

moy = moyenne(notes)

print(f"Notes : {notes}")
print(f"Somme : {sum(notes)}")
print(f"Nombre de notes : {len(notes)}")
print(f"Moyenne : {moy}")
````

## 2ï¸âƒ£ Calculer la mÃ©diane

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– DÃ©finition :</strong> La <strong>mÃ©diane</strong> est la valeur qui sÃ©pare la sÃ©rie ordonnÃ©e en deux parties de mÃªme effectif.
</div>

### ğŸ”§ MÃ©thode

1. Trier la sÃ©rie avec `sorted()`
2. Si $n$ est **impair** : mÃ©diane = valeur centrale
3. Si $n$ est **pair** : mÃ©diane = moyenne des deux valeurs centrales

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Calculer la mÃ©diane

def mediane(donnees):
    """Calcule la mÃ©diane d'une sÃ©rie"""
    # Trier la sÃ©rie
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    
    if n % 2 == 1:  # Impair
        # Indice de l'Ã©lÃ©ment central
        indice = n // 2
        return donnees_triees[indice]
    else:  # Pair
        # Moyenne des deux Ã©lÃ©ments centraux
        indice1 = n // 2 - 1
        indice2 = n // 2
        return (donnees_triees[indice1] + donnees_triees[indice2]) / 2

# Test avec effectif impair
serie1 = [15, 8, 12, 20, 10]
print(f"SÃ©rie : {serie1}")
print(f"TriÃ©e : {sorted(serie1)}")
print(f"MÃ©diane : {mediane(serie1)}")

# Test avec effectif pair
serie2 = [12, 15, 8, 14, 16, 9]
print(f"\nSÃ©rie : {serie2}")
print(f"TriÃ©e : {sorted(serie2)}")
print(f"MÃ©diane : {mediane(serie2)}")
````

        <li>SÃ©rie 2 triÃ©e : [8, 9, 12, 14, 15, 16] â†’ mÃ©diane = (12 + 14) / 2 = 13</li>

## 3ï¸âƒ£ Calculer l'Ã©cart-type

<div style="background-color: #fff9c4; padding: 15px; border-radius: 10px; border-left: 5px solid #FFC107;">
    <strong>ğŸ¯ Objectif :</strong> Mesurer la <strong>dispersion</strong> des valeurs autour de la moyenne.
</div>

### ğŸ“ Formule mathÃ©matique

$$\sigma = \sqrt{\dfrac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2}$$

oÃ¹ $\bar{x}$ est la moyenne.

### ğŸ” InterprÃ©tation

- **Ã‰cart-type faible** : valeurs proches de la moyenne (donnÃ©es homogÃ¨nes)
- **Ã‰cart-type Ã©levÃ©** : valeurs dispersÃ©es (donnÃ©es hÃ©tÃ©rogÃ¨nes)

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Calculer l'Ã©cart-type
import math

def moyenne(donnees):
    return sum(donnees) / len(donnees)

def ecart_type(donnees):
    """Calcule l'Ã©cart-type d'une sÃ©rie"""
    moy = moyenne(donnees)
    n = len(donnees)
    
    # Somme des carrÃ©s des Ã©carts
    somme_ecarts_carres = 0
    for valeur in donnees:
        ecart = valeur - moy
        somme_ecarts_carres += ecart ** 2
    
    # Variance
    variance = somme_ecarts_carres / n
    
    # Ã‰cart-type = racine de la variance
    return math.sqrt(variance)

# Test avec deux sÃ©ries
serie_homogene = [10, 11, 10, 11, 10]  # Valeurs proches
serie_heterogene = [5, 10, 15, 2, 18]  # Valeurs dispersÃ©es

print("SÃ©rie homogÃ¨ne :", serie_homogene)
print(f"Moyenne : {moyenne(serie_homogene)}")
print(f"Ã‰cart-type : {ecart_type(serie_homogene):.2f}")

print("\nSÃ©rie hÃ©tÃ©rogÃ¨ne :", serie_heterogene)
print(f"Moyenne : {moyenne(serie_heterogene)}")
print(f"Ã‰cart-type : {ecart_type(serie_heterogene):.2f}")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li>SÃ©rie homogÃ¨ne : Ã©cart-type â‰ˆ 0.49 (faible, valeurs groupÃ©es)</li>
        <li>SÃ©rie hÃ©tÃ©rogÃ¨ne : Ã©cart-type â‰ˆ 5.29 (Ã©levÃ©, valeurs dispersÃ©es)</li>
        <li>Plus l'Ã©cart-type est grand, plus les donnÃ©es sont dispersÃ©es</li>
    </ul>
</div>

## 4ï¸âƒ£ Analyse statistique complÃ¨te

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– Objectif :</strong> CrÃ©er une fonction qui affiche tous les indicateurs d'une sÃ©rie.
</div>

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Analyse statistique complÃ¨te
import math

def moyenne(donnees):
    return sum(donnees) / len(donnees)

def mediane(donnees):
    donnees_triees = sorted(donnees)
    n = len(donnees_triees)
    if n % 2 == 1:
        return donnees_triees[n // 2]
    else:
        return (donnees_triees[n // 2 - 1] + donnees_triees[n // 2]) / 2

def ecart_type(donnees):
    moy = moyenne(donnees)
    variance = sum((x - moy)**2 for x in donnees) / len(donnees)
    return math.sqrt(variance)

def analyser_serie(donnees, nom="SÃ©rie"):
    """Analyse complÃ¨te d'une sÃ©rie de donnÃ©es"""
    print(f"ğŸ“Š Analyse de {nom}")
    print("=" * 50)
    print(f"DonnÃ©es : {donnees}")
    print(f"Effectif : {len(donnees)}")
    print()
    print(f"Minimum : {min(donnees)}")
    print(f"Maximum : {max(donnees)}")
    print(f"Ã‰tendue : {max(donnees) - min(donnees)}")
    print()
    print(f"Moyenne : {moyenne(donnees):.2f}")
    print(f"MÃ©diane : {mediane(donnees):.2f}")
    print(f"Ã‰cart-type : {ecart_type(donnees):.2f}")
    print("=" * 50)

# Test
notes_classe = [12, 15, 8, 14, 16, 9, 18, 11, 13, 17]
analyser_serie(notes_classe, "Notes du contrÃ´le")
````
