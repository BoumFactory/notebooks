## üí° Introduction : Rappels g√©om√©triques

### 1Ô∏è‚É£ Alignement de trois points

Trois points $A(x_A, y_A)$, $B(x_B, y_B)$, $C(x_C, y_C)$ sont **align√©s** si et seulement si le **d√©terminant** suivant est nul :

$$\text{det} = (x_B - x_A)(y_C - y_A) - (y_B - y_A)(x_C - x_A) = 0$$

**Interpr√©tation g√©om√©trique :** Ce d√©terminant mesure l'aire du parall√©logramme form√© par les vecteurs $\overrightarrow{AB}$ et $\overrightarrow{AC}$. Si cette aire est nulle, les vecteurs sont colin√©aires, donc les points sont align√©s.

### 2Ô∏è‚É£ √âquation d'une droite

Une droite passant par deux points $A(x_A, y_A)$ et $B(x_B, y_B)$ avec $x_A \neq x_B$ a pour √©quation :

$$y = ax + b$$

o√π :
- **Coefficient directeur** : $a = \dfrac{y_B - y_A}{x_B - x_A}$
- **Ordonn√©e √† l'origine** : $b = y_A - ax_A$

### 3Ô∏è‚É£ Distance entre deux points

La distance entre deux points $A(x_A, y_A)$ et $B(x_B, y_B)$ dans un rep√®re orthonormal est :

$$AB = \sqrt{(x_B - x_A)^2 + (y_B - y_A)^2}$$

C'est le **th√©or√®me de Pythagore** appliqu√© au triangle rectangle form√© par les projections.

### 4Ô∏è‚É£ Longueur d'un arc de courbe

Pour estimer la longueur d'un arc de courbe $y = f(x)$ entre $x = a$ et $x = b$, on peut **subdiviser** l'intervalle :

1. Diviser $[a, b]$ en $n$ segments de longueur $\text{pas} = \dfrac{b - a}{n}$
2. Calculer la distance entre les points successifs $(x_i, f(x_i))$ et $(x_{i+1}, f(x_{i+1}))$
3. Sommer toutes ces distances

**Principe :** Plus $n$ est grand, plus l'approximation est pr√©cise (on approche la courbe par une ligne bris√©e).

```python
def longueur_arc(f, a, b, n):
    longueur_totale = 0
    pas = (b - a) / n
    
    for i in range(n):
        x1 = a + i * pas
        x2 = x1 + pas
        y1 = f(x1)
        y2 = f(x2)
        longueur_totale += distance(x1, y1, x2, y2)
    
    return longueur_totale
```

**Explication :**
- On divise l'intervalle en `n` segments de taille `pas`
- √Ä chaque it√©ration, on calcule les coordonn√©es de deux points cons√©cutifs
- On calcule la distance entre ces points et on l'ajoute au total
- `+=` est un raccourci pour `longueur_totale = longueur_totale + ...`

**Observation :** Plus $n$ est grand, plus l'estimation est pr√©cise (environ $12{,}57$ m pour $n = 1000$).

**Pourquoi ?** Quand on augmente $n$, on subdivise plus finement la courbe, donc la ligne bris√©e qui relie les points se rapproche de plus en plus de la vraie courbe.
