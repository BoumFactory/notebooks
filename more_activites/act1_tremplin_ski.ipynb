### ‚úçÔ∏è Exercice 6 : Longueur de l'arc parabolique

Pour estimer la longueur de l'arc, on divise l'intervalle $[16, 28]$ en $n$ segments et on calcule la somme des longueurs de ces segments.

**Algorithme :**
1. Diviser $[16, 28]$ en $n$ intervalles de longueur $\text{pas} = \dfrac{28 - 16}{n}$
2. Pour chaque segment, calculer la distance entre $(x_i, f(x_i))$ et $(x_{i+1}, f(x_{i+1}))$
3. Sommer toutes ces distances

√âcrivez une fonction `longueur_arc(n, borne_inf, borne_sup)` qui estime la longueur de l'arc.

````python
def longueur_arc(n, borne_inf, borne_sup):
    longueur_totale = 0
    pas = (borne_sup - borne_inf) / n
    
    # Compl√©tez ici
    # Pour i de 0 √† n-1 :
    #   x1 = borne_inf + i * pas
    #   x2 = x1 + pas
    #   Calculer distance entre (x1, f(x1)) et (x2, f(x2))
    #   Ajouter √† longueur_totale
    
    return longueur_totale

# Testez avec n=10, n=100, n=1000
print(f"Longueur avec n=10 : {longueur_arc(10, 16, 28):.2f} m")
print(f"Longueur avec n=100 : {longueur_arc(100, 16, 28):.2f} m")
print(f"Longueur avec n=1000 : {longueur_arc(1000, 16, 28):.2f} m")
````

    pas = (borne_sup - borne_inf) / n
