## üí° Introduction th√©orique

### Chute libre sans frottements

Dans le vide, un objet en chute libre est soumis uniquement √† son poids $\vec{P} = m\vec{g}$.

D'apr√®s la 2√®me loi de Newton : $m\vec{a} = m\vec{g}$

En projetant sur l'axe vertical descendant :
$$\frac{dv}{dt} = g$$

La vitesse augmente **ind√©finiment** !

### Chute avec frottements

En r√©alit√©, l'air exerce une force de frottement $\vec{F}_{frot} = -k\vec{v}$ (proportionnelle √† la vitesse, oppos√©e au mouvement).

L'√©quation devient :
$$m\frac{dv}{dt} = mg - kv$$

Soit :
$$\frac{dv}{dt} = g - \frac{k}{m}v$$

### Vitesse limite

√Ä l'√©quilibre, quand $\frac{dv}{dt} = 0$ :
$$v_{\infty} = \frac{mg}{k}$$

**La vitesse ne cro√Æt plus ind√©finiment**, elle tend vers une limite !

### Application : le parachutiste

Un parachutiste subit **deux phases** :
1. **Chute libre** (parachute ferm√©) : $k$ petit ‚Üí $v_{\infty}$ grande (‚âà 200 km/h)
2. **Parachute ouvert** : $k$ grand ‚Üí $v_{\infty}$ petite (‚âà 20 km/h)

Le parachute augmente les frottements pour r√©duire la vitesse d'atterrissage.

def euler_chute_simple(v0, y0, g, h, n):
    t = [0]
    v = [v0]
    y = [y0]
    
    for i in range(n):
        # Calcul des nouvelles valeurs selon Euler
        v_new = v[-1] + h * g
        y_new = y[-1] - h * v[-1]
        
        # Arr√™t si le sol est atteint
        if y_new <= 0:
            break
        
        # Ajout des nouvelles valeurs
        t.append(t[-1] + h)
        v.append(v_new)
        y.append(y_new)
    
    return t, y, v
```

**Explications :**

- `h * g` est la variation de vitesse sur un pas de temps

- Le `break` arr√™te la boucle quand on touche le sol

def simulation_parachutiste():
    m, g = 80, 9.81
    h = 0.01
    
    t, v, y = [0], [0], [3000]
    
    # Phase 1 : chute libre (20 premi√®res secondes)
    k1 = 1
    for i in range(int(20/h)):
        v_new = v[-1] + h * (g - (k1/m) * v[-1])
        y_new = y[-1] - h * v[-1]
        t.append(t[-1] + h)
        v.append(v_new)
        y.append(y_new)
    
    # Phase 2 : parachute ouvert
    k2 = 50
    while y[-1] > 0:
        v_new = v[-1] + h * (g - (k2/m) * v[-1])
        y_new = y[-1] - h * v[-1]
        t.append(t[-1] + h)
        v.append(v_new)
        y.append(y_new)
    
    return t, y, v
```

**Points cl√©s :**
- `int(20/h)` donne le nombre d'it√©rations pour 20 secondes
- La boucle `while y[-1] > 0` s'arr√™te quand on touche le sol
- Le changement de $k$ mod√©lise l'ouverture du parachute

### ‚ú® Ce que vous avez appris

#### Concepts physiques
- ‚úÖ **Chute libre** : acc√©l√©ration constante $v' = g$
- ‚úÖ **Frottements fluides** : force proportionnelle √† la vitesse $F = -kv$
- ‚úÖ **Vitesse limite** : √©quilibre entre poids et frottements $v_{\infty} = \frac{mg}{k}$
- ‚úÖ **Applications** : parachutisme, m√©t√©orites, gouttes de pluie

#### Comp√©tences math√©matiques
- ‚úÖ **√âquations diff√©rentielles** : $\frac{dv}{dt} = g - \frac{k}{m}v$
- ‚úÖ **M√©thode d'Euler** : approximation num√©rique par pas successifs
- ‚úÖ **Analyse asymptotique** : convergence vers une limite

#### Comp√©tences Python
- ‚úÖ Impl√©menter des √©quations physiques avec des boucles `for`
- ‚úÖ G√©rer plusieurs phases avec `if` et `while`
- ‚úÖ Visualiser des donn√©es avec `matplotlib` (subplots)
- ‚úÖ Comparer simulation et th√©orie

### üîë Formules cl√©s

**√âquation du mouvement avec frottements :**
$$m\frac{dv}{dt} = mg - kv$$

**Vitesse limite :**
$$v_{\infty} = \frac{mg}{k}$$

**M√©thode d'Euler :**
$$v_{n+1} = v_n + h \times v'(t_n, v_n)$$
$$y_{n+1} = y_n - h \times v_n$$

### üéØ Applications r√©elles

| Situation | $k$ | $v_{\infty}$ |
|-----------|-----|-------------|
| Parachutiste (chute libre) | Faible | ‚âà 200 km/h |
| Parachutiste (parachute ouvert) | Fort | ‚âà 20 km/h |
| Goutte de pluie | Moyen | ‚âà 30 km/h |
| M√©t√©orite | Tr√®s fort | Variable |

### üí° Pour aller plus loin

- Frottements quadratiques : $F = -kv^2$ (vitesses √©lev√©es)
- Chute dans un fluide visqueux (loi de Stokes)
- Trajectoire balistique avec frottements
