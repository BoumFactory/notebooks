## üí° Introduction : Qu'est-ce qu'une √©preuve de Bernoulli ?

### D√©finition

Une **√©preuve de Bernoulli** est une exp√©rience al√©atoire qui n'a que **deux issues possibles** :
- Un **succ√®s** (on note souvent 1)
- Un **√©chec** (on note souvent 0)

On note $p$ la probabilit√© de succ√®s, donc $(1-p)$ est la probabilit√© d'√©chec.

### Exemples concrets

| Exp√©rience | Succ√®s | √âchec | Probabilit√© $p$ |
|------------|--------|-------|----------------|
| Lancer une pi√®ce √©quilibr√©e | Pile | Face | $p = 0{,}5$ |
| Lancer un d√© | Obtenir un 6 | Ne pas obtenir 6 | $p = \dfrac{1}{6} \approx 0{,}167$ |
| Tir au basket | Panier r√©ussi | Panier rat√© | $p = 0{,}7$ (selon le joueur) |
| Test m√©dical | Positif | N√©gatif | $p = 0{,}05$ (selon la maladie) |
| Naissance | Fille | Gar√ßon | $p \approx 0{,}49$ |

### Variable al√©atoire

On note $X$ la variable al√©atoire de Bernoulli :
- $P(X = 1) = p$ (probabilit√© de succ√®s)
- $P(X = 0) = 1 - p$ (probabilit√© d'√©chec)

On √©crit : $X \sim \mathcal{B}(p)$ (lire "$X$ suit une loi de Bernoulli de param√®tre $p$")

**Esp√©rance :** $E(X) = p$

## üí° De Bernoulli √† la loi binomiale

### D√©finition

Maintenant, au lieu de faire **une seule** √©preuve de Bernoulli, on en fait **$n$ √©preuves ind√©pendantes**.

On s'int√©resse au **nombre total de succ√®s** sur ces $n$ √©preuves.

**Exemple :** Je lance 10 fois une pi√®ce (10 √©preuves de Bernoulli avec $p=0{,}5$). Combien de fois vais-je obtenir Pile ?

### Variable al√©atoire binomiale

On note $X$ le nombre de succ√®s sur $n$ √©preuves de Bernoulli ind√©pendantes de param√®tre $p$.

On dit que $X$ suit une **loi binomiale** de param√®tres $n$ et $p$ :

$$X \sim \mathcal{B}(n, p)$$

**Formule de probabilit√© :**

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

o√π $\binom{n}{k} = \dfrac{n!}{k!(n-k)!}$ est le coefficient binomial (nombre de fa√ßons de choisir $k$ √©l√©ments parmi $n$).

**Esp√©rance et variance :**
- $E(X) = np$
- $V(X) = np(1-p)$

### Exemples

| Situation | $n$ | $p$ | Question |
|-----------|-----|-----|----------|
| 10 lancers de pi√®ce | 10 | 0,5 | Combien de Pile ? |
| 20 lancers de d√© | 20 | $\frac{1}{6}$ | Combien de 6 ? |
| 100 tirs au basket | 100 | 0,7 | Combien de paniers r√©ussis ? |
