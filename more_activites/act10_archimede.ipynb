## üí° Introduction : La m√©thode d'Archim√®de

### Un exploit math√©matique du IIIe si√®cle avant J.-C.

En 250 avant J√©sus-Christ, le math√©maticien grec **Archim√®de de Syracuse** a d√©velopp√© une m√©thode ing√©nieuse pour calculer une approximation de œÄ, sans conna√Ætre les d√©cimales modernes !

### Le principe : encadrer le cercle

L'id√©e d'Archim√®de √©tait d'**encadrer un cercle** entre deux polygones r√©guliers :
- Un polygone **inscrit** (√† l'int√©rieur du cercle)
- Un polygone **circonscrit** (√† l'ext√©rieur du cercle)

```
       Circonscrit
      _______________
     /               \
    /    ‚¨°  Cercle   \
   |   ‚óè  (rayon=1)  ‚óè|
    \     Inscrit    /
     \_______________/
```

### Les formules math√©matiques

Pour un cercle de **rayon 1**, Archim√®de a d√©montr√© :

**Polygone inscrit √† n c√¥t√©s :**
$$P_n = 2n \sin\left(\dfrac{\pi}{n}\right)$$

**Polygone circonscrit √† n c√¥t√©s :**
$$Q_n = 2n \tan\left(\dfrac{\pi}{n}\right)$$

### Pourquoi √ßa marche ?

Le p√©rim√®tre d'un cercle de rayon 1 est exactement $2\pi$.

En augmentant le nombre de c√¥t√©s $n$, les polygones se rapprochent de plus en plus du cercle, donc :

$$\dfrac{P_n}{2} \quad \xrightarrow[n \to \infty]{} \quad \pi \quad \xleftarrow[n \to \infty]{} \quad \dfrac{Q_n}{2}$$

### Le r√©sultat d'Archim√®de

Avec des polygones √† **96 c√¥t√©s** (et beaucoup de calculs √† la main !), Archim√®de a obtenu :

$$3{,}1408 < \pi < 3{,}1429$$

Ce qui donne **3 d√©cimales correctes** de œÄ ! Un exploit pour l'√©poque.
