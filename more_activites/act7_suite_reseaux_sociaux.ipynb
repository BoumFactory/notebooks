## üí° Introduction : Les deux types de croissance

### Situation

Un nouveau r√©seau social d√©marre avec **1000 membres** ($u_0 = 1000$).

Deux mod√®les de croissance sont possibles :

---

### üìä Mod√®le 1 : Suite arithm√©tique (croissance lin√©aire)

**D√©finition :** Chaque mois, le r√©seau gagne **un nombre fixe** de nouveaux membres.

**Formule de r√©currence :**
$$u_{n+1} = u_n + r$$

o√π $r$ est la **raison** (nombre constant ajout√© √† chaque √©tape).

**Formule explicite :**
$$u_n = u_0 + n \times r$$

**Exemple :** Si chaque mois le r√©seau gagne $r = 50$ membres :
- Mois 0 : $u_0 = 1000$ membres
- Mois 1 : $u_1 = 1000 + 50 = 1050$ membres
- Mois 2 : $u_2 = 1050 + 50 = 1100$ membres
- Mois 3 : $u_3 = 1100 + 50 = 1150$ membres

---

### üìà Mod√®le 2 : Suite g√©om√©trique (croissance exponentielle)

**D√©finition :** Chaque mois, le nombre de membres est **multipli√©** par un facteur constant.

**Formule de r√©currence :**
$$u_{n+1} = u_n \times q$$

o√π $q$ est la **raison** (facteur multiplicatif constant).

**Formule explicite :**
$$u_n = u_0 \times q^n$$

**Exemple :** Si chaque mois le nombre de membres augmente de 5% ($q = 1{,}05$) :
- Mois 0 : $u_0 = 1000$ membres
- Mois 1 : $u_1 = 1000 \times 1{,}05 = 1050$ membres
- Mois 2 : $u_2 = 1050 \times 1{,}05 = 1102{,}5$ membres
- Mois 3 : $u_3 = 1102{,}5 \times 1{,}05 = 1157{,}625$ membres

---

### ü§î Question de r√©flexion

Avant de commencer √† coder, r√©fl√©chissez :

**Question :** Sur le long terme (plusieurs ann√©es), quel mod√®le pensez-vous donnera le plus de membres ?

<details>
<summary>üí° R√©ponse</summary>

La **suite g√©om√©trique** finira toujours par d√©passer la suite arithm√©tique, m√™me si au d√©but elles semblent proches.

C'est le principe de la **croissance exponentielle** : elle d√©marre lentement mais explose ensuite.

Exemple r√©el : Les r√©seaux sociaux comme Facebook ont connu une croissance exponentielle (chaque utilisateur am√®ne d'autres utilisateurs).

### üîë Formules cl√©s

| Type | R√©currence | Formule explicite | Croissance |
|------|-----------|------------------|------------|
| **Arithm√©tique** | $u_{n+1} = u_n + r$ | $u_n = u_0 + n \times r$ | Lin√©aire (droite) |
| **G√©om√©trique** | $u_{n+1} = u_n \times q$ | $u_n = u_0 \times q^n$ | Exponentielle (courbe) |

### üåç Applications dans le monde r√©el

**Suites arithm√©tiques :**

**Suites g√©om√©triques :**
- Croissance des r√©seaux sociaux (effet viral)
- Int√©r√™ts compos√©s en finance
- Propagation d'√©pid√©mies

- Croissance de populations (bact√©ries, animaux)

### üí° Point important

Les ph√©nom√®nes de **croissance virale** ou **explosive** dans notre monde moderne (r√©seaux sociaux, vid√©os virales, technologies) suivent presque toujours des mod√®les **exponentiels** (suites g√©om√©triques), pas lin√©aires !
