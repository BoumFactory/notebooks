## üìö Partie 3 : Simulation et estimation

### ‚úçÔ∏è Exercice 3 : √âchantillon

√âcrivez une fonction `echantillon(n)` qui simule $n$ parties et renvoie la **fr√©quence** de victoires.

````python
from random import randint

def echantillon(n):
    gagne = 0
    
    # Compl√©tez ici
    # Pour i de 0 √† n-1 :
    #   Tirer 3 d√©s al√©atoires avec randint(1, 6)
    #   Si partie gagn√©e : incr√©menter gagne
    
    return gagne / n

# Tests
print("√âchantillon de 10 parties :", echantillon(10))
print("√âchantillon de 100 parties :", echantillon(100))
print("√âchantillon de 1000 parties :", echantillon(1000))
print("√âchantillon de 10000 parties :", echantillon(10000))
````

<summary>üí° Correction</summary>

```python
from random import randint

def echantillon(n):
    gagne = 0
    
    for i in range(n):
        d1 = randint(1, 6)
        d2 = randint(1, 6)
        d3 = randint(1, 6)
        gagne += partie(d1, d2, d3)
    
    return gagne / n
```

## üìö Partie 4 : Loi des grands nombres

### ‚úçÔ∏è Exercice 4 : Fluctuation

## üèÜ Challenge final : Intervalle de fluctuation

Pour un √©chantillon de taille $n$ et une probabilit√© $p$, on admet que dans environ 95% des cas, la fr√©quence $f$ v√©rifie :

$$p - \dfrac{1}{\sqrt{n}} \leq f \leq p + \dfrac{1}{\sqrt{n}}$$

**Objectif :** Simulez 100 √©chantillons de taille 1000 et comptez combien de fr√©quences sont dans cet intervalle.

````python
from math import sqrt
import matplotlib.pyplot as plt

n = 1000
p = 111 / 216
N = 100  # Nombre d'√©chantillons

mini = p - 1 / sqrt(n)
maxi = p + 1 / sqrt(n)

dans_intervalle = 0

plt.clf()

for repet in range(N):
    f = echantillon(n)
    
    if mini < f < maxi:
        plt.plot(repet, f, 'go')  # Vert si dans l'intervalle
        dans_intervalle += 1
    else:
        plt.plot(repet, f, 'ro')  # Rouge sinon

# Lignes de r√©f√©rence
plt.plot([0, N], [p, p], 'm--', label=f'p = {p:.4f}')
plt.plot([0, N], [mini, mini], 'k--', label=f'p - 1/‚àön')
plt.plot([0, N], [maxi, maxi], 'k--', label=f'p + 1/‚àön')

plt.xlabel('Num√©ro de l\'√©chantillon')
plt.ylabel('Fr√©quence')
plt.title(f'Fluctuation d\'√©chantillonnage (n={n}, N={N})')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nNombre d'√©chantillons dans l'intervalle : {dans_intervalle} / {N}")
print(f"Proportion : {dans_intervalle / N * 100:.1f}%")
````

<details>
<summary>üí° Interpr√©tation</summary>

On devrait observer environ **95% des points en vert** (dans l'intervalle).

C'est l'**intervalle de fluctuation** : il donne une zone dans laquelle la fr√©quence observ√©e a de grandes chances de se trouver.
</details>
