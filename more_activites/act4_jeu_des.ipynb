## üí° Introduction : Le jeu

### La r√®gle du jeu

√Ä la F√™te de la science, des √©tudiants organisent un jeu :

**Le joueur lance 3 d√©s** (rouge, vert, bleu). Si les trois valeurs obtenues permettent de former un **triangle non aplati**, il gagne. Sinon, il perd.

**Exemple :**
- üé≤ Lancer : (2, 3, 4) ‚Üí Triangle possible ‚úÖ **GAGN√â**
- üé≤ Lancer : (1, 2, 6) ‚Üí Triangle impossible ‚ùå **PERDU** (1+2 < 6)
- üé≤ Lancer : (1, 2, 3) ‚Üí Triangle aplati ‚ùå **PERDU** (1+2 = 3)

### üìê Rappel math√©matique : L'in√©galit√© triangulaire

On peut construire un triangle **non aplati** de c√¥t√©s $a$, $b$, $c$ si et seulement si :

$$\text{max}(a, b, c) < a + b + c - \text{max}(a, b, c)$$

Autrement dit : **le plus grand c√¥t√© doit √™tre strictement inf√©rieur √† la somme des deux autres.**

**Pourquoi cette formule ?**
- $a + b + c$ est la somme des trois c√¥t√©s
- $\text{max}(a, b, c)$ est le plus grand c√¥t√©
- $a + b + c - \text{max}(a, b, c)$ est donc la somme des deux autres c√¥t√©s

### üîç Deux m√©thodes pour calculer la probabilit√©

**Question :** Quelle est la probabilit√© de gagner √† ce jeu ?

#### M√©thode 1 : D√©nombrement exhaustif
On teste **tous les cas possibles** : (1,1,1), (1,1,2), ..., (6,6,6)
- Total de cas : $6 \times 6 \times 6 = 216$
- On compte combien de cas permettent de former un triangle
- $P(\text{gagner}) = \dfrac{\text{nombre de cas favorables}}{216}$

#### M√©thode 2 : Simulation
On simule $n$ parties (avec `random.randint`) et on observe la **fr√©quence** de victoires.
- Plus $n$ est grand, plus la fr√©quence se rapproche de la probabilit√© exacte
- C'est la **loi des grands nombres** !

```python
def partie(d1, d2, d3):
    grand = max(d1, d2, d3)
    somme_trois = d1 + d2 + d3
    somme_deux_autres = somme_trois - grand
    
    if grand < somme_deux_autres:
        return 1
    else:
        return 0
```

**Explications :**

```python
import random

def echantillon(n):
    gagne = 0
    
    for i in range(n):
        d1 = random.randint(1, 6)
        d2 = random.randint(1, 6)
```

## üèÜ Challenge final : Intervalle de fluctuation

### üí° Th√©orie

En statistiques, on admet que pour un √©chantillon de taille $n$ et une probabilit√© $p$, la fr√©quence observ√©e $f$ v√©rifie **dans environ 95% des cas** :

$$p - \dfrac{1}{\sqrt{n}} \leq f \leq p + \dfrac{1}{\sqrt{n}}$$

C'est l'**intervalle de fluctuation au seuil de 95%**.
