## üìö Partie 3 : Esp√©rance de gain

### ‚úçÔ∏è Exercice 3 : Estimation par simulation

Simulez 10000 parties et calculez le gain moyen.

````python
N = 10000
gains = [jouer() for i in range(N)]
gain_moyen = sum(gains) / N

print(f"Gain moyen observ√© sur {N} parties : {gain_moyen:.2f}‚Ç¨")

# Calcul th√©orique
esperance_theo = (1*100 + 5*20 + 10*5 + 84*0) / 100
print(f"Esp√©rance th√©orique : {esperance_theo:.2f}‚Ç¨")
````

### üìñ Calcul th√©orique

$$E(X) = \dfrac{1 \times 100 + 5 \times 20 + 10 \times 5 + 84 \times 0}{100} = \dfrac{250}{100} = 2{,}50‚Ç¨$$

**Conclusion :** Pour que le jeu soit √©quitable, le billet doit co√ªter **2,50‚Ç¨**.

## üèÜ Challenge final : Distribution des gains

Tracez un histogramme des gains sur 10000 parties.

````python
import matplotlib.pyplot as plt
from collections import Counter

plt.clf()

N = 10000
gains = [jouer() for i in range(N)]
compteur = Counter(gains)

# Histogramme
valeurs = sorted(compteur.keys())
frequences = [compteur[v]/N for v in valeurs]

plt.bar(valeurs, frequences, edgecolor='black')
plt.xlabel('Gain (‚Ç¨)')
plt.ylabel('Fr√©quence')
plt.title(f'Distribution des gains ({N} parties)')
plt.axhline(y=0.01, color='r', linestyle='--', label='1% (gros lot)')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.show()

print("\nDistribution observ√©e :")
for v in valeurs:
    print(f"  {v:3d}‚Ç¨ : {compteur[v]:4d} fois ({compteur[v]/N*100:.1f}%)")
````
