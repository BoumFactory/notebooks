## üí° Introduction : Qu'est-ce que l'esp√©rance math√©matique ?

### D√©finition intuitive

L'**esp√©rance math√©matique** d'une variable al√©atoire repr√©sente la **valeur moyenne** qu'on peut attendre si on r√©p√®te l'exp√©rience un tr√®s grand nombre de fois.

**Exemple simple :** Vous lancez un d√© √©quilibr√©. √Ä chaque lancer, vous gagnez le nombre indiqu√© en euros.
- Si vous obtenez 1, vous gagnez 1‚Ç¨
- Si vous obtenez 2, vous gagnez 2‚Ç¨
- ...
- Si vous obtenez 6, vous gagnez 6‚Ç¨

**Question :** Si vous jouez 6000 fois, combien gagnerez-vous en moyenne par partie ?

**R√©ponse intuitive :** Chaque face appara√Ætra environ 1000 fois, donc :
- Gain total ‚âà 1000√ó1 + 1000√ó2 + 1000√ó3 + 1000√ó4 + 1000√ó5 + 1000√ó6 = 21000‚Ç¨
- Gain moyen par partie ‚âà 21000 / 6000 = 3,5‚Ç¨

### D√©finition math√©matique

Si $X$ est une variable al√©atoire qui peut prendre les valeurs $x_1, x_2, ..., x_n$ avec les probabilit√©s $p_1, p_2, ..., p_n$, alors :

$$E(X) = x_1 \times p_1 + x_2 \times p_2 + ... + x_n \times p_n = \sum_{i=1}^{n} x_i \times p_i$$

**Pour le d√© :**
$$E(X) = 1 \times \frac{1}{6} + 2 \times \frac{1}{6} + 3 \times \frac{1}{6} + 4 \times \frac{1}{6} + 5 \times \frac{1}{6} + 6 \times \frac{1}{6} = \frac{21}{6} = 3{,}5$$

### Notion de jeu √©quitable

Un jeu est dit **√©quitable** si l'esp√©rance de gain est √©gale au prix pay√© pour jouer.

**Exemple :**
- Si un billet de loterie co√ªte 2‚Ç¨ et que l'esp√©rance de gain est de 2‚Ç¨ ‚Üí **jeu √©quitable**
- Si un billet co√ªte 2‚Ç¨ et que l'esp√©rance de gain est de 1,50‚Ç¨ ‚Üí **jeu d√©savantageux** (pour le joueur)
- Si un billet co√ªte 2‚Ç¨ et que l'esp√©rance de gain est de 3‚Ç¨ ‚Üí **jeu avantageux** (pour le joueur)

**Remarque :** Dans la r√©alit√©, les loteries sont toujours d√©savantageuses pour le joueur (sinon l'organisateur perdrait de l'argent !)

## üìö Situation : La loterie des 100 billets

Une loterie fonctionne ainsi :
- Il y a **100 billets** num√©rot√©s de 1 √† 100
- **Billet n¬∞1** : gagne 100‚Ç¨
- **Billets n¬∞2 √† n¬∞6** (5 billets) : gagnent 20‚Ç¨ chacun
- **Billets n¬∞7 √† n¬∞16** (10 billets) : gagnent 5‚Ç¨ chacun
- **Billets n¬∞17 √† n¬∞100** (84 billets) : ne gagnent rien (0‚Ç¨)

### ü§î Questions de r√©flexion

Avant de coder, r√©fl√©chissez √† ces questions :

**Question 1 :** Quelle est la probabilit√© de gagner 100‚Ç¨ ?

**Question 2 :** Quelle est la probabilit√© de gagner 20‚Ç¨ ?

**Question 3 :** Quel est le gain total distribu√© par la loterie ?

**Question 4 :** Si 100 personnes jouent (chacune prend un billet diff√©rent), quel sera le gain moyen par personne ?

<details>
<summary>üí° R√©ponses</summary>

**R√©ponse 1 :** $P(\text{gagner } 100‚Ç¨) = \dfrac{1}{100} = 0{,}01 = 1\%$

**R√©ponse 2 :** $P(\text{gagner } 20‚Ç¨) = \dfrac{5}{100} = 0{,}05 = 5\%$

**R√©ponse 3 :** Gain total = 1√ó100 + 5√ó20 + 10√ó5 + 84√ó0 = 100 + 100 + 50 = 250‚Ç¨

**R√©ponse 4 :** Gain moyen = 250 / 100 = 2,50‚Ç¨

Cette valeur de 2,50‚Ç¨ est l'**esp√©rance math√©matique** du jeu.
