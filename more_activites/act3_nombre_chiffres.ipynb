## üí° Introduction : Puissances de 10 et nombre de chiffres

### Observation fondamentale

Regardons les puissances de 10 :

| Nombre | Valeur | Nombre de chiffres |
|--------|--------|-------------------|
| $10^0$ | 1 | 1 |
| $10^1$ | 10 | 2 |
| $10^2$ | 100 | 3 |
| $10^3$ | 1000 | 4 |
| $10^k$ | 10...0 | $k+1$ |

**R√®gle importante :** $10^k$ a exactement $k+1$ chiffres.

### G√©n√©ralisation

Pour un nombre quelconque $n$ :

**Si $10^k \leq n < 10^{k+1}$, alors $n$ a exactement $k+1$ chiffres.**

### Exemples

| Nombre | Encadrement | Nombre de chiffres |
|--------|-------------|-------------------|
| 100 | $10^2 \leq 100 < 10^3$ | 3 |
| 999 | $10^2 \leq 999 < 10^3$ | 3 |
| 1024 | $10^3 \leq 1024 < 10^4$ | 4 |
| 9999 | $10^3 \leq 9999 < 10^4$ | 4 |

### ü§î Question de r√©flexion

On sait que $2^{10} = 1024 \approx 10^3$.

**Question :** Combien de chiffres a approximativement $2^{100} = (2^{10})^{10}$ ?

<details>
<summary>üí° R√©ponse</summary>

$2^{100} = (2^{10})^{10} \approx (10^3)^{10} = 10^{30}$

Donc $2^{100}$ a environ **31 chiffres** (car $10^{30}$ a 31 chiffres).

$$k \leq \log_{10}(n) < k+1$$

Donc : $k = \lfloor \log_{10}(n) \rfloor$ (partie enti√®re du logarithme)

**Nombre de chiffres = $\lfloor \log_{10}(n) \rfloor + 1$**
