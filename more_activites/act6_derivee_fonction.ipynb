## üí° Rappel math√©matique

La **d√©riv√©e** de $f$ en $a$ est d√©finie par :

$$f'(a) = \lim_{h \to 0} \dfrac{f(a+h) - f(a)}{h}$$

Le **taux d'accroissement** $\dfrac{f(a+h) - f(a)}{h}$ approxime $f'(a)$ quand $h$ est petit.
