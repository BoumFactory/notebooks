## üí° Introduction th√©orique

### Qu'est-ce qu'un taux d'accroissement ?

Le **taux d'accroissement** d'une fonction $f$ entre deux points $a$ et $a+h$ mesure la variation moyenne de $f$ sur cet intervalle :

$$\text{Taux d'accroissement} = \dfrac{f(a+h) - f(a)}{h}$$

**Interpr√©tation g√©om√©trique :** C'est la pente de la droite passant par les points $(a, f(a))$ et $(a+h, f(a+h))$ (appel√©e **s√©cante**).

### La d√©riv√©e : une limite

La **d√©riv√©e** de $f$ en $a$ est d√©finie comme la limite du taux d'accroissement quand $h$ tend vers 0 :

$$f'(a) = \lim_{h \to 0} \dfrac{f(a+h) - f(a)}{h}$$

**Interpr√©tation g√©om√©trique :** C'est la pente de la **tangente** √† la courbe de $f$ au point $(a, f(a))$.

### Exemples concrets

| Contexte | Fonction $f(t)$ | D√©riv√©e $f'(t)$ |
|----------|-----------------|------------------|
| Mouvement | Position | Vitesse instantan√©e |
| √âconomie | Co√ªt total | Co√ªt marginal |
| Physique | Charge √©lectrique | Intensit√© du courant |
| Biologie | Population | Taux de croissance |

### Approximation num√©rique

En pratique, pour **approximer** $f'(a)$ avec Python, on calcule :

$$f'(a) \approx \dfrac{f(a+h) - f(a)}{h} \text{ avec } h \text{ tr√®s petit}$$

**Principe cl√© :** Plus $h$ est petit, meilleure est l'approximation.

**Exemple :** Pour $f(x) = x^2$ en $a = 2$, on sait que $f'(2) = 4$.
- Avec $h = 0.1$ : $\dfrac{f(2.1) - f(2)}{0.1} = \dfrac{4.41 - 4}{0.1} = 4.1$
- Avec $h = 0.01$ : $\dfrac{f(2.01) - f(2)}{0.01} = \dfrac{4.0401 - 4}{0.01} = 4.01$
- Avec $h = 0.001$ : $\approx 4.001$

On observe la **convergence** vers 4 !

### üîë Formules √† retenir

**Taux d'accroissement :**
$$\tau = \dfrac{f(a+h) - f(a)}{h}$$

**D√©riv√©e :**
$$f'(a) = \lim_{h \to 0} \dfrac{f(a+h) - f(a)}{h}$$

**Approximation num√©rique :**
$$f'(a) \approx \dfrac{f(a+h) - f(a)}{h} \text{ avec } h \approx 10^{-4}$$

### üìä D√©riv√©es usuelles

| Fonction $f(x)$ | D√©riv√©e $f'(x)$ |
|-----------------|------------------|
| $x^n$ | $nx^{n-1}$ |
| $\sin(x)$ | $\cos(x)$ |
