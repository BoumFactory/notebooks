## üí° Introduction : Qu'est-ce que $\sqrt{2}$ ?

### Un nombre irrationnel

$\sqrt{2}$ est le nombre positif dont le carr√© vaut 2. C'est un **nombre irrationnel** : il poss√®de une infinit√© de d√©cimales non p√©riodiques.

$$\sqrt{2} = 1{,}41421356237309504880168872420969807856967187537694...$$

On ne peut donc pas l'√©crire exactement, mais on peut l'**approximer** avec une pr√©cision aussi grande qu'on veut.

### Lien avec une fonction

Consid√©rons la fonction $f(x) = x^2 - 2$.

**Question importante :** Que vaut $f(\sqrt{2})$ ?

<details>
<summary>üí° R√©ponse</summary>

$$f(\sqrt{2}) = (\sqrt{2})^2 - 2 = 2 - 2 = 0$$

**Donc : $\sqrt{2}$ est un z√©ro de la fonction $f$.**
</details>

### Principe de la m√©thode

Pour approximer $\sqrt{2}$, on va :
1. **Chercher le z√©ro** de la fonction $f(x) = x^2 - 2$
2. **Balayer** l'intervalle $[0, 2]$ par petits pas
3. **S'arr√™ter** quand $f(x)$ devient positif
4. **Encadrer** $\sqrt{2}$ entre la derni√®re valeur n√©gative et la premi√®re valeur positive

## üìö Partie 3 : Recherche du z√©ro par balayage

### üí≠ Principe du balayage

On va "balayer" l'intervalle $[0, 2]$ en avan√ßant par petits pas jusqu'√† ce que $f(x)$ devienne **positif ou nul**.

**Algorithme :**
1. Partir de $x = 0$
2. **Tant que** $f(x) < 0$ :
   - Avancer : $x \leftarrow x + \text{pas}$
3. S'arr√™ter d√®s que $f(x) \geq 0$

**Outil n√©cessaire :** Boucle `while` (pas `for` !) car on ne sait pas √† l'avance combien d'it√©rations sont n√©cessaires.

### ü§î Questions de r√©flexion

**Question 1 :** Pourquoi utiliser `while` et pas `for` ?

**Question 2 :** Que repr√©sente la derni√®re valeur de $x$ avant que $f(x)$ devienne positif ?

<details>
<summary>üí° R√©ponses</summary>

**R√©ponse 1 :** 
- `for` : on conna√Æt √† l'avance le nombre d'it√©rations
- `while` : on continue **tant qu'une condition est vraie**, sans savoir combien de fois
- Ici, on ne sait pas √† l'avance combien de pas il faut faire avant que $f(x) \geq 0$

**R√©ponse 2 :**
- C'est une **approximation par d√©faut** de $\sqrt{2}$
- Elle est l√©g√®rement inf√©rieure √† $\sqrt{2}$
</details>

La fonction `balayage` nous donne la **premi√®re valeur o√π $f(x) \geq 0$**.

**Id√©e :**
- Soit $x$ la valeur renvoy√©e par `balayage(precision)`

- Donc : $x - \text{precision} < \sqrt{2} \leq x$

**Largeur de l'encadrement :** exactement √©gale √† `precision` !

### ‚ú® Ce que vous avez appris

#### Concepts math√©matiques
- ‚úÖ **Approximation de nombres irrationnels** : $\sqrt{2}$ a une infinit√© de d√©cimales
- ‚úÖ **Z√©ro d'une fonction** : $\sqrt{2}$ est le z√©ro de $f(x) = x^2 - 2$
- ‚úÖ **M√©thode du balayage** : avancer par petits pas pour trouver un z√©ro
- ‚úÖ **Encadrement** : trouver un intervalle contenant la valeur cherch√©e
- ‚úÖ **Convergence** : l'erreur diminue quand la pr√©cision augmente

#### Comp√©tences Python
- ‚úÖ **Fonction simple** : `def f(x): return x**2 - 2`
- ‚úÖ **Boucle `for`** : pour tracer une courbe point par point
- ‚úÖ **Boucle `while`** : pour r√©p√©ter jusqu'√† ce qu'une condition soit remplie
- ‚úÖ **Manipulation de listes** : cr√©er, remplir, parcourir
- ‚úÖ **Visualisation** : tracer des courbes avec matplotlib
- ‚úÖ **Tuples** : renvoyer plusieurs valeurs `(inf, sup)`

### üîë Formules et concepts cl√©s

| Concept | Formule/Code | Explication |
|---------|--------------|-------------|
| Z√©ro de fonction | $f(\sqrt{2}) = 0$ | $\sqrt{2}$ annule $f(x) = x^2 - 2$ |
| Pr√©cision | $p = 10^{-n}$ | Pas de balayage |
| Boucle while | `while f(x) < 0:` | Continue tant que condition vraie |
| Encadrement | $(x-p, x]$ | Largeur = pr√©cision |

### üéØ Diff√©rence `for` vs `while`

| `for` | `while` |
|-------|--------|
| Nombre d'it√©rations **connu** | Nombre d'it√©rations **inconnu** |
| Parcourir une liste, un intervalle | R√©p√©ter jusqu'√† condition |
| Ex: tracer 100 points | Ex: chercher quand $f(x) \geq 0$ |
