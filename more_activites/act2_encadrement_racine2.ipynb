## üí° Introduction

√âcrivez une fonction `representation(a, b, N)` qui trace la courbe de $f$ sur $[a, b]$ avec $N$ points.

**M√©thode :**
1. Calculer le pas : $\text{pas} = \dfrac{b - a}{N}$
2. Pour $i$ de 0 √† $N$ :
   - Calculer $x = a + i \times \text{pas}$

## üìö Partie 2 : Approximation par d√©faut

### üìñ Principe du balayage

On "balaye" l'intervalle $[0, 2]$ en avan√ßant par petits pas jusqu'√† ce que $f(x)$ devienne positif.

La derni√®re valeur de $x$ pour laquelle $f(x) < 0$ est une **approximation par d√©faut** de $\sqrt{2}$.

### ‚úçÔ∏è Exercice 3 : Balayage simple

Compl√©tez le programme suivant pour trouver une approximation de $\sqrt{2}$ √† $10^{-2}$ pr√®s.

````python
a = 0
b = 2
precision = 0.01  # 10^-2
N = int((b - a) / precision)  # Nombre de subdivisions
pas = (b - a) / N

estimation = a

# Compl√©tez ici :
# Tant que f(estimation) < 0 :
#     estimation = estimation + pas

print(f"Estimation de ‚àö2 : {estimation}")
print(f"V√©rification : {estimation}¬≤ = {estimation**2}")
````

### üìñ Principe de l'encadrement

Si on s'arr√™te au premier $x$ tel que $f(x) \geq 0$, alors :
- Pour $x - \text{pas}$ : $f(x - \text{pas}) < 0$
- Pour $x$ : $f(x) \geq 0$

Donc : $x - \text{pas} \leq \sqrt{2} < x$

```python
def encadrer_racine2(n):
    a = 0
    b = 2
    precision = 10**(-n)
    N = int((b - a) / precision)
    pas = (b - a) / N
    
    x = a
    while f(x) < 0:
        x = x + pas
    
    borne_inf = x - pas
    borne_sup = x
    
    return borne_inf, borne_sup
```
