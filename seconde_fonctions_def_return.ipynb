## üìö Notion 5 : Fonctions avec conditions

### üìñ Cours

Certaines fonctions math√©matiques sont **d√©finies par morceaux**, c'est-√†-dire que la formule change selon la valeur de $x$.

**Exemple :** La valeur absolue
$$|x| = \begin{cases}
x & \text{si } x \geq 0 \\
-x & \text{si } x < 0
\end{cases}$$

**En Python :** On utilise `if` et `else`

### üéØ √Ä vous de coder !

**Fonction math√©matique :** La valeur absolue $|x|$

Traduisez cette fonction en Python, puis testez avec $|5|$, $|-3|$ et $|0|$.

<details>
<summary>üõ†Ô∏è Outils</summary>

- **Structure** : `if condition:` suivi de `return ...`, puis `else:` suivi de `return ...`
- **Condition** : `x >= 0` pour tester si $x$ est positif ou nul
- **Oppos√©** : `-x` pour obtenir l'oppos√© d'un nombre
</details>

````python
# D√©finissez la fonction valeur absolue
````

<details>
<summary>üí° Correction</summary>

```python
def valeur_absolue(x):
    if x >= 0:
        return x
    else:
        return -x

print("|5| =", valeur_absolue(5))    # 5
print("|-3| =", valeur_absolue(-3))  # 3
print("|0| =", valeur_absolue(0))    # 0
```
</details>

### Distance entre deux points

En math√©matiques, la distance entre deux points $A(x_1, y_1)$ et $B(x_2, y_2)$ est donn√©e par :
$$d(A, B) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$

√âcrivez une fonction `distance(x1, y1, x2, y2)` qui calcule cette distance.

**Testez avec :**
- $A(0, 0)$ et $B(3, 4)$ ‚Üí distance = 5
- $A(1, 2)$ et $B(4, 6)$ ‚Üí distance = 5

<details>
<summary>üõ†Ô∏è Outils</summary>

- **Racine carr√©e** : `a ** 0.5` ou `import math` puis `math.sqrt(a)`
- **Fonction √† 4 param√®tres** : `def distance(x1, y1, x2, y2):`
- **Formule** : `((x2-x1)**2 + (y2-y1)**2) ** 0.5`
</details>

````python
# √âcrivez votre code ici
````

<details>
<summary>üèÜ Correction</summary>

```python
# === SOLUTION ATTENDUE ===
def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    return (dx**2 + dy**2) ** 0.5

print("Distance A(0,0) √† B(3,4) :", distance(0, 0, 3, 4))  # 5.0
print("Distance A(1,2) √† B(4,6) :", distance(1, 2, 4, 6))  # 5.0
```

**Explication math√©matique :**
- Pour A(0,0) et B(3,4) : $d = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$
- C'est un triangle rectangle 3-4-5 (triplet pythagoricien)
