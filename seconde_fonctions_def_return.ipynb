## üí° Introduction

En math√©matiques, une fonction associe √† chaque nombre $x$ un nombre $f(x)$.

Par exemple, si $f(x) = 2x + 3$, alors :
- $f(0) = 3$
- $f(1) = 5$
- $f(2) = 7$

En Python, on peut **d√©finir une fonction** pour automatiser ce calcul. Plut√¥t que de calculer √† la main chaque image, on √©crit une fois la formule et Python calcule pour nous !

**Avantages** :
- ‚úÖ On √©crit la formule une seule fois
- ‚úÖ On peut calculer autant d'images qu'on veut
- ‚úÖ Le code est plus lisible et r√©utilisable
- ‚úÖ On √©vite les erreurs de calcul

## üìö Notion 1 : D√©finir une fonction simple

### üìñ Cours

Pour d√©finir une fonction en Python, on utilise le mot-cl√© **`def`** (pour "define" en anglais).

**Syntaxe :**
```python
def nom_fonction(parametre):
    return resultat
```

**√âl√©ments cl√©s :**
- `def` : mot-cl√© pour d√©finir une fonction
- `nom_fonction` : le nom que vous donnez √† votre fonction
- `parametre` : la variable d'entr√©e (correspond au $x$ en maths)
- `return` : renvoie le r√©sultat du calcul
- **Indentation** : obligatoire apr√®s les `:` (utilisez Tab)

**Exemple r√©solu :**

````python
# D√©finition de la fonction f(x) = 2x + 3
def f(x):
    return 2 * x + 3

# Utilisation de la fonction
print("f(0) =", f(0))  # 3
print("f(1) =", f(1))  # 5
print("f(5) =", f(5))  # 13
````

## üìö Notion 2 : Fonction carr√©e et puissances

‚ö†Ô∏è **Attention** : Ne pas confondre `x ** 2` (puissance) et `x * 2` (multiplication) !

## üìö Notion 3 : Fonctions polynomiales

Une **fonction polynomiale** combine plusieurs termes avec diff√©rentes puissances de $x$.

**Ordre d'√©valuation en Python :**
1. Puissances : `**`

````python
# Fonction polynomiale : h(x) = x¬≤ - 3x + 2
def h(x):
    return x**2 - 3*x + 2

# Calculons quelques images
print("h(0) =", h(0))  # 2
print("h(1) =", h(1))  # 0
print("h(2) =", h(2))  # 0
print("h(3) =", h(3))  # 2
````

## üìö Notion 5 : Fonctions avec conditions

### üìñ Cours

Certaines fonctions math√©matiques sont **d√©finies par morceaux**, c'est-√†-dire que la formule change selon la valeur de $x$.

**Exemple :** La valeur absolue
$$|x| = \begin{cases}
x & \text{si } x \geq 0 \\
-x & \text{si } x < 0
\end{cases}$$

**En Python :** On utilise `if` et `else`

**Exemple r√©solu :**

````python
# Valeur absolue
def valeur_absolue(x):
    if x >= 0:
        return x
    else:
        return -x

# Tests
print("|5| =", valeur_absolue(5))    # 5
print("|-3| =", valeur_absolue(-3))  # 3
print("|0| =", valeur_absolue(0))    # 0
````
