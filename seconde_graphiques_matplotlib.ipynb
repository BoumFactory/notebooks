## 2ï¸âƒ£ Utiliser numpy.linspace() pour plus de points

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– DÃ©finition :</strong> <code>numpy.linspace(a, b, n)</code> gÃ©nÃ¨re <strong>n</strong> valeurs rÃ©guliÃ¨rement espacÃ©es entre <strong>a</strong> et <strong>b</strong> (inclus).

### ğŸ”§ Pourquoi utiliser linspace ?

- Pour obtenir une courbe **lisse**, on a besoin de beaucoup de points
- Ã‰crire manuellement 100 valeurs est impossible
- `linspace()` gÃ©nÃ¨re ces valeurs automatiquement

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Tracer f(x) = xÂ² avec 100 points
import matplotlib.pyplot as plt
import numpy as np

# GÃ©nÃ©rer 100 valeurs de x entre -3 et 3
x = np.linspace(-3, 3, 100)

# Calculer les y correspondants
y = x**2

# Tracer
plt.plot(x, y)
plt.show()
````

## 3ï¸âƒ£ Personnaliser un graphique

<div style="background-color: #fff9c4; padding: 15px; border-radius: 10px; border-left: 5px solid #FFC107;">
    <strong>ğŸ¯ Objectif :</strong> Ajouter un titre, des lÃ©gendes d'axes, une grille et personnaliser les couleurs.
</div>

### ğŸ”§ Fonctions principales

- `plt.title("Titre")` : ajoute un titre
- `plt.xlabel("Axe x")` : lÃ©gende de l'axe horizontal
- `plt.ylabel("Axe y")` : lÃ©gende de l'axe vertical
- `plt.grid(True)` : affiche une grille
- `plt.axhline(0, color='black', linewidth=0.8)` : axe horizontal y=0
- `plt.axvline(0, color='black', linewidth=0.8)` : axe vertical x=0

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Graphique complet de f(x) = xÂ² - 4
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4, 200)
y = x**2 - 4

plt.figure(figsize=(8, 6))  # Taille du graphique
plt.plot(x, y, color='blue', linewidth=2, label='f(x) = xÂ² - 4')

# Personnalisation
plt.title("Courbe reprÃ©sentative de f(x) = xÂ² - 4", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.axhline(0, color='black', linewidth=0.8)  # Axe des x
plt.axvline(0, color='black', linewidth=0.8)  # Axe des y
plt.grid(True, alpha=0.3)  # Grille semi-transparente
plt.legend()  # Afficher la lÃ©gende

plt.show()
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li><code>figsize=(8, 6)</code> dÃ©finit la taille en pouces</li>
        <li><code>label</code> permet d'identifier la courbe dans la lÃ©gende</li>
        <li><code>alpha=0.3</code> rend la grille semi-transparente</li>
        <li><code>plt.legend()</code> affiche la boÃ®te de lÃ©gende</li>
    </ul>
</div>

## 4ï¸âƒ£ Tracer plusieurs courbes sur le mÃªme graphique

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– Astuce :</strong> On peut appeler <code>plt.plot()</code> plusieurs fois avant <code>plt.show()</code> pour superposer des courbes.
</div>

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Comparer deux fonctions
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 200)
y1 = x**2
y2 = x**3

plt.figure(figsize=(8, 6))
plt.plot(x, y1, color='blue', label='f(x) = xÂ²')
plt.plot(x, y2, color='red', label='g(x) = xÂ³')

plt.title("Comparaison de xÂ² et xÂ³")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````

    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li>Chaque <code>plt.plot()</code> ajoute une courbe au mÃªme graphique</li>
        <li>La lÃ©gende permet de distinguer les courbes</li>
