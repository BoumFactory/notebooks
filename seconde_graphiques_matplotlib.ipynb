## üìö Notion 2 : Utiliser numpy.linspace() pour plus de points

### üìñ Cours

Pour obtenir une **courbe lisse**, on a besoin de beaucoup de points. La fonction `numpy.linspace()` g√©n√®re automatiquement des valeurs r√©guli√®rement espac√©es.

**Syntaxe :**
```python
import numpy as np
x = np.linspace(a, b, n)  # n valeurs entre a et b (inclus)
```

**Avantage :** On peut calculer directement `y = x**2` pour tous les x (op√©ration vectorielle).

**Exemple r√©solu :**

````python
# Tracer f(x) = x¬≤ avec 100 points
import matplotlib.pyplot as plt
import numpy as np

plt.clf()  # Nettoyer les graphiques pr√©c√©dents

# G√©n√©rer 100 valeurs de x entre -3 et 3
x = np.linspace(-3, 3, 100)

# Calculer les y correspondants
y = x**2

# Tracer
plt.plot(x, y)
plt.show()
````

## üìö Notion 3 : Ajouter un titre et des labels

### üìñ Cours

Pour rendre un graphique **professionnel et lisible**, on ajoute :
- Un titre avec `plt.title("...")`
- Une l√©gende pour l'axe horizontal avec `plt.xlabel("...")`
- Une l√©gende pour l'axe vertical avec `plt.ylabel("...")`

**Exemple r√©solu :**

````python
# Graphique avec titre et labels
import matplotlib.pyplot as plt
import numpy as np

plt.clf()  # Nettoyer les graphiques pr√©c√©dents

x = np.linspace(-4, 4, 100)
y = x**2 - 4

plt.plot(x, y)
plt.title("Courbe de f(x) = x¬≤ - 4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
````

## üìö Notion 4 : Ajouter une grille et des axes

### üìñ Cours

Pour am√©liorer la **lisibilit√©** du graphique, on peut ajouter :
- Une **grille** avec `plt.grid(True)`
- Les **axes x=0 et y=0** avec `plt.axhline(0)` et `plt.axvline(0)`

**Exemple r√©solu :**

````python
# Graphique avec grille et axes
import matplotlib.pyplot as plt
import numpy as np

plt.clf()  # Nettoyer les graphiques pr√©c√©dents

x = np.linspace(-3, 3, 100)
y = x**2 - 2

plt.plot(x, y, color='blue')
plt.title("f(x) = x¬≤ - 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)  # Grille
plt.axhline(0, color='black', linewidth=0.8)  # Axe y=0
plt.axvline(0, color='black', linewidth=0.8)  # Axe x=0
plt.show()
````

## üìö Notion 5 : Tracer plusieurs courbes

### üìñ Cours

On peut **superposer plusieurs courbes** sur le m√™me graphique en appelant `plt.plot()` plusieurs fois avant `plt.show()`.

**Pour diff√©rencier les courbes :**
- Utilisez des **couleurs** diff√©rentes avec `color=`
- Ajoutez une **l√©gende** avec `label=` dans `plot()` puis `plt.legend()`

**Exemple r√©solu :**

````python
# Comparer deux fonctions
import matplotlib.pyplot as plt
import numpy as np

plt.clf()  # Nettoyer les graphiques pr√©c√©dents

x = np.linspace(-3, 3, 100)
y1 = x**2
y2 = x**3

plt.plot(x, y1, color='blue', label='f(x) = x¬≤')
plt.plot(x, y2, color='red', label='g(x) = x¬≥')

plt.title("Comparaison de x¬≤ et x¬≥")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()  # Affiche la l√©gende
plt.show()
````
