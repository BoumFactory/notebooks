## üìö Notion 1 : Cr√©er une classe simple

### üìñ Cours

Une **classe** se d√©finit avec le mot-cl√© `class`. Elle contient :
- Un **constructeur** `__init__` : initialise les attributs de l'objet
- Des **attributs** : les donn√©es de l'objet (stock√©es avec `self.nom_attribut`)
- Des **m√©thodes** : les fonctions de l'objet

**Structure de base :**
```python
class NomDeLaClasse:
    def __init__(self, parametre1, parametre2):
        self.attribut1 = parametre1
        self.attribut2 = parametre2
    
    def ma_methode(self):
        return self.attribut1 + self.attribut2
```

**Le mot-cl√© `self` :** Fait r√©f√©rence √† l'objet lui-m√™me (comme "je" en fran√ßais).

**Exemple :** Une classe Point pour repr√©senter un point du plan.

````python
class Point:
    """Classe repr√©sentant un point dans le plan cart√©sien."""
    
    def __init__(self, nom, x, y):
        """Constructeur : initialise un point avec son nom et ses coordonn√©es."""
        self.nom = nom.upper()  # En majuscules
        self.x = x
        self.y = y
    
    def __str__(self):
        """Renvoie une repr√©sentation textuelle du point."""
        return f"{self.nom}({self.x} ; {self.y})"
    
    def distance_origine(self):
        """Calcule la distance du point √† l'origine."""
        return (self.x**2 + self.y**2)**0.5

# Cr√©er un point A de coordonn√©es (3, 4)
A = Point("A", 3, 4)
print(A)
print("Abscisse de A :", A.x)
print("Distance √† l'origine :", A.distance_origine())
````

## üìö Notion 2 : Ajouter des m√©thodes avanc√©es

### üìñ Cours

On peut ajouter des m√©thodes qui prennent d'**autres objets** en param√®tre.

**Exemple :** M√©thode pour calculer la distance entre deux points.

$$d(A, B) = \sqrt{(x_B - x_A)^2 + (y_B - y_A)^2}$$

````python
class Point:
    def __init__(self, nom, x, y):
        self.nom = nom.upper()
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"{self.nom}({self.x} ; {self.y})"
    
    def distance(self, autre_point):
        """Calcule la distance entre ce point et un autre."""
        dx = self.x - autre_point.x
        dy = self.y - autre_point.y
        return (dx**2 + dy**2)**0.5
    
    def milieu(self, autre_point, nom_milieu):
        """Renvoie le milieu de ce point et d'un autre."""
        x_milieu = (self.x + autre_point.x) / 2
        y_milieu = (self.y + autre_point.y) / 2
        return Point(nom_milieu, x_milieu, y_milieu)

# Cr√©er deux points
A = Point("A", 1, 2)
B = Point("B", 4, 6)

print("Distance AB :", A.distance(B))
print("Milieu de [AB] :", A.milieu(B, "M"))
````

## üìö Notion 3 : Les m√©thodes sp√©ciales (op√©rateurs)

### üìñ Cours

Python permet de **surcharger les op√©rateurs** (`+`, `-`, `*`, etc.) avec des m√©thodes sp√©ciales :

| M√©thode | Op√©ration | Exemple |
|---------|-----------|----------|
| `__init__(self, ...)` | Constructeur | `Point("A", 3, 4)` |
| `__str__(self)` | Conversion en cha√Æne | `print(point)` |
| `__add__(self, autre)` | Addition `+` | `a + b` |
| `__sub__(self, autre)` | Soustraction `-` | `a - b` |
| `__mul__(self, autre)` | Multiplication `*` | `a * b` |
| `__eq__(self, autre)` | √âgalit√© `==` | `a == b` |

**Exemple :** Classe Vecteur avec addition et multiplication par un scalaire.

````python
class Vecteur:
    """Classe repr√©sentant un vecteur du plan."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Vecteur({self.x}, {self.y})"
    
    def __add__(self, autre):
        """Addition de deux vecteurs avec l'op√©rateur +."""
        return Vecteur(self.x + autre.x, self.y + autre.y)
    
    def __mul__(self, scalaire):
        """Multiplication par un scalaire avec l'op√©rateur *."""
        return Vecteur(self.x * scalaire, self.y * scalaire)
    
    def norme(self):
        """Calcule la norme du vecteur."""
        return (self.x**2 + self.y**2)**0.5

# Test
u = Vecteur(3, 4)
v = Vecteur(1, 2)

print("u =", u)
print("v =", v)
print("u + v =", u + v)       # Utilise __add__
print("2 * u =", u * 2)       # Utilise __mul__
print("Norme de u :", u.norme())
````

## üìö Notion 4 : Cr√©er une classe Fraction

### üìñ Cours

Une fraction $\dfrac{a}{b}$ est d√©finie par un **num√©rateur** et un **d√©nominateur**.

**Op√©rations :**
- Addition : $\dfrac{a}{b} + \dfrac{c}{d} = \dfrac{ad + bc}{bd}$
- Soustraction : $\dfrac{a}{b} - \dfrac{c}{d} = \dfrac{ad - bc}{bd}$
- Simplification : diviser num√©rateur et d√©nominateur par leur PGCD

**PGCD (algorithme d'Euclide) :**

````python
def pgcd(a, b):
    """Calcule le PGCD de a et b par l'algorithme d'Euclide."""
    a, b = abs(a), abs(b)
    while b != 0:
        a, b = b, a % b
    return a

# Test
print("PGCD(12, 8) =", pgcd(12, 8))  # 4
print("PGCD(15, 25) =", pgcd(15, 25))  # 5
````

**Exemple d'une classe Fraction simple :**

````python
class Fraction:
    def __init__(self, numerateur, denominateur):
        self.num = numerateur
        self.den = denominateur
    
    def __str__(self):
        return f"{self.num}/{self.den}"
    
    def simplifier(self):
        """Simplifie la fraction en place."""
        diviseur = pgcd(self.num, self.den)
        self.num = self.num // diviseur
        self.den = self.den // diviseur

# Test
f = Fraction(6, 8)
print("Avant :", f)
f.simplifier()
print("Apr√®s :", f)  # 3/4
````

class Affine:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def __str__(self):
        if self.b >= 0:
            return f"f(x) = {self.a}x + {self.b}"
        else:
            return f"f(x) = {self.a}x - {abs(self.b)}"
    
    def evaluer(self, x):
        return self.a * x + self.b
    
    def zero(self):
        return -self.b / self.a

# Test

class Segment:
    def __init__(self, point1, point2):

        self.point2 = point2

class Triangle:
