## üí° Introduction

En g√©om√©trie, le **produit scalaire** est une op√©ration fondamentale entre deux vecteurs. Il permet de :
- Calculer des angles entre vecteurs

### üìê Rappels math√©matiques

Pour deux vecteurs $\vec{u}(x_1, y_1)$ et $\vec{v}(x_2, y_2)$ :

**Produit scalaire :**
$$\vec{u} \cdot \vec{v} = x_1 \times x_2 + y_1 \times y_2$$

**Norme (longueur) :**
$$\|\vec{u}\| = \sqrt{x_1^2 + y_1^2}$$

**Orthogonalit√© :**
$$\vec{u} \perp \vec{v} \Leftrightarrow \vec{u} \cdot \vec{v} = 0$$

import math

def norme(u):
    """Calcule la norme d'un vecteur"""
    return math.sqrt(u[0]**2 + u[1]**2)

def normaliser(u):
    """Retourne le vecteur unitaire de m√™me direction"""
    n = norme(u)
    return (u[0]/n, u[1]/n)

# Calcul des normes

for v in vecteurs:
    print(f"||{v}|| = {norme(v):.4f}")
# R√©sultats : 10, 13, 1.4142, 5

# === BONUS (non demand√©) ===
# Normalisation et v√©rification

u_norm = normaliser(u)
print(f"\nNormalisation de {u} : ({u_norm[0]:.4f}, {u_norm[1]:.4f})")
print(f"V√©rification : norme = {norme(u_norm):.4f}")  # 1.0000

## üìö Notion 5 : Calculer l'angle entre deux vecteurs

### üìñ Cours

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>üìê Formule :</strong> Pour deux vecteurs $\vec{u}$ et $\vec{v}$ non nuls, l'angle $\theta$ entre eux est donn√© par :
    $$\cos(\theta) = \dfrac{\vec{u} \cdot \vec{v}}{\|\vec{u}\| \times \|\vec{v}\|}$$
</div>

**Donc :**
$$\theta = \arccos\left(\dfrac{\vec{u} \cdot \vec{v}}{\|\vec{u}\| \times \|\vec{v}\|}\right)$$

**En Python :**
- `math.acos(x)` retourne l'arc cosinus en **radians**
- `math.degrees(rad)` convertit des radians en degr√©s

**Cas particuliers :**
- Si $\theta = 0¬∞$ : vecteurs colin√©aires de m√™me sens
- Si $\theta = 90¬∞$ : vecteurs orthogonaux
- Si $\theta = 180¬∞$ : vecteurs colin√©aires de sens oppos√©s

````python
# ‚úÖ EXEMPLE RESOLU : Calcul d'angle
import math

def angle_degres(u, v):
    """Calcule l'angle en degr√©s entre deux vecteurs"""
    ps = u[0] * v[0] + u[1] * v[1]
    nu = math.sqrt(u[0]**2 + u[1]**2)
    nv = math.sqrt(v[0]**2 + v[1]**2)
    
    cos_theta = ps / (nu * nv)
    # Correction des erreurs d'arrondi
    cos_theta = max(-1, min(1, cos_theta))
    
    theta_rad = math.acos(cos_theta)
    return math.degrees(theta_rad)

# Tests
print(f"Angle entre (1,0) et (0,1) : {angle_degres((1,0), (0,1)):.1f}¬∞")  # 90¬∞
print(f"Angle entre (1,0) et (1,1) : {angle_degres((1,0), (1,1)):.1f}¬∞")  # 45¬∞
````
