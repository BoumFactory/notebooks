## 3ï¸âƒ£ VÃ©rifier l'orthogonalitÃ©

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– PropriÃ©tÃ© :</strong> Deux vecteurs $\vec{u}$ et $\vec{v}$ sont <strong>orthogonaux</strong> (perpendiculaires) si et seulement si :
    $$\vec{u} \cdot \vec{v} = 0$$
</div>

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : VÃ©rifier si deux vecteurs sont orthogonaux

def produit_scalaire(u, v):
    return u[0] * v[0] + u[1] * v[1]

def sont_orthogonaux(u, v):
    """VÃ©rifie si deux vecteurs sont orthogonaux"""
    return produit_scalaire(u, v) == 0

# Tests
u1 = (3, 4)
v1 = (-4, 3)  # Orthogonal Ã  u1

u2 = (1, 2)
v2 = (3, 1)  # Non orthogonal Ã  u2

print("Test 1 :")
print(f"uâƒ— = {u1}, vâƒ— = {v1}")
print(f"uâƒ— Â· vâƒ— = {produit_scalaire(u1, v1)}")
print(f"Orthogonaux ? {sont_orthogonaux(u1, v1)}")

print("\nTest 2 :")
print(f"uâƒ— = {u2}, vâƒ— = {v2}")
print(f"uâƒ— Â· vâƒ— = {produit_scalaire(u2, v2)}")
print(f"Orthogonaux ? {sont_orthogonaux(u2, v2)}")
````

## 4ï¸âƒ£ Calculer la norme d'un vecteur

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– DÃ©finition :</strong> La <strong>norme</strong> (longueur) d'un vecteur $\vec{u}(x, y)$ est :
    $$\|\vec{u}\| = \sqrt{x^2 + y^2}$$
</div>

### ğŸ”— Lien avec le produit scalaire

On a : $\|\vec{u}\|^2 = \vec{u} \cdot \vec{u}$

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Calculer la norme d'un vecteur
import math

def norme(u):
    """Calcule la norme d'un vecteur"""
    return math.sqrt(u[0]**2 + u[1]**2)

# Tests
vecteurs = [(3, 4), (1, 0), (5, 12), (-3, -4)]

print("Calcul de normes :\n")
for v in vecteurs:
    n = norme(v)
    print(f"||{v}|| = {n:.2f}")

# VÃ©rification avec le produit scalaire
u = (3, 4)
print(f"\nVÃ©rification pour {u} :")
print(f"||uâƒ—||Â² = {norme(u)**2}")
print(f"uâƒ— Â· uâƒ— = {u[0]**2 + u[1]**2}")
````
