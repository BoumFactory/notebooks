### ğŸ“ Rappels mathÃ©matiques

Pour deux vecteurs $\vec{u}(x_1, y_1)$ et $\vec{v}(x_2, y_2)$ :

**Produit scalaire :**
$$\vec{u} \cdot \vec{v} = x_1 \times x_2 + y_1 \times y_2$$

**Norme (longueur) :**
$$\|\vec{u}\| = \sqrt{x_1^2 + y_1^2}$$

**OrthogonalitÃ© :**
$$\vec{u} \perp \vec{v} \Leftrightarrow \vec{u} \cdot \vec{v} = 0$$

## ğŸ“š Notion 2 : Calculer le produit scalaire (formule algÃ©brique)

### ğŸ“– Cours

<div style="background-color: #fff9c4; padding: 15px; border-radius: 10px; border-left: 5px solid #FFC107;">
    <strong>ğŸ“ Formule :</strong> Pour deux vecteurs $\vec{u}(x_1, y_1)$ et $\vec{v}(x_2, y_2)$, le produit scalaire est :
    $$\vec{u} \cdot \vec{v} = x_1 \times x_2 + y_1 \times y_2$$
</div>

**ImplÃ©mentation Python :**
```python
def produit_scalaire(u, v):
    return u[0] * v[0] + u[1] * v[1]
```

**PropriÃ©tÃ©s importantes :**
- **CommutativitÃ©** : $\vec{u} \cdot \vec{v} = \vec{v} \cdot \vec{u}$
- **LinÃ©aritÃ©** : $\vec{u} \cdot (k\vec{v}) = k(\vec{u} \cdot \vec{v})$
- Le produit scalaire peut Ãªtre **positif, nÃ©gatif ou nul**

**InterprÃ©tation gÃ©omÃ©trique :**
- Si $\vec{u} \cdot \vec{v} > 0$ : angle aigu entre les vecteurs
- Si $\vec{u} \cdot \vec{v} < 0$ : angle obtus entre les vecteurs
- Si $\vec{u} \cdot \vec{v} = 0$ : vecteurs perpendiculaires (orthogonaux)

````python
# âœ… EXEMPLE COMPLET : Calculer des produits scalaires

def produit_scalaire(u, v):
    """Calcule le produit scalaire de deux vecteurs du plan"""
    return u[0] * v[0] + u[1] * v[1]

# Tests avec diffÃ©rents vecteurs
print("ğŸ“Š Calculs de produits scalaires\n")

# Exemple 1
u = (3, 4)
v = (1, -2)
ps = produit_scalaire(u, v)

print(f"uâƒ— = {u}, vâƒ— = {v}")
print(f"Calcul : {u[0]} Ã— {v[0]} + {u[1]} Ã— {v[1]}")
print(f"       = {u[0]*v[0]} + {u[1]*v[1]}")
print(f"       = {ps}")

# Exemple 2
print("\n" + "-"*40)
a = (2, 5)
b = (3, -1)
print(f"aâƒ— Â· bâƒ— = {produit_scalaire(a, b)}")

# VÃ©rification de la commutativitÃ©
print("\nğŸ”„ VÃ©rification de la commutativitÃ© :")
print(f"uâƒ— Â· vâƒ— = {produit_scalaire(u, v)}")
print(f"vâƒ— Â· uâƒ— = {produit_scalaire(v, u)}")
print(f"Ã‰gaux ? {produit_scalaire(u, v) == produit_scalaire(v, u)}")
````

## ğŸ“š Notion 3 : VÃ©rifier l'orthogonalitÃ© de deux vecteurs

### ğŸ“– Cours

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“ ThÃ©orÃ¨me :</strong> Deux vecteurs $\vec{u}$ et $\vec{v}$ sont <strong>orthogonaux</strong> (perpendiculaires) si et seulement si :
    $$\vec{u} \cdot \vec{v} = 0$$
</div>

**Notation :** On Ã©crit $\vec{u} \perp \vec{v}$ pour dire que $\vec{u}$ et $\vec{v}$ sont orthogonaux.

**ImplÃ©mentation Python :**
```python
def sont_orthogonaux(u, v):
    return produit_scalaire(u, v) == 0
```

**Application pratique :**
Pour trouver un vecteur orthogonal Ã  $\vec{u}(a, b)$, on peut prendre :
- $\vec{v}(-b, a)$ ou
- $\vec{v}(b, -a)$

En effet : $\vec{u} \cdot \vec{v} = a \times (-b) + b \times a = -ab + ab = 0$ âœ“

````python
# âœ… EXEMPLE COMPLET : Tester l'orthogonalitÃ©

def produit_scalaire(u, v):
    return u[0] * v[0] + u[1] * v[1]

def sont_orthogonaux(u, v):
    """VÃ©rifie si deux vecteurs sont orthogonaux"""
    return produit_scalaire(u, v) == 0

# Tests avec diffÃ©rents couples
print("ğŸ“Š Test d'orthogonalitÃ©\n")

couples = [
    ((3, 4), (-4, 3)),    # Orthogonaux
    ((1, 0), (0, 1)),     # Vecteurs de base (orthogonaux)
    ((2, 5), (5, -2)),    # Orthogonaux
    ((1, 2), (3, 1))      # Non orthogonaux
]

for u, v in couples:
    ps = produit_scalaire(u, v)
    ortho = "âœ… OUI" if sont_orthogonaux(u, v) else "âŒ NON"
    print(f"{u} Â· {v} = {ps:3} â†’ Orthogonaux : {ortho}")

# Trouver un vecteur orthogonal
print("\nğŸ”§ Construire un vecteur orthogonal :")
u = (5, 3)
v_ortho = (-u[1], u[0])  # (-3, 5)
print(f"uâƒ— = {u}")
print(f"vâƒ— orthogonal = {v_ortho}")
print(f"VÃ©rification : uâƒ— Â· vâƒ— = {produit_scalaire(u, v_ortho)}")
````

## ğŸ“š Notion 4 : Calculer la norme d'un vecteur

### ğŸ“– Cours

<div style="background-color: #fff9c4; padding: 15px; border-radius: 10px; border-left: 5px solid #FFC107;">
    <strong>ğŸ“ DÃ©finition :</strong> La <strong>norme</strong> (longueur) d'un vecteur $\vec{u}(x, y)$ est :
    $$\|\vec{u}\| = \sqrt{x^2 + y^2}$$
</div>

**Lien avec le produit scalaire :**
$$\|\vec{u}\|^2 = \vec{u} \cdot \vec{u} = x^2 + y^2$$

**ImplÃ©mentation Python :**
```python
import math

def norme(u):
    return math.sqrt(u[0]**2 + u[1]**2)
```

**Vecteur unitaire :** Un vecteur est **unitaire** si $\|\vec{u}\| = 1$.

Pour **normaliser** un vecteur (obtenir un vecteur unitaire de mÃªme direction) :
$$\vec{u}_{normalisÃ©} = \dfrac{\vec{u}}{\|\vec{u}\|} = \left(\dfrac{x}{\|\vec{u}\|}, \dfrac{y}{\|\vec{u}\|}\right)$$

````python
# âœ… EXEMPLE COMPLET : Calcul de normes et normalisation
import math

def norme(u):
    """Calcule la norme d'un vecteur"""
    return math.sqrt(u[0]**2 + u[1]**2)

def normaliser(u):
    """Retourne le vecteur unitaire de mÃªme direction"""
    n = norme(u)
    return (u[0]/n, u[1]/n)

# Tests avec des triplets pythagoriciens classiques
print("ğŸ“Š Calcul de normes\n")

vecteurs = [(3, 4), (5, 12), (8, 15), (1, 0), (1, 1)]

for v in vecteurs:
    n = norme(v)
    print(f"||{v}|| = {n:.4f}")

# Normalisation
print("\nğŸ”§ Normalisation (vecteur unitaire) :\n")
u = (3, 4)
u_norm = normaliser(u)

print(f"Vecteur original : {u}")
print(f"Norme : {norme(u)}")
print(f"Vecteur normalisÃ© : ({u_norm[0]:.4f}, {u_norm[1]:.4f})")
print(f"Norme du normalisÃ© : {norme(u_norm):.4f}")
````

## ğŸ“š Notion 5 : Calculer l'angle entre deux vecteurs

### ğŸ“– Cours

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“ Formule :</strong> Pour deux vecteurs $\vec{u}$ et $\vec{v}$ non nuls, l'angle $\theta$ entre eux est donnÃ© par :
    $$\cos(\theta) = \dfrac{\vec{u} \cdot \vec{v}}{\|\vec{u}\| \times \|\vec{v}\|}$$
</div>

**Donc :**
$$\theta = \arccos\left(\dfrac{\vec{u} \cdot \vec{v}}{\|\vec{u}\| \times \|\vec{v}\|}\right)$$

**ImplÃ©mentation Python :**
```python
import math

def angle_degres(u, v):
    ps = produit_scalaire(u, v)
    nu = norme(u)
    nv = norme(v)
    
    cos_theta = ps / (nu * nv)
    theta_rad = math.acos(cos_theta)
    theta_deg = math.degrees(theta_rad)
    
    return theta_deg
```

**Cas particuliers :**
- Si $\theta = 0Â°$ : vecteurs colinÃ©aires de mÃªme sens
- Si $\theta = 90Â°$ : vecteurs orthogonaux
- Si $\theta = 180Â°$ : vecteurs colinÃ©aires de sens opposÃ©s

**ğŸ’¡ Note :** `math.acos()` retourne un angle en **radians**, on utilise `math.degrees()` pour convertir en degrÃ©s.

````python
# âœ… EXEMPLE COMPLET : Calculer des angles
import math

def produit_scalaire(u, v):
    return u[0] * v[0] + u[1] * v[1]

def norme(u):
    return math.sqrt(u[0]**2 + u[1]**2)

def angle_degres(u, v):
    """Calcule l'angle en degrÃ©s entre deux vecteurs"""
    ps = produit_scalaire(u, v)
    nu = norme(u)
    nv = norme(v)
    
    cos_theta = ps / (nu * nv)
    # Gestion des erreurs d'arrondi
    cos_theta = max(-1, min(1, cos_theta))
    
    theta_rad = math.acos(cos_theta)
    theta_deg = math.degrees(theta_rad)
    
    return theta_deg

# Tests avec des cas classiques
print("ğŸ“Š Calcul d'angles entre vecteurs\n")

couples = [
    ((1, 0), (1, 1), "45Â° attendu"),
    ((1, 0), (0, 1), "90Â° attendu"),
    ((1, 0), (-1, 0), "180Â° attendu"),
    ((1, 1), (1, 1), "0Â° attendu"),
    ((3, 4), (-4, 3), "90Â° attendu")
]

for u, v, description in couples:
    angle = angle_degres(u, v)
    print(f"{u} et {v}")
    print(f"  Angle : {angle:.1f}Â° ({description})\n")
````
