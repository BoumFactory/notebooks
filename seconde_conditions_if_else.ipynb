## üìö Notion 1 : Les op√©rateurs de comparaison

### üìñ Cours

En programmation, on a souvent besoin de **comparer** des valeurs. Python utilise des **op√©rateurs de comparaison** qui renvoient `True` (vrai) ou `False` (faux) :

| Op√©rateur | Signification | Exemple |
|-----------|---------------|----------|
| `==` | √âgal √† | `x == 5` |
| `!=` | Diff√©rent de | `x != 0` |
| `<` | Strictement inf√©rieur | `x < 10` |
| `>` | Strictement sup√©rieur | `x > 0` |
| `<=` | Inf√©rieur ou √©gal | `x <= 5` |
| `>=` | Sup√©rieur ou √©gal | `x >= -3` |

‚ö†Ô∏è **Attention** : Pour tester l'√©galit√©, on utilise `==` (deux signes √©gal), **pas** `=` !

**Exemple r√©solu :**

````python
# Comparer des nombres
x = 7

print(x == 7)   # True (vrai)
print(x == 5)   # False (faux)
print(x > 5)    # True
print(x < 3)    # False
````

## üìö Notion 2 : La structure if

### üìñ Cours

La structure `if` permet d'**ex√©cuter un bloc de code** seulement si une condition est vraie.

**Syntaxe :**
```python

    # Instructions (avec indentation obligatoire !)
```

‚ö†Ô∏è **Important** : L'indentation (4 espaces) est **obligatoire** en Python !

**Exemple r√©solu :**

````python
# Afficher un message si le nombre est positif
x = 12

if x > 0:
    print(f"{x} est positif")
````

````python
# Si la condition est fausse, rien ne s'affiche
x = -5

if x > 0:
    print(f"{x} est positif")

print("Fin du programme")
````

Pour ex√©cuter une instruction quand la condition est **fausse**, on utilise `else` (sinon).

**Syntaxe :**

````python
# Tester la parit√© d'un nombre
n = 8

if n % 2 == 0:
    print(f"{n} est pair")
else:
    print(f"{n} est impair")
````

## üìö Notion 4 : La structure if/elif/else

### üìñ Cours

Pour tester **plusieurs conditions**, on utilise `elif` (abr√©viation de "else if" = "sinon si").

**Syntaxe :**
```python
if condition1:
    # Si condition1 est vraie
elif condition2:
    # Si condition1 fausse ET condition2 vraie
elif condition3:
    # Si condition1 et 2 fausses ET condition3 vraie
else:
    # Si toutes les conditions sont fausses
```

‚ö†Ô∏è **Important** : D√®s qu'une condition est vraie, les autres ne sont **pas test√©es** !

**Exemple r√©solu :**

````python
# D√©terminer le signe d'un nombre (3 cas)
x = 0

if x > 0:
    print(f"{x} est strictement positif")
elif x < 0:
    print(f"{x} est strictement n√©gatif")
else:
    print(f"{x} est nul")
````

````python
# Mention selon la note
note = 15

if note >= 16:
    print("Tr√®s bien")
elif note >= 14:
    print("Bien")
elif note >= 12:
    print("Assez bien")
elif note >= 10:
    print("Passable")
else:
    print("Insuffisant")
````

## üìö Notion 5 : Combiner des conditions

### üìñ Cours

On peut **combiner** plusieurs tests avec les op√©rateurs logiques :

| Op√©rateur | Signification | Exemple |
|-----------|---------------|----------|
| `and` | ET (toutes les conditions vraies) | `x > 0 and x < 10` |
| `or` | OU (au moins une vraie) | `x == 0 or x == 1` |
| `not` | NON (inverse la condition) | `not (x > 0)` |

**Exemple r√©solu :**

````python
# Tester si un nombre est dans un intervalle
x = 7

if x >= 5 and x <= 10:
    print(f"{x} est dans [5, 10]")
else:
    print(f"{x} n'est pas dans [5, 10]")
````

````python
# Tester si divisible par 2 OU par 3
n = 15

if n % 2 == 0 or n % 3 == 0:
    print(f"{n} est divisible par 2 ou par 3")
else:
    print(f"{n} n'est divisible ni par 2 ni par 3")
````

````python
# Tester si divisible par 2 ET par 3 (donc par 6)
n = 12

if n % 2 == 0 and n % 3 == 0:
    print(f"{n} est divisible par 6")
````
