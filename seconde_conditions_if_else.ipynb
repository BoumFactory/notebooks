## üìö Notion 1 : Comparaisons et structure if/else

### üìñ Cours

**Op√©rateurs de comparaison** (renvoient `True` ou `False`) :

| Op√©rateur | Signification |
|-----------|---------------|
| `==` | √âgal √† |
| `!=` | Diff√©rent de |
| `<`, `>` | Strictement inf√©rieur/sup√©rieur |
| `<=`, `>=` | Inf√©rieur/sup√©rieur ou √©gal |

‚ö†Ô∏è **Attention** : `==` pour comparer, `=` pour affecter !

**Structure conditionnelle** :
```python
if condition:
    # Si condition vraie
else:
    # Si condition fausse
```

‚ö†Ô∏è **L'indentation (d√©calage) est obligatoire !**

**Exemples :**

````python
# Comparaisons simples
x = 7
print(x > 5)   # True
print(x == 10) # False
print(x != 0)  # True
````

````python
# Tester la parit√© d'un nombre
n = 8

if n % 2 == 0:
    print(n, "est pair")
else:
    print(n, "est impair")
````

````python
# Tester le signe d'un nombre (3 cas avec elif)
x = -5

if x > 0:
    print("Positif")
elif x < 0:
    print("N√©gatif")
else:
    print("Nul")
````

## üìö Notion 2 : Combiner des conditions

### üìñ Cours

On peut combiner plusieurs tests avec :

| Op√©rateur | Signification |
|-----------|---------------|
| `and` | ET (toutes vraies) |
| `or` | OU (au moins une vraie) |

**Exemples :**

````python
# Tester si x est dans l'intervalle [5, 10]
x = 7

if x >= 5 and x <= 10:
    print(x, "est dans [5, 10]")
else:
    print(x, "n'est pas dans [5, 10]")
````

````python
# Divisible par 2 ou par 3 ?
n = 15

if n % 2 == 0 or n % 3 == 0:
    print(n, "est divisible par 2 ou par 3")
````
