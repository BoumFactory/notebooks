## üí° Introduction

Le **taux d'accroissement** mesure la variation d'une fonction entre deux points.

<div style="background-color: #f0f4c3; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ü§î Probl√®me :</strong> Comment mesurer la pente d'une courbe en un point ?<br>
    <strong>‚ú® Solution :</strong> Le taux d'accroissement approche le <strong>nombre d√©riv√©</strong> quand h‚Üí0 !
</div>

**Formule :** $\text{Taux} = \dfrac{f(a+h) - f(a)}{h}$

---

## üìö Notion 1 : Calculer le taux d'accroissement

### üìñ Cours

Le taux d'accroissement entre $a$ et $a+h$ est : $\tau = \dfrac{f(a+h) - f(a)}{h}$

**En Python :**
```python
def taux(f, a, h):
    return (f(a+h) - f(a)) / h
```

**Exemple r√©solu :** Pour $f(x) = x^2$, $a=3$, $h=1$

````python
def f(x):
    return x**2

def taux(f, a, h):
    return (f(a+h) - f(a)) / h

a, h = 3, 1
resultat = taux(f, a, h)
print(f"Taux entre x={a} et x={a+h} : {resultat}")
````

## üìö Notion 2 : Approcher le nombre d√©riv√©

### üìñ Cours

Le **nombre d√©riv√©** $f'(a)$ est la limite du taux quand $h \to 0$ :

$$f'(a) = \lim_{h \to 0} \dfrac{f(a+h) - f(a)}{h}$$

**Approximation :** Prendre $h$ tr√®s petit (ex: $h=10^{-6}$).

**Exemple r√©solu :** Approcher $f'(3)$ pour $f(x)=x^2$

````python
def f(x):
    return x**2

def derive_approchee(f, a):
    h = 1e-6
    return (f(a+h) - f(a)) / h

a = 3
derivee = derive_approchee(f, a)
print(f"f'({a}) ‚âà {derivee:.6f}")
print(f"Th√©orique : 2√ó{a} = {2*a}")
````

## üìö Notion 3 : √âquation de la tangente

### üìñ Cours

La **tangente** √† la courbe de f en x=a a pour √©quation :

$$y = f(a) + f'(a)(x - a)$$

**M√©thode Python :**
1. Calculer f(a)
2. Approcher f'(a)
3. Construire l'√©quation

**Exemple r√©solu :** Tangente √† f(x)=x¬≤ en x=3

````python
def f(x):
    return x**2

def tangente(f, a, x):
    """Calcule y pour la tangente en x=a"""
    h = 1e-6
    f_a = f(a)
    f_prime_a = (f(a+h) - f(a)) / h
    return f_a + f_prime_a * (x - a)

a = 3
print(f"Point de tangence : ({a}, {f(a)})")
print(f"Pente : {(f(a+1e-6)-f(a))/1e-6:.2f}")
print(f"En x=4 : y = {tangente(f, a, 4)}")
````

## üìö Notion 4 : Taux pour diff√©rentes fonctions

### üìñ Cours

Le calcul du taux s'applique √† **toutes les fonctions** :
- Polyn√¥mes : $x^n$
- Racine : $\sqrt{x}$
- Inverse : $1/x$
- Exponentielles : $e^x$
- Trigonom√©triques : $\sin(x)$, $\cos(x)$

**Exemple r√©solu :** Comparer les d√©riv√©es

````python
import math

def derive(f, a):
    h = 1e-6
    return (f(a+h) - f(a)) / h

# Diff√©rentes fonctions
fonctions = [
    (lambda x: x**2, "x¬≤", "2x"),
    (lambda x: x**3, "x¬≥", "3x¬≤"),
    (lambda x: math.sqrt(x), "‚àöx", "1/(2‚àöx)"),
    (lambda x: math.sin(x), "sin(x)", "cos(x)")
]

a = 2
print(f"D√©riv√©es en x={a} :\n")
for f, nom, formule in fonctions:
    d = derive(f, a)
    print(f"{nom:8} : f'({a}) ‚âà {d:.6f} ({formule})")
````

## üìö Notion 5 : Visualisation graphique

### üìñ Cours

Tracer la courbe et sa tangente permet de **visualiser** le nombre d√©riv√©.

**M√©thode :**
1. Tracer la courbe de f
2. Calculer la tangente en un point
3. Tracer la tangente

**Exemple r√©solu :** f(x)=x¬≤, tangente en x=2

````python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2

# Point de tangence
a = 2
h = 1e-6
f_a = f(a)
f_prime_a = (f(a+h) - f(a)) / h

# Courbe
x = np.linspace(0, 4, 200)
y = f(x)

# Tangente
x_tang = np.linspace(0, 4, 100)
y_tang = f_a + f_prime_a * (x_tang - a)

# Trac√©
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x)=x¬≤')
plt.plot(x_tang, y_tang, 'r--', linewidth=2, label=f'Tangente (pente={f_prime_a:.0f})')
plt.scatter([a], [f_a], color='green', s=150, zorder=5)
plt.title("Courbe et tangente")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````
