## üí° Introduction

Le **taux d'accroissement** mesure la variation d'une fonction entre deux points.

<div style="background-color: #f0f4c3; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ü§î Probl√®me :</strong> Comment mesurer la pente d'une courbe en un point ?<br>
    <strong>‚ú® Solution :</strong> Le taux d'accroissement approche le <strong>nombre d√©riv√©</strong> quand h‚Üí0 !
</div>

**Formule du taux d'accroissement :** $\tau = \dfrac{f(a+h) - f(a)}{h}$

---

## üìö Notion 1 : Calculer le taux d'accroissement

### üìñ Cours

Le **taux d'accroissement** de $f$ entre $a$ et $a+h$ est d√©fini par :

$$\tau = \dfrac{f(a+h) - f(a)}{h}$$

- **G√©om√©triquement** : c'est le coefficient directeur de la droite passant par les points $(a, f(a))$ et $(a+h, f(a+h))$
- **Physiquement** : c'est une vitesse moyenne

**Exemple :** Pour $f(x) = x^2$, calculons le taux entre $x=3$ et $x=4$ (donc $a=3$, $h=1$) :

$$\tau = \dfrac{f(4) - f(3)}{1} = \dfrac{16 - 9}{1} = 7$$

````python
# Exemple : taux d'accroissement de f(x)=x¬≤ entre x=3 et x=4
def f(x):
    return x**2

a = 3
h = 1
taux = (f(a + h) - f(a)) / h

print(f"f({a}) = {f(a)}")
print(f"f({a+h}) = {f(a+h)}")
print(f"Taux d'accroissement = {taux}")
````

## üìö Notion 2 : Approcher le nombre d√©riv√©

### üìñ Cours

Le **nombre d√©riv√©** $f'(a)$ est la limite du taux d'accroissement quand $h \to 0$ :

$$f'(a) = \lim_{h \to 0} \dfrac{f(a+h) - f(a)}{h}$$

**Interpr√©tation :** C'est la pente de la tangente √† la courbe au point d'abscisse $a$.

**Approximation num√©rique :** On prend $h$ tr√®s petit (par exemple $h = 10^{-6}$) pour approcher $f'(a)$.

**Exemple :** Pour $f(x) = x^2$, on sait que $f'(x) = 2x$, donc $f'(3) = 6$. V√©rifions num√©riquement :

````python
# Convergence du taux vers le nombre d√©riv√©
def f(x):
    return x**2

a = 3
valeurs_h = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]

print("Convergence vers f'(3) = 6 :")
print()
for h in valeurs_h:
    taux = (f(a + h) - f(a)) / h
    print(f"h = {h:8.5f}  ‚Üí  taux = {taux:.6f}")
````

## üìö Notion 3 : √âquation de la tangente

### üìñ Cours

La **tangente** √† la courbe de $f$ au point d'abscisse $a$ a pour √©quation :

$$y = f(a) + f'(a)(x - a)$$

**Interpr√©tation :**
- Elle passe par le point $(a, f(a))$
- Elle a pour pente $f'(a)$

**Exemple :** Pour $f(x) = x^2$ en $a = 3$ :
- $f(3) = 9$
- $f'(3) = 6$
- √âquation : $y = 9 + 6(x - 3) = 6x - 9$

````python
# Exemple : √©quation de la tangente √† f(x)=x¬≤ en x=3
def f(x):
    return x**2

a = 3
f_a = f(a)  # = 9
h = 1e-6 # tr√®s petit 1x10^(-6)
f_prime_a = (f(a + h) - f(a)) / h  

print(f"Point de tangence : ({a}, {f_a})")
print(f"Pente : {f_prime_a:.2f}")
print(f"√âquation : y = {f_prime_a:.0f}x - {f_prime_a*a - f_a:.0f}")

# V√©rification : la tangente passe bien par (3, 9)
y_tang_en_3 = f_a + f_prime_a * (3 - a)
print(f"V√©rification : en x=3, y_tangente = {y_tang_en_3}")
````

## üìö Notion 5 : Visualisation graphique

### üìñ Cours

Tracer la courbe et sa tangente permet de **visualiser** le nombre d√©riv√©.

**M√©thode :**
1. Cr√©er un tableau de valeurs pour la courbe
2. Calculer $f(a)$ et $f'(a)$ pour la tangente
3. Calculer les ordonn√©es de la tangente : $y = f(a) + f'(a)(x - a)$
4. Tracer les deux courbes avec `matplotlib`

````python
# Exemple : courbe de f(x)=x¬≤ et tangente en x=2
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2

# Point de tangence
a = 2
h = 1e-6
f_a = f(a)  # = 4
f_prime_a = (f(a + h) - f(a)) / h  # ‚âà 4

# Points pour la courbe
x = np.linspace(0, 4, 200)
y_courbe = f(x)

# Points pour la tangente
y_tangente = f_a + f_prime_a * (x - a)

# Trac√©
plt.figure(figsize=(10, 6))
plt.plot(x, y_courbe, 'b-', linewidth=2, label='f(x) = x¬≤')
plt.plot(x, y_tangente, 'r--', linewidth=2, label=f'Tangente en x={a} (pente={f_prime_a:.0f})')
plt.scatter([a], [f_a], color='green', s=150, zorder=5, label=f'Point ({a}, {f_a})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Courbe et tangente')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-2, 16)
plt.show()
````

# 1. Fonction √©nergie
def E(x):
    return 0.5 * k * x**2 + m * g * x

def derive(f, a):
    h = 1e-6
    return (f(a + h) - f(a)) / h

# 2. Calcul des d√©riv√©es

    print(f"x = {x:6.2f} m : E'(x) = {d:8.2f}")
    if abs(d) < 1:  # Proche de 0
        x_equilibre = x
