## ğŸ’¡ Introduction

En mathÃ©matiques, le **taux d'accroissement** mesure la variation moyenne d'une fonction entre deux points.

Pour une fonction $f$ et deux points $a$ et $a+h$ :

$$\text{Taux d'accroissement} = \dfrac{f(a+h) - f(a)}{h}$$

<div style="background-color: #f0f4c3; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ¤” ProblÃ¨me :</strong> Comment mesurer la "vitesse" Ã  laquelle une fonction varie en un point ?<br>
    <strong>âœ¨ Solution :</strong> En calculant le taux d'accroissement et en faisant tendre $h$ vers 0, on obtient le <strong>nombre dÃ©rivÃ©</strong> !
</div>

### ğŸ” InterprÃ©tation gÃ©omÃ©trique

Le taux d'accroissement reprÃ©sente la **pente de la corde** reliant les points $(a, f(a))$ et $(a+h, f(a+h))$.

Quand $h \to 0$, cette corde devient la **tangente** Ã  la courbe en $a$.

## 1ï¸âƒ£ Calculer un taux d'accroissement

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– DÃ©finition :</strong> Le taux d'accroissement de $f$ entre $a$ et $a+h$ est :
    $$\tau_h = \dfrac{f(a+h) - f(a)}{h}$$
</div>

### ğŸ”§ Formule Python

```python
def taux_accroissement(f, a, h):
    return (f(a + h) - f(a)) / h
```

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Taux d'accroissement de f(x) = xÂ² en a = 3 avec h = 1

def f(x):
    return x**2

def taux_accroissement(f, a, h):
    return (f(a + h) - f(a)) / h

# Calcul
a = 3
h = 1
tau = taux_accroissement(f, a, h)

print(f"Fonction : f(x) = xÂ²")
print(f"Point : a = {a}")
print(f"Pas : h = {h}")
print(f"\nf({a}) = {f(a)}")
print(f"f({a + h}) = {f(a + h)}")
print(f"\nTaux d'accroissement : Ï„ = {tau}")
print(f"\nInterprÃ©tation : Entre x=3 et x=4, f varie en moyenne de {tau} par unitÃ©")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li>$f(3) = 9$ et $f(4) = 16$</li>
        <li>Taux = $(16 - 9) / 1 = 7$</li>
        <li>La fonction xÂ² augmente en moyenne de 7 entre x=3 et x=4</li>
    </ul>
</div>

## 3ï¸âƒ£ Approcher le nombre dÃ©rivÃ©

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– DÃ©finition :</strong> Le nombre dÃ©rivÃ© de $f$ en $a$ est la limite du taux d'accroissement quand $h \to 0$ :
    $$f'(a) = \lim_{h \to 0} \dfrac{f(a+h) - f(a)}{h}$$
</div>

### ğŸ¯ Approximation numÃ©rique

En pratique, on calcule le taux pour un $h$ trÃ¨s petit (par exemple $h = 10^{-6}$).

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Approximer f'(a) pour f(x) = xÂ² - 3x + 2

def f(x):
    return x**2 - 3*x + 2

def nombre_derive_approche(f, a):
    """Approxime f'(a) avec h trÃ¨s petit"""
    h = 1e-6  # h = 0.000001
    return (f(a + h) - f(a)) / h

# Test en plusieurs points
points = [0, 1, 2, 3]

print("ğŸ“ Approximation de f'(a) pour f(x) = xÂ² - 3x + 2\n")
print("a  | f'(a) approchÃ© | f'(a) exact (2a - 3)")
print("-" * 45)

for a in points:
    derive_approchee = nombre_derive_approche(f, a)
    derive_exacte = 2*a - 3  # Formule thÃ©orique
    print(f"{a}  | {derive_approchee:14.6f} | {derive_exacte:6}")

print("\nâœ… L'approximation est trÃ¨s prÃ©cise !")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li>Pour $f(x) = x^2 - 3x + 2$, on a $f'(x) = 2x - 3$ (formule thÃ©orique)</li>
        <li>L'approximation avec $h = 10^{-6}$ donne des rÃ©sultats trÃ¨s proches</li>
        <li>Python peut ainsi approcher le nombre dÃ©rivÃ© mÃªme sans formule !</li>
    </ul>
</div>

## 4ï¸âƒ£ Visualiser gÃ©omÃ©triquement

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 10px; border-left: 5px solid #4CAF50;">
    <strong>ğŸ“– Rappel :</strong> Le taux d'accroissement est la pente de la corde. Quand $h \to 0$, la corde tend vers la tangente.
</div>

### ğŸ“ Exemple complet rÃ©solu

````python
# âœ… EXEMPLE RÃ‰SOLU : Visualiser avec matplotlib
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2

# ParamÃ¨tres
a = 2
h = 1.5

# Points
x = np.linspace(0, 5, 200)
y = f(x)

# Calcul du taux (pente de la corde)
pente = (f(a + h) - f(a)) / h

# Ã‰quation de la corde : y - f(a) = pente Ã— (x - a)
x_corde = np.linspace(a - 0.5, a + h + 0.5, 100)
y_corde = f(a) + pente * (x_corde - a)

# TracÃ©
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = xÂ²')
plt.plot(x_corde, y_corde, 'r--', linewidth=2, label=f'Corde (pente = {pente:.2f})')
plt.scatter([a, a+h], [f(a), f(a+h)], color='red', s=100, zorder=5)
plt.scatter([a], [f(a)], color='green', s=150, zorder=5, label=f'Point a = {a}')

plt.title("Taux d'accroissement = Pente de la corde", fontsize=14)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(f"Taux d'accroissement entre a={a} et a+h={a+h} : {pente:.4f}")
````

<div style="background-color: #e1f5fe; padding: 10px; border-radius: 5px; margin: 10px 0;">
    <strong>ğŸ’¬ Explication :</strong>
    <ul>
        <li>La droite rouge relie les points $(a, f(a))$ et $(a+h, f(a+h))$</li>
        <li>Sa pente est exactement le taux d'accroissement</li>
        <li>Si on diminue $h$, la droite se rapproche de la tangente</li>
    </ul>
</div>
